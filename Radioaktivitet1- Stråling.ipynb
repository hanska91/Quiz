```python
import ipywidgets as widgets
from IPython.display import display, HTML, Javascript, clear_output, display
import random
import matplotlib.pyplot as plt

# --- CSS: FIXED GOD TIER ---
custom_css = """
<style>
    /* --- 1. LEVENDE BAKGRUNN --- */
    .master-container {
        background: linear-gradient(-45deg, #ee7752, #e73c7e, #23a6d5, #23d5ab);
        background-size: 400% 400%;
        animation: gradientBG 15s ease infinite;
        padding: 50px;
        border-radius: 20px;
        font-family: 'Inter', 'Segoe UI', Helvetica, Arial, sans-serif;
    }

    @keyframes gradientBG {
        0% { background-position: 0% 50%; }
        50% { background-position: 100% 50%; }
        100% { background-position: 0% 50%; }
    }

    /* --- 2. GLASS-KORTET --- */
    .quiz-card-container {
        background-color: rgba(255, 255, 255, 0.92);
        backdrop-filter: blur(15px);
        border-radius: 24px;
        box-shadow: 0 25px 80px rgba(0,0,0,0.3);
        width: 100%;
        margin: 0 auto;
        overflow: visible; /* VIKTIG: Lar konfettien flyte fritt */
        border: 1px solid rgba(255, 255, 255, 0.6);
        position: relative;
    }

    /* Header - FIKSET TEKSTKUTT */
    .quiz-header-widget {
        background: transparent;
        padding: 30px 0 10px 0;
        text-align: center;
        width: 100%;
        border-bottom: 2px solid rgba(0,0,0,0.05);
    }
    
    .quiz-title {
        margin: 0;
        font-size: 48px;
        font-weight: 900;
        letter-spacing: 5px;
        text-transform: uppercase;
        
        /* Gradient tekst */
        background: linear-gradient(to right, #2563eb, #9333ea);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        
        /* FIKS FOR KUTTET TEKST: Mer høyde og padding */
        line-height: 1.4; 
        padding-bottom: 10px; 
        display: inline-block; /* Sikrer at boksen omslutter teksten korrekt */
    }

    /* Innhold */
    .quiz-content-area {
        padding: 20px 50px 50px 50px;
        position: relative;
        z-index: 1;
    }

    /* Animasjoner */
    .slide-in-up {
        animation: slideUp 0.6s cubic-bezier(0.16, 1, 0.3, 1) forwards;
        opacity: 0;
        transform: translateY(20px);
    }
    @keyframes slideUp { to { opacity: 1; transform: translateY(0); } }

    /* Progresjon */
    .progress-wrapper {
        margin-top: 15px;
        margin-bottom: 40px;
    }
    .progress-bg {
        width: 100%;
        background: #e5e7eb;
        height: 12px;
        border-radius: 6px;
    }
    .progress-fill {
        background: linear-gradient(90deg, #2563eb, #9333ea);
        height: 12px;
        border-radius: 6px;
        box-shadow: 0 0 15px rgba(147, 51, 234, 0.4);
        transition: width 0.6s cubic-bezier(0.4, 0, 0.2, 1);
    }

    /* Spørsmål */
    .question-text {
        font-size: 28px;
        color: #1f2937;
        font-weight: 800;
        margin-bottom: 35px;
        line-height: 1.3;
        text-align: center;
    }

    /* Knapper */
    .widget-button { 
        box-sizing: border-box !important;
        font-size: 20px !important; 
        border-radius: 16px !important; 
        border: 2px solid #e5e7eb !important; 
        background-color: white !important;
        color: #4b5563 !important;
        margin: 10px 0 !important; 
        padding: 22px 30px !important;
        height: auto !important;
        width: 98% !important; 
        margin-left: 1% !important;
        text-align: left !important;
        outline: none !important;
        box-shadow: 0 4px 6px rgba(0,0,0,0.03) !important;
        transition: all 0.3s cubic-bezier(0.25, 0.8, 0.25, 1) !important;
    }
    
    .widget-button:hover:enabled {
        border-color: #8b5cf6 !important;
        transform: translateY(-4px) scale(1.005);
        box-shadow: 0 15px 30px rgba(139, 92, 246, 0.2) !important;
        z-index: 10;
    }

    /* Feedback Box */
    .feedback-box {
        margin-top: 30px;
        padding: 30px;
        border-radius: 16px;
        background-color: #f9fafb;
        border-left: 8px solid #ccc;
        font-size: 18px;
        line-height: 1.6;
        animation: slideUp 0.4s ease-out;
    }
    
    /* Resultat Melding */
    .result-msg {
        font-size: 24px;
        margin-top: 25px;
        font-weight: 700;
        padding: 30px;
        border-radius: 20px;
        text-align: center;
        box-shadow: 0 10px 30px rgba(0,0,0,0.1);
        position: relative;
        z-index: 20; /* Foran konfetti */
    }

    /* --- KONFETTI MOTOR (FIKSET) --- */
    .confetti-container {
        position: absolute;
        top: 0;
        left: 0;
        width: 100%;
        height: 100%;
        pointer-events: none; /* Lar deg klikke gjennom konfettien */
        z-index: 100; /* Høyere enn alt annet */
        overflow: hidden;
    }
    
    .confetti {
        position: absolute;
        width: 12px; 
        height: 12px;
        opacity: 0.8;
        top: -20px; /* Starter over kanten */
        animation: fall linear forwards;
    }
    
    @keyframes fall {
        0% { transform: translateY(-20px) rotate(0deg); opacity: 1;}
        100% { transform: translateY(600px) rotate(720deg); opacity: 0; }
    }
    
    /* Styling for Flash/Highlight */
    .correct-flash { animation: flash-success 0.6s forwards !important; font-weight: bold !important; z-index: 20 !important; }
    @keyframes flash-success {
        0% { background-color: white; border-color: #10b981; }
        50% { background-color: #10b981; border-color: #10b981; color: white; transform: scale(1.02); }
        100% { background-color: #10b981; border-color: #10b981; color: white; }
    }
    .wrong-answer { background-color: #ef4444 !important; color: white !important; border-color: #ef4444 !important; }
    .fasit-highlight { background-color: #ecfdf5 !important; color: #047857 !important; border-color: #10b981 !important; border-width: 3px !important; font-weight: bold !important; }

</style>
"""

# --- DATA ---
original_quiz_data = [
        {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet.",
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling.",
        },
        {
            "question": "Hva er gammastråling?",
            "options": ["A. Partikkelstråling", "B. Synlig lys", "C. Ultrafiolett stråling", "D. Svært energirik elektromagnetisk stråling"],
            "answer": "D",
            "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner.",
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling med lav energi", "B. Stråling som ikke påvirker atomer", "C. Stråling som består av partikler", "D. Stråling som har nok energi til å fjerne elektroner fra et atom eller molekyl"],
            "answer": "D",
            "explanation": "Ioniserende stråling har nok energi til å fjerne elektroner fra et atom eller molekyl.",
        },
        {
            "question": "Hva er rødforskyvning?",
            "options": ["A. Når lysbølger stopper helt opp", "B. Når lysbølger blir kortere og bølgelengden minker", "C. Når lysbølger endrer farge til blått", "D. Når lysbølger blir strukket ut og bølgelengden øker"],
            "answer": "D",
            "explanation": "Rødforskyvning skjer når lysbølger blir strukket ut og bølgelengden øker.",
        },
        {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet."
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling."
        },
        {
            "question": "Hva er gammastråling?",
            "options": ["A. Partikkelstråling", "B. Synlig lys", "C. Ultrafiolett stråling", "D. Svært energirik elektromagnetisk stråling"],
            "answer": "D",
            "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner."
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling med lav energi", "B. Stråling som ikke påvirker atomer", "C. Stråling som består av partikler", "D. Stråling som har nok energi til å fjerne elektroner fra et atom eller molekyl"],
            "answer": "D",
             "explanation": "Ioniserende stråling har nok energi til å fjerne elektroner fra et atom eller molekyl."
        },
        {
            "question": "Hva er rødforskyvning?",
            "options": ["A. Når lysbølger stopper helt opp", "B. Når lysbølger blir kortere og bølgelengden minker", "C. Når lysbølger endrer farge til blått", "D. Når lysbølger blir strukket ut og bølgelengden øker"],
            "answer": "D",
             "explanation": "Rødforskyvning skjer når lysbølger blir strukket ut og bølgelengden øker."
        },
        {
            "question": "Hva er et foton?",
            "options": ["A. En type atom", "B. En bølge", "C. En lyspartikkel som frakter energi", "D. En partikkel med masse"],
            "answer": "C",
            "explanation": "Et foton er en lyspartikkel som frakter energi."
        },
        {
            "question": "Hva slags stråling sender den radioaktive kilden ut hvis ikke all strålingen blir stanset av papir, men derimot både av en treplate og en blyplate?",
            "options": ["A. Alfastråling", "B. Betastråling", "C. Gammastråling"],
            "answer": "B",
            "explanation": "Betastråling blir stanset av en treplate, men ikke av papir."
        },
        {
            "question": "Hvis halveringstiden for Ce-137 er 30 år, hvilken strålingsaktivitet hadde kilden for 90 år siden hvis den nå er 1000 Bq?",
            "options": ["A. 500 Bq", "B. 1000 Bq", "C. 2000 Bq", "D. 4000 Bq", "E. 8000 Bq"],
            "answer": "E",
            "explanation": "Etter 90 år, som er tre halveringstider, vil aktiviteten ha blitt redusert til en åttendedel av den opprinnelige aktiviteten, så den opprinnelige aktiviteten var 8000 Bq."
        },
        {
            "question": "Hvor gammelt er et lårbein fra en moskusokse hvis innholdet av isotopen karbon-14 bare er om lag 25 % av det vi finner i en levende moskusokse? Karbon-14 har en halveringstid på 5730 år.",
            "options": ["A. 1432 år", "B. 5730 år", "C. 7163 år", "D. 11 460 år", "E. 17 190 år"],
            "answer": "D",
            "explanation": "Hvis innholdet av karbon-14 er redusert til 25 %, betyr det at to halveringstider har gått, så lårbeinet er 11 460 år gammelt."
        },
        {
        "question": "Når Thorium-234 sender ut en betapartikkel, vil atomnummeret",
        "options": ["A. Ikke endres", "B. Økes med en", "C. Minkes med en", "D. Minkes med to", "E. Minkes med tre"],
        "answer": "B",
        "explanation": "Når Thorium-234 sender ut en betapartikkel, omdannes en nøytron til et proton, noe som øker atomnummeret med en."
    },
    {
        "question": "Når Radon-222 sender ut en alfapartikkel, vil atomnummeret",
        "options": ["A. Minskes med to", "B. Minskes med fire", "C. Økes med to", "D. Økes med fire", "E. Økes med tre"],
        "answer": "A",
        "explanation": "Når Radon-222 sender ut en alfapartikkel, mister det to protoner og to nøytroner, noe som reduserer atomnummeret med to."
    },
    {
        "question": "Når Radon-222 sender ut gammastråling, vil atomnummeret",
        "options": ["A. Økes med en", "B. Minkes med en", "C. Minkes med to", "D. Ikke endres", "E. Økes med tre"],
        "answer": "D",
        "explanation": "Gammastråling endrer ikke antall protoner eller nøytroner i kjernen, så atomnummeret forblir uendret."
    },
    {
        "question": "Hva kan man si om galaksene hvis lyset fra de fleste galakser er rødforskjøvet?",
        "options": ["A. Beveger seg mot oss", "B. Beveger seg fra oss", "C. Beveger seg med samme hastighet som oss", "D. Står stille i forhold til oss"],
        "answer": "B",
        "explanation": "Rødforskyvning indikerer at galaksene beveger seg bort fra oss, noe som støtter teorien om at universet utvider seg."
    },
    {
        "question": "Hva bestemmer protoner og nøytroner i atomene?",
        "options": ["A. Protoner bestemmer atomnummeret, nøytroner bestemmer isotopen", "B. Protoner bestemmer massen, nøytroner bestemmer ladningen", "C. Protoner bestemmer ladningen, nøytroner bestemmer massen", "D. Protoner bestemmer isotopen, nøytroner bestemmer atomnummeret"],
        "answer": "A",
        "explanation": "Antall protoner i kjernen bestemmer atomnummeret og dermed hvilket grunnstoff det er. Antall nøytroner bestemmer isotopen av grunnstoffet."
    },
    {
        "question": "Hva er halveringstiden for et radioaktivt stoff som reduseres til 12,5 % av den opprinnelige massen i løpet av en time?",
        "options": ["A. 15 minutter", "B. 20 minutter", "C. 30 minutter", "D. 40 minutter"],
        "answer": "B",
        "explanation": "Hvis et stoff reduseres til 12,5 % av den opprinnelige massen, har det gjennomgått tre halveringstider (100 % -> 50 % -> 25 % -> 12,5 %). En time delt på tre gir en halveringstid på 20 minutter."
    },
    {
        "question": "Hva er Radon?",
        "options": ["A. En edelgass", "B. En metall", "C. En halvmetall", "D. En ikke-metall"],
        "answer": "A",
        "explanation": "Radon er en edelgass som er radioaktiv og finnes naturlig i jordskorpen."
    },
    {
        "question": "Hva brukes PET-scan til?",
        "options": ["A. Radioaktive markører som enten må bli svelget, inhalert eller injeksert i kroppen og kan derfor viser hvordan ulike vev eller organer i kroppen jobber på ett celle-nivå. Brukes for å diagnostisere kreft, hjerteproblemer eller hjerneproblemer", "B. PET-scan brukes for å måle PET(dyret på engelsk)-en din, altså katten, hunden eller hamsteren din", "C. Å måle blodtrykk", "D. Å måle blodsukker"],
        "answer": "A",
        "explanation": "PET-scan bruker radioaktive markører for å vise hvordan vev og organer fungerer på cellenivå, og brukes til å diagnostisere sykdommer som kreft, hjerteproblemer og hjerneproblemer."
    },
    {
        "question": "Hva brukes CT-scan til?",
        "options": ["A. Røntgenundersøkelse der det blir tatt tverrsnittbilde av aktuelle områder i kroppen (bløtvev, blodårer, organer, bein og vev)", "B. Viser hvordan ulike vev eller organer i kroppen jobber på ett celle-nivå", "C. Å måle blodtrykk", "D. Å måle blodsukker"],
        "answer": "A",
        "explanation": "CT-scan er en røntgenundersøkelse som tar tverrsnittbilder av kroppen, inkludert bløtvev, blodårer, organer, bein og vev."
    },
    {
        "question": "Hva er kosmisk bakgrunnstråling?",
        "options": ["A. Stråling fra solen", "B. Stråling fra jorden", "C. Stråling fra universets begynnelse", "D. Stråling fra stjerner"],
        "answer": "C",
        "explanation": "Kosmisk bakgrunnstråling er stråling som stammer fra universets begynnelse og er en viktig observasjon som støtter Big Bang-teorien."
    },
    {
        "question": "Hva er Big Bang teorien?",
        "options": ["A. Teorien om at universet alltid har eksistert", "B. Teorien om at universet ble skapt av en stor eksplosjon", "C. Teorien om at universet vil ende i en stor eksplosjon", "D. Teorien om at universet er uendelig"],
        "answer": "B",
        "explanation": "Big Bang-teorien foreslår at universet ble skapt av en stor eksplosjon for omtrent 13,8 milliarder år siden."
    },
    {
        "question": "Hvilke observasjoner støtter Big Bang teorien?",
        "options": ["A. Rødforskyvning av galakser, kosmisk bakgrunnstråling, fordeling av grunnstoffer", "B. Blåforskyvning av galakser, kosmisk bakgrunnstråling, fordeling av grunnstoffer", "C. Rødforskyvning av galakser, stråling fra solen, fordeling av grunnstoffer", "D. Blåforskyvning av galakser, stråling fra solen, fordeling av grunnstoffer"],
        "answer": "A",
        "explanation": "Rødforskyvning av galakser, kosmisk bakgrunnstråling og fordelingen av grunnstoffer i universet er observasjoner som støtter Big Bang-teorien."
    },
    {
        "question": "Hva er halveringstiden til karbon-14?",
        "options": ["A. 5730 år", "B. 5000 år", "C. 6000 år", "D. 7000 år"],
        "answer": "A",
        "explanation": "Halveringstiden til karbon-14 er 5730 år, noe som betyr at det tar 5730 år for halvparten av karbon-14 i en prøve å brytes ned."
    },
    {
        "question": "Hvorfor skal vi være redde for radon i huset vårt?",
        "options": [
            "A. Radon er en edelgass som kan forårsake eksplosjoner",
            "B. Radon er en radioaktiv gass som kan øke risikoen for lungekreft",
            "C. Radon kan føre til elektriske feil i huset",
            "D. Radon kan forårsake vannforurensning"
        ],
        "answer": "B",
        "explanation": "Radon er en radioaktiv gass som kan samle seg i bygninger og øke risikoen for lungekreft ved langvarig eksponering."
    },
        {
        "question": "Hva er enheten for radioaktivitet?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "A",
        "explanation": "Becquerel (Bq) er enheten for radioaktivitet og måler antall henfall per sekund."
    },
    {
        "question": "Hva er enheten for absorbert dose?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "C",
        "explanation": "Gray (Gy) er enheten for absorbert dose og måler mengden energi som absorberes per enhet masse."
    },
    {
        "question": "Hva er enheten for ekvivalent dose?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "B",
        "explanation": "Sievert (Sv) er enheten for ekvivalent dose og tar hensyn til biologisk effekt av strålingen."
    },
    {
        "question": "Hva er enheten for effektiv dose?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "B",
        "explanation": "Sievert (Sv) er også enheten for effektiv dose, som vurderer risikoen for stråling på ulike organer og vev."
    },
    {
        "question": "Hva er elektromagnetisk stråling?",
        "options": ["A. Stråling som består av partikler", "B. Stråling som består av elektriske og magnetiske felt", "C. Stråling som består av lyd", "D. Stråling som består av varme"],
        "answer": "B",
        "explanation": "Elektromagnetisk stråling består av elektriske og magnetiske felt som svinger i takt."
    },
    {
        "question": "Hvilken type stråling har høyest energi?",
        "options": ["A. Radiobølger", "B. Mikrobølger", "C. Synlig lys", "D. Gammastråling"],
        "answer": "D",
        "explanation": "Gammastråling har høyest energi blant de nevnte typene stråling."
    },
    {
        "question": "Hva er halveringstid?",
        "options": ["A. Tiden det tar for halvparten av atomene i et radioaktivt stoff å henfalle", "B. Tiden det tar for et stoff å miste halvparten av sin energi", "C. Tiden det tar for et stoff å fordoble sin masse", "D. Tiden det tar for et stoff å nå sin maksimale energi"],
        "answer": "A",
        "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling."
    },
    {
        "question": "Hva er trådløs kommunikasjon?",
        "options": ["A. Kommunikasjon som bruker kabler", "B. Kommunikasjon som bruker radiobølger", "C. Kommunikasjon som bruker fiberoptikk", "D. Kommunikasjon som bruker lyd"],
        "answer": "B",
        "explanation": "Trådløs kommunikasjon bruker radiobølger til å sende og motta informasjon uten bruk av kabler."
    },
    {
        "question": "Hva brukes radiobølger til i trådløs kommunikasjon?",
        "options": ["A. Å sende og motta informasjon", "B. Å varme opp mat", "C. Å lage lyd", "D. Å produsere lys"],
        "answer": "A",
        "explanation": "Radiobølger brukes til å sende og motta informasjon i trådløs kommunikasjon."
    },
    {
        "question": "Hva er en bølgelengde?",
        "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
        "answer": "A",
        "explanation": "Bølgelengde er avstanden mellom to bølgetopper eller to bølgebunner."
    },
    {
        "question": "Hva er frekvens?",
        "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
        "answer": "D",
        "explanation": "Frekvens er antall bølger som passerer et punkt per sekund."
    },
    {
        "question": "Hva er sammenhengen mellom bølgelengde og frekvens?",
        "options": ["A. De er uavhengige av hverandre", "B. Når bølgelengden øker, øker også frekvensen", "C. Når bølgelengden øker, minker frekvensen", "D. Når bølgelengden minker, minker også frekvensen"],
        "answer": "C",
        "explanation": "Når bølgelengden øker, minker frekvensen."
    },
    {
        "question": "Hva er elektromagnetisk stråling?",
        "options": ["A. Stråling som består av partikler", "B. Stråling som består av elektriske og magnetiske felt", "C. Stråling som består av lyd", "D. Stråling som består av varme"],
        "answer": "B",
        "explanation": "Elektromagnetisk stråling består av elektriske og magnetiske felt."
    },
    {
        "question": "Hva er ioniserende stråling?",
        "options": ["A. Stråling som kan ionisere atomer og molekyler", "B. Stråling som kan varme opp objekter", "C. Stråling som kan kjøle ned objekter", "D. Stråling som kan skape lys"],
        "answer": "A",
        "explanation": "Ioniserende stråling kan ionisere atomer og molekyler."
    },
    {
        "question": "Hva er alfastråling?",
        "options": ["A. Stråling som består av elektroner", "B. Stråling som består av heliumkjerner", "C. Stråling som består av fotoner", "D. Stråling som består av nøytroner"],
        "answer": "B",
        "explanation": "Alfastråling består av heliumkjerner."
    },
    {
        "question": "Hva er betastråling?",
        "options": ["A. Stråling som består av elektroner eller positroner", "B. Stråling som består av heliumkjerner", "C. Stråling som består av fotoner", "D. Stråling som består av nøytroner"],
        "answer": "A",
        "explanation": "Betastråling består av elektroner eller positroner."
    },
    {
        "question": "Hva er gammastråling?",
        "options": ["A. Stråling som består av elektroner", "B. Stråling som består av heliumkjerner", "C. Stråling som består av fotoner", "D. Stråling som består av nøytroner"],
        "answer": "C",
        "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner."
    },
        {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet.",
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling.",
        },
        {
            "question": "Hva er gammastråling?",
            "options": ["A. Partikkelstråling", "B. Synlig lys", "C. Ultrafiolett stråling", "D. Svært energirik elektromagnetisk stråling"],
            "answer": "D",
            "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner.",
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling med lav energi", "B. Stråling som ikke påvirker atomer", "C. Stråling som består av partikler", "D. Stråling som har nok energi til å fjerne elektroner fra et atom eller molekyl"],
            "answer": "D",
            "explanation": "Ioniserende stråling har nok energi til å fjerne elektroner fra et atom eller molekyl.",
        },
        {
            "question": "Hvorfor skal vi være redde for radon i huset vårt?",
            "options": [
                "A. Radon er en edelgass som kan forårsake eksplosjoner",
                "B. Radon er en radioaktiv gass som kan øke risikoen for lungekreft",
                "C. Radon kan føre til elektriske feil i huset",
                "D. Radon kan forårsake vannforurensning"
            ],
            "answer": "B",
            "explanation": "Radon er en radioaktiv gass som kan øke risikoen for lungekreft.",
        },
        {
            "question": "Hva er enheten for radioaktivitet?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "A",
            "explanation": "Enheten for radioaktivitet er Becquerel (Bq).",
        },
        {
            "question": "Hva er enheten for absorbert dose?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "C",
            "explanation": "Enheten for absorbert dose er Gray (Gy).",
        },
        {
            "question": "Hva er enheten for ekvivalent dose?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "B",
            "explanation": "Enheten for ekvivalent dose er Sievert (Sv).",
        },
        {
            "question": "Hva er enheten for effektiv dose?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "B",
            "explanation": "Enheten for effektiv dose er Sievert (Sv).",
        },
        {
            "question": "Hva er elektromagnetisk stråling?",
            "options": ["A. Stråling som består av partikler", "B. Stråling som består av elektriske og magnetiske felt", "C. Stråling som består av lyd", "D. Stråling som består av varme"],
            "answer": "B",
            "explanation": "Elektromagnetisk stråling består av elektriske og magnetiske felt.",
        },
        {
            "question": "Hvilken type stråling har høyest energi?",
            "options": ["A. Radiobølger", "B. Mikrobølger", "C. Synlig lys", "D. Gammastråling"],
            "answer": "D",
            "explanation": "Gammastråling har høyest energi.",
        },
        {
            "question": "Hva er trådløs kommunikasjon?",
            "options": ["A. Kommunikasjon som bruker kabler", "B. Kommunikasjon som bruker radiobølger", "C. Kommunikasjon som bruker fiberoptikk", "D. Kommunikasjon som bruker lyd"],
            "answer": "B",
            "explanation": "Trådløs kommunikasjon bruker radiobølger.",
        },
        {
            "question": "Hva brukes radiobølger til i trådløs kommunikasjon?",
            "options": ["A. Å sende og motta informasjon", "B. Å varme opp mat", "C. Å lage lyd", "D. Å produsere lys"],
            "answer": "A",
            "explanation": "Radiobølger brukes til å sende og motta informasjon.",
        },
        {
            "question": "Hva er en bølgelengde?",
            "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
            "answer": "A",
            "explanation": "Bølgelengde er avstanden mellom to bølgetopper eller to bølgebunner.",
        },
        {
            "question": "Hva er frekvens?",
            "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
            "answer": "D",
            "explanation": "Frekvens er antall bølger som passerer et punkt per sekund.",
        },
        {
            "question": "Hva er sammenhengen mellom bølgelengde og frekvens?",
            "options": ["A. De er uavhengige av hverandre", "B. Når bølgelengden øker, øker også frekvensen", "C. Når bølgelengden øker, minker frekvensen", "D. Når bølgelengden minker, minker også frekvensen"],
            "answer": "C",
            "explanation": "Når bølgelengden øker, minker frekvensen.",
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling som kan ionisere atomer og molekyler", "B. Stråling som kan varme opp objekter", "C. Stråling som kan kjøle ned objekter", "D. Stråling som kan skape lys"],
            "answer": "A",
            "explanation": "Ioniserende stråling kan ionisere atomer og molekyler.",
        }
]

# --- LOGIC ENGINE ---

class CapoFinalFixQuiz:
    def __init__(self, data):
        self.original_data = data
        self.quiz_data = []
        self.current_index = 0
        self.score = 0
        
        display(HTML(custom_css))
        
#---------------------------------------------------------------------# 1. Overskrift på quizen skrives under her ---------------------------------------
        self.header_widget = widgets.HTML(
            value="<div class='quiz-header-widget'><h1 class='quiz-title'>Quiz</h1></div>",
            layout=widgets.Layout(width='100%')
        )
        
        # 2. Content Box
        self.content_box = widgets.VBox([])
        self.content_box.add_class('quiz-content-area')
        self.content_box.layout = widgets.Layout(width='100%')
        
        # 3. Glass Card Container
        self.glass_card = widgets.VBox([self.header_widget, self.content_box])
        self.glass_card.add_class('quiz-card-container')
        
        # 4. Master Background
        self.master_bg = widgets.VBox([self.glass_card])
        self.master_bg.add_class('master-container')
        
        display(self.master_bg)
        self.start_game()

    def start_game(self):
        self.quiz_data = self.original_data.copy()
        random.shuffle(self.quiz_data)
        self.current_index = 0
        self.score = 0
        self.render_question()

    def render_question(self):
        if self.current_index >= len(self.quiz_data):
            self.render_results()
            return

        q_data = self.quiz_data[self.current_index]
        total_q = len(self.quiz_data)
        
        prog_val = ((self.current_index) / total_q) * 100
        
        html_content = f"""
        <div class="slide-in-up">
            <div class="progress-wrapper">
                <div style="display:flex; justify-content:space-between; color:#6b7280; font-size:13px; margin-bottom:10px; font-weight:700; letter-spacing:1px; text-transform:uppercase;">
                    <span>Progresjon</span>
                    <span>{self.current_index + 1} / {total_q}</span>
                </div>
                <div class="progress-bg">
                    <div class="progress-fill" style="width:{prog_val}%;"></div>
                </div>
            </div>
            <div class='question-text'>{q_data['question']}</div>
        </div>
        """
        top_html = widgets.HTML(html_content)

        self.buttons = []
        for idx, option in enumerate(q_data['options']):
            btn = widgets.Button(description=option)
            btn.layout = widgets.Layout(width='100%', height='auto')
            btn.on_click(lambda b, i=idx, code=chr(65+idx): self.handle_answer(b, i, code))
            self.buttons.append(btn)
        
        btn_box = widgets.VBox(self.buttons)
        btn_box.layout = widgets.Layout(width='100%')
        btn_box.add_class('slide-in-up')
        
        self.feedback_out = widgets.Output()
        self.feedback_out.layout = widgets.Layout(width='100%')

        self.content_box.children = [top_html, btn_box, self.feedback_out]

    def handle_answer(self, clicked_btn, index, selected_code):
        q_data = self.quiz_data[self.current_index]
        correct_code = q_data['answer']
        is_correct = (selected_code == correct_code)

        if is_correct:
            self.score += 1

        for btn in self.buttons:
            btn.disabled = True
            if btn.description.startswith(correct_code) and not is_correct:
                btn.add_class('fasit-highlight')
            
        if is_correct:
            clicked_btn.add_class('correct-flash')
            color = "#10b981"
            title = "✨ HELT RIKTIG!"
        else:
            clicked_btn.add_class('wrong-answer')
            color = "#ef4444"
            title = "❌ DESVERRE FEIL"

        with self.feedback_out:
            clear_output()
            feedback_html = f"""
            <div class='feedback-box' style='border-left-color: {color};'>
                <strong style='color:{color}; font-size:20px; letter-spacing:1px;'>{title}</strong><br>
                <div style="margin-top:10px; color:#374151;">{q_data['explanation']}</div>
            </div>
            """
            display(HTML(feedback_html))
            
            btn_next = widgets.Button(
                description="GÅ VIDERE ➜", 
                button_style='primary', 
                layout=widgets.Layout(width='100%', margin='30px 0 0 0') 
            )
            btn_next.style.font_weight = '900'
            btn_next.style.font_size = '18px'
            btn_next.style.button_color = '#2563eb'
            btn_next.on_click(lambda b: self.next_q())
            display(btn_next)

    def next_q(self):
        self.current_index += 1
        self.render_question()

    def generate_confetti(self):
        """Genererer en 'sky' av konfetti divs som flyter over skjermen."""
        confetti_html = "<div class='confetti-container'>"
        colors = ['#ef476f', '#ffd166', '#06d6a0', '#118ab2', '#073b4c', '#e73c7e']
        
        # Genererer 60 biter konfetti
        for i in range(60): 
            left = random.randint(0, 100)
            # Raskere fall: mellom 2s og 5s
            duration = random.uniform(2, 5) 
            delay = random.uniform(0, 3)
            bg = random.choice(colors)
            size = random.randint(8, 14)
            
            confetti_html += f"""
            <div class='confetti' style='
                left:{left}%; 
                animation-duration:{duration}s; 
                animation-delay:{delay}s; 
                background-color:{bg};
                width:{size}px;
                height:{size}px;
            '></div>
            """
        confetti_html += "</div>"
        return confetti_html

    def render_results(self):
        total = len(self.quiz_data)
        percent = (self.score / total) * 100 if total > 0 else 0
        
        # Generer konfetti HTML hvis man består
        confetti_effect = self.generate_confetti() if percent >= 50 else ""
        
        if percent >= 80:
            msg = "Fantastisk! Du er en ekte mester! 🏆"
            bg = "#d1fae5"
            txt = "#065f46"
        elif percent >= 50:
            msg = "Veldig bra jobbet! 📚"
            bg = "#fef3c7"
            txt = "#92400e"
        else:
            msg = "Prøv igjen – øvelse gjør mester! 💪"
            bg = "#fee2e2"
            txt = "#991b1b"

        # Her injiserer vi konfetti ØVERST i HTML-blokken, men inni containeren
        res_html = widgets.HTML(f"""
        <div style="position:relative; width:100%; height:100%;">
            {confetti_effect} 
            <div style="text-align:center; padding-top:20px; position:relative; z-index:10;" class="slide-in-up">
                <div style="font-size:16px; color:#6b7280; letter-spacing:3px; font-weight:700; text-transform:uppercase;">Resultat</div>
                <div style="font-size:100px; font-weight:900; color:#111827; margin: 10px 0; line-height:1; letter-spacing:-4px;">
                    {self.score}/{total}
                </div>
                <div class='result-msg' style="background-color: {bg}; color: {txt};">
                    {msg}
                </div>
            </div>
        </div>
        """)
        
        out_chart = widgets.Output()
        with out_chart:
            fig, ax = plt.subplots(figsize=(6, 6), facecolor='none') 
            sizes = [self.score, total - self.score]
            colors = ['#10b981', '#ef4444']
            wedges, texts, autotexts = ax.pie(
                sizes, labels=['Riktig', 'Feil'], colors=colors, 
                autopct='%1.0f%%', startangle=90, textprops={'color':"#374151", 'fontsize': 14, 'weight':'bold'},
                explode=(0.05, 0)
            )
            plt.setp(autotexts, size=18, weight="bold", color="white")
            plt.close(fig)
            display(fig)

        chart_box = widgets.HBox([out_chart], layout=widgets.Layout(justify_content='center'))

        btn_restart = widgets.Button(
            description="START PÅ NYTT 🔄",
            layout=widgets.Layout(width='100%', height='80px', margin='40px 0 0 0')
        )
        btn_restart.style.button_color = '#2563eb'
        btn_restart.style.text_color = 'white'
        btn_restart.style.font_weight = 'bold'
        btn_restart.style.font_size = '24px'
        btn_restart.on_click(lambda b: self.start_game())

        self.content_box.children = [res_html, chart_box, btn_restart]

# --- START APP ---
app = CapoFinalFixQuiz(original_quiz_data)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, Javascript, clear_output, display
import random
import matplotlib.pyplot as plt

# --- CSS: THE SINGULARITY EDITION ---
custom_css = """
<style>
    /* --- 1. ATMOSFÆRE & BAKGRUNN --- */
    .universe-container {
        background: radial-gradient(circle at center, #1e293b, #0f172a);
        position: relative;
        padding: 60px;
        border-radius: 24px;
        overflow: hidden;
        font-family: 'Inter', system-ui, -apple-system, sans-serif;
        box-shadow: inset 0 0 100px rgba(0,0,0,0.5);
    }

    /* Flytende Orber */
    .orb {
        position: absolute;
        border-radius: 50%;
        filter: blur(40px);
        opacity: 0.4;
        animation: floatOrb 20s infinite ease-in-out;
        z-index: 0;
    }
    .orb-1 { top: -10%; left: -10%; width: 300px; height: 300px; background: #3b82f6; animation-delay: 0s; }
    .orb-2 { bottom: -10%; right: -10%; width: 400px; height: 400px; background: #8b5cf6; animation-delay: -5s; }
    .orb-3 { top: 40%; left: 60%; width: 200px; height: 200px; background: #ec4899; animation-delay: -10s; }

    @keyframes floatOrb {
        0%, 100% { transform: translate(0, 0) scale(1); }
        33% { transform: translate(30px, -50px) scale(1.1); }
        66% { transform: translate(-20px, 20px) scale(0.9); }
    }

    /* --- 2. THE HUD CARD (GLASSMORPHISM ULTIMATE) --- */
    .quiz-card-container {
        background: rgba(255, 255, 255, 0.85);
        backdrop-filter: blur(20px);
        -webkit-backdrop-filter: blur(20px);
        border-radius: 24px;
        box-shadow: 
            0 20px 50px rgba(0,0,0,0.2), 
            0 0 0 1px rgba(255, 255, 255, 0.5) inset;
        width: 100%;
        max-width: 900px; /* Bred kino-format */
        margin: 0 auto;
        position: relative;
        z-index: 10;
        overflow: hidden;
        transition: transform 0.3s ease;
    }

    /* Riste-animasjon ved feil svar */
    .shake-effect {
        animation: shake 0.5s cubic-bezier(.36,.07,.19,.97) both;
    }
    @keyframes shake {
        10%, 90% { transform: translate3d(-1px, 0, 0); }
        20%, 80% { transform: translate3d(2px, 0, 0); }
        30%, 50%, 70% { transform: translate3d(-4px, 0, 0); }
        40%, 60% { transform: translate3d(4px, 0, 0); }
    }

    /* Header */
    .quiz-header-widget {
        display: flex;
        justify-content: space-between;
        align-items: center;
        padding: 25px 40px;
        border-bottom: 1px solid rgba(0,0,0,0.05);
        background: rgba(255,255,255,0.4);
    }
    
    .brand-title {
        font-size: 28px;
        font-weight: 900;
        letter-spacing: 2px;
        background: linear-gradient(135deg, #2563eb, #db2777);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin: 0;
    }

    .streak-counter {
        font-size: 18px;
        font-weight: 700;
        color: #f59e0b;
        background: rgba(254, 243, 199, 0.8);
        padding: 8px 16px;
        border-radius: 20px;
        box-shadow: 0 2px 10px rgba(245, 158, 11, 0.2);
        display: flex;
        align-items: center;
        gap: 8px;
    }

    /* Innhold */
    .quiz-content-area {
        padding: 40px 50px;
    }

    /* Spørsmål */
    .question-text {
        font-size: 30px;
        color: #0f172a;
        font-weight: 800;
        line-height: 1.3;
        margin-bottom: 30px;
        animation: fadeSlideIn 0.5s ease-out;
    }

    @keyframes fadeSlideIn {
        from { opacity: 0; transform: translateY(10px); }
        to { opacity: 1; transform: translateY(0); }
    }

    /* --- KNAPPER MED RIPPLE & GLOW --- */
    .widget-button { 
        position: relative;
        box-sizing: border-box !important;
        font-size: 18px !important; 
        font-weight: 600 !important;
        border-radius: 16px !important; 
        border: 2px solid rgba(0,0,0,0.05) !important;
        background: white !important;
        color: #334155 !important;
        padding: 20px 30px !important;
        margin-bottom: 12px !important;
        width: 100% !important;
        text-align: left !important;
        transition: all 0.2s cubic-bezier(0.4, 0, 0.2, 1) !important;
        overflow: hidden; /* For ripple effekten */
    }
    
    .widget-button:hover:enabled {
        border-color: #3b82f6 !important;
        background: #f8fafc !important;
        transform: scale(1.01) translateY(-2px);
        box-shadow: 0 10px 25px rgba(59, 130, 246, 0.15) !important;
        z-index: 2;
    }

    /* Fasit & Svar Styles */
    .correct-flash { 
        background: #10b981 !important; 
        color: white !important; 
        border-color: #10b981 !important;
        box-shadow: 0 0 30px rgba(16, 185, 129, 0.4) !important;
        font-weight: 800 !important;
    }
    
    .wrong-answer {
        background: #ef4444 !important;
        color: white !important;
        border-color: #ef4444 !important;
    }

    .fasit-highlight {
        background: #ecfdf5 !important;
        color: #059669 !important;
        border: 2px solid #10b981 !important;
    }

    /* Feedback Box */
    .feedback-container {
        margin-top: 30px;
        padding: 25px;
        background: #f1f5f9;
        border-radius: 16px;
        border-left: 6px solid;
        animation: popIn 0.4s cubic-bezier(0.175, 0.885, 0.32, 1.275);
    }
    
    @keyframes popIn {
        from { opacity: 0; transform: scale(0.95); }
        to { opacity: 1; transform: scale(1); }
    }

    /* Progresjonslinje */
    .progress-track {
        height: 6px;
        background: #e2e8f0;
        border-radius: 3px;
        margin-bottom: 30px;
        overflow: hidden;
    }
    .progress-bar-fill {
        height: 100%;
        background: linear-gradient(90deg, #3b82f6, #8b5cf6, #ec4899);
        background-size: 200% 100%;
        animation: gradientMove 3s linear infinite;
        transition: width 0.8s cubic-bezier(0.4, 0, 0.2, 1);
    }
    @keyframes gradientMove { 0% { background-position: 100% 0; } 100% { background-position: -100% 0; } }

    /* Resultat */
    .score-big {
        font-size: 100px;
        font-weight: 900;
        line-height: 1;
        background: -webkit-linear-gradient(#3b82f6, #8b5cf6);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin: 20px 0;
    }

    /* Konfetti er 'must' */
    .confetti {
        position: absolute;
        width: 10px; height: 10px;
        top: -10px;
        z-index: 100;
        animation: confettiFall linear forwards;
    }
    @keyframes confettiFall {
        to { transform: translateY(600px) rotate(720deg); }
    }

</style>
"""

# --- DATA ---
original_quiz_data = [
  {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet.",
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling.",
        },
        {
            "question": "Hva er gammastråling?",
            "options": ["A. Partikkelstråling", "B. Synlig lys", "C. Ultrafiolett stråling", "D. Svært energirik elektromagnetisk stråling"],
            "answer": "D",
            "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner.",
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling med lav energi", "B. Stråling som ikke påvirker atomer", "C. Stråling som består av partikler", "D. Stråling som har nok energi til å fjerne elektroner fra et atom eller molekyl"],
            "answer": "D",
            "explanation": "Ioniserende stråling har nok energi til å fjerne elektroner fra et atom eller molekyl.",
        },
        {
            "question": "Hva er rødforskyvning?",
            "options": ["A. Når lysbølger stopper helt opp", "B. Når lysbølger blir kortere og bølgelengden minker", "C. Når lysbølger endrer farge til blått", "D. Når lysbølger blir strukket ut og bølgelengden øker"],
            "answer": "D",
            "explanation": "Rødforskyvning skjer når lysbølger blir strukket ut og bølgelengden øker.",
        },
        {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet."
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling."
        },
        {
            "question": "Hva er gammastråling?",
            "options": ["A. Partikkelstråling", "B. Synlig lys", "C. Ultrafiolett stråling", "D. Svært energirik elektromagnetisk stråling"],
            "answer": "D",
            "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner."
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling med lav energi", "B. Stråling som ikke påvirker atomer", "C. Stråling som består av partikler", "D. Stråling som har nok energi til å fjerne elektroner fra et atom eller molekyl"],
            "answer": "D",
             "explanation": "Ioniserende stråling har nok energi til å fjerne elektroner fra et atom eller molekyl."
        },
        {
            "question": "Hva er rødforskyvning?",
            "options": ["A. Når lysbølger stopper helt opp", "B. Når lysbølger blir kortere og bølgelengden minker", "C. Når lysbølger endrer farge til blått", "D. Når lysbølger blir strukket ut og bølgelengden øker"],
            "answer": "D",
             "explanation": "Rødforskyvning skjer når lysbølger blir strukket ut og bølgelengden øker."
        },
        {
            "question": "Hva er et foton?",
            "options": ["A. En type atom", "B. En bølge", "C. En lyspartikkel som frakter energi", "D. En partikkel med masse"],
            "answer": "C",
            "explanation": "Et foton er en lyspartikkel som frakter energi."
        },
        {
            "question": "Hva slags stråling sender den radioaktive kilden ut hvis ikke all strålingen blir stanset av papir, men derimot både av en treplate og en blyplate?",
            "options": ["A. Alfastråling", "B. Betastråling", "C. Gammastråling"],
            "answer": "B",
            "explanation": "Betastråling blir stanset av en treplate, men ikke av papir."
        },
        {
            "question": "Hvis halveringstiden for Ce-137 er 30 år, hvilken strålingsaktivitet hadde kilden for 90 år siden hvis den nå er 1000 Bq?",
            "options": ["A. 500 Bq", "B. 1000 Bq", "C. 2000 Bq", "D. 4000 Bq", "E. 8000 Bq"],
            "answer": "E",
            "explanation": "Etter 90 år, som er tre halveringstider, vil aktiviteten ha blitt redusert til en åttendedel av den opprinnelige aktiviteten, så den opprinnelige aktiviteten var 8000 Bq."
        },
        {
            "question": "Hvor gammelt er et lårbein fra en moskusokse hvis innholdet av isotopen karbon-14 bare er om lag 25 % av det vi finner i en levende moskusokse? Karbon-14 har en halveringstid på 5730 år.",
            "options": ["A. 1432 år", "B. 5730 år", "C. 7163 år", "D. 11 460 år", "E. 17 190 år"],
            "answer": "D",
            "explanation": "Hvis innholdet av karbon-14 er redusert til 25 %, betyr det at to halveringstider har gått, så lårbeinet er 11 460 år gammelt."
        },
        {
        "question": "Når Thorium-234 sender ut en betapartikkel, vil atomnummeret",
        "options": ["A. Ikke endres", "B. Økes med en", "C. Minkes med en", "D. Minkes med to", "E. Minkes med tre"],
        "answer": "B",
        "explanation": "Når Thorium-234 sender ut en betapartikkel, omdannes en nøytron til et proton, noe som øker atomnummeret med en."
    },
    {
        "question": "Når Radon-222 sender ut en alfapartikkel, vil atomnummeret",
        "options": ["A. Minskes med to", "B. Minskes med fire", "C. Økes med to", "D. Økes med fire", "E. Økes med tre"],
        "answer": "A",
        "explanation": "Når Radon-222 sender ut en alfapartikkel, mister det to protoner og to nøytroner, noe som reduserer atomnummeret med to."
    },
    {
        "question": "Når Radon-222 sender ut gammastråling, vil atomnummeret",
        "options": ["A. Økes med en", "B. Minkes med en", "C. Minkes med to", "D. Ikke endres", "E. Økes med tre"],
        "answer": "D",
        "explanation": "Gammastråling endrer ikke antall protoner eller nøytroner i kjernen, så atomnummeret forblir uendret."
    },
    {
        "question": "Hva kan man si om galaksene hvis lyset fra de fleste galakser er rødforskjøvet?",
        "options": ["A. Beveger seg mot oss", "B. Beveger seg fra oss", "C. Beveger seg med samme hastighet som oss", "D. Står stille i forhold til oss"],
        "answer": "B",
        "explanation": "Rødforskyvning indikerer at galaksene beveger seg bort fra oss, noe som støtter teorien om at universet utvider seg."
    },
    {
        "question": "Hva bestemmer protoner og nøytroner i atomene?",
        "options": ["A. Protoner bestemmer atomnummeret, nøytroner bestemmer isotopen", "B. Protoner bestemmer massen, nøytroner bestemmer ladningen", "C. Protoner bestemmer ladningen, nøytroner bestemmer massen", "D. Protoner bestemmer isotopen, nøytroner bestemmer atomnummeret"],
        "answer": "A",
        "explanation": "Antall protoner i kjernen bestemmer atomnummeret og dermed hvilket grunnstoff det er. Antall nøytroner bestemmer isotopen av grunnstoffet."
    },
    {
        "question": "Hva er halveringstiden for et radioaktivt stoff som reduseres til 12,5 % av den opprinnelige massen i løpet av en time?",
        "options": ["A. 15 minutter", "B. 20 minutter", "C. 30 minutter", "D. 40 minutter"],
        "answer": "B",
        "explanation": "Hvis et stoff reduseres til 12,5 % av den opprinnelige massen, har det gjennomgått tre halveringstider (100 % -> 50 % -> 25 % -> 12,5 %). En time delt på tre gir en halveringstid på 20 minutter."
    },
    {
        "question": "Hva er Radon?",
        "options": ["A. En edelgass", "B. En metall", "C. En halvmetall", "D. En ikke-metall"],
        "answer": "A",
        "explanation": "Radon er en edelgass som er radioaktiv og finnes naturlig i jordskorpen."
    },
    {
        "question": "Hva brukes PET-scan til?",
        "options": ["A. Radioaktive markører som enten må bli svelget, inhalert eller injeksert i kroppen og kan derfor viser hvordan ulike vev eller organer i kroppen jobber på ett celle-nivå. Brukes for å diagnostisere kreft, hjerteproblemer eller hjerneproblemer", "B. PET-scan brukes for å måle PET(dyret på engelsk)-en din, altså katten, hunden eller hamsteren din", "C. Å måle blodtrykk", "D. Å måle blodsukker"],
        "answer": "A",
        "explanation": "PET-scan bruker radioaktive markører for å vise hvordan vev og organer fungerer på cellenivå, og brukes til å diagnostisere sykdommer som kreft, hjerteproblemer og hjerneproblemer."
    },
    {
        "question": "Hva brukes CT-scan til?",
        "options": ["A. Røntgenundersøkelse der det blir tatt tverrsnittbilde av aktuelle områder i kroppen (bløtvev, blodårer, organer, bein og vev)", "B. Viser hvordan ulike vev eller organer i kroppen jobber på ett celle-nivå", "C. Å måle blodtrykk", "D. Å måle blodsukker"],
        "answer": "A",
        "explanation": "CT-scan er en røntgenundersøkelse som tar tverrsnittbilder av kroppen, inkludert bløtvev, blodårer, organer, bein og vev."
    },
    {
        "question": "Hva er kosmisk bakgrunnstråling?",
        "options": ["A. Stråling fra solen", "B. Stråling fra jorden", "C. Stråling fra universets begynnelse", "D. Stråling fra stjerner"],
        "answer": "C",
        "explanation": "Kosmisk bakgrunnstråling er stråling som stammer fra universets begynnelse og er en viktig observasjon som støtter Big Bang-teorien."
    },
    {
        "question": "Hva er Big Bang teorien?",
        "options": ["A. Teorien om at universet alltid har eksistert", "B. Teorien om at universet ble skapt av en stor eksplosjon", "C. Teorien om at universet vil ende i en stor eksplosjon", "D. Teorien om at universet er uendelig"],
        "answer": "B",
        "explanation": "Big Bang-teorien foreslår at universet ble skapt av en stor eksplosjon for omtrent 13,8 milliarder år siden."
    },
    {
        "question": "Hvilke observasjoner støtter Big Bang teorien?",
        "options": ["A. Rødforskyvning av galakser, kosmisk bakgrunnstråling, fordeling av grunnstoffer", "B. Blåforskyvning av galakser, kosmisk bakgrunnstråling, fordeling av grunnstoffer", "C. Rødforskyvning av galakser, stråling fra solen, fordeling av grunnstoffer", "D. Blåforskyvning av galakser, stråling fra solen, fordeling av grunnstoffer"],
        "answer": "A",
        "explanation": "Rødforskyvning av galakser, kosmisk bakgrunnstråling og fordelingen av grunnstoffer i universet er observasjoner som støtter Big Bang-teorien."
    },
    {
        "question": "Hva er halveringstiden til karbon-14?",
        "options": ["A. 5730 år", "B. 5000 år", "C. 6000 år", "D. 7000 år"],
        "answer": "A",
        "explanation": "Halveringstiden til karbon-14 er 5730 år, noe som betyr at det tar 5730 år for halvparten av karbon-14 i en prøve å brytes ned."
    },
    {
        "question": "Hvorfor skal vi være redde for radon i huset vårt?",
        "options": [
            "A. Radon er en edelgass som kan forårsake eksplosjoner",
            "B. Radon er en radioaktiv gass som kan øke risikoen for lungekreft",
            "C. Radon kan føre til elektriske feil i huset",
            "D. Radon kan forårsake vannforurensning"
        ],
        "answer": "B",
        "explanation": "Radon er en radioaktiv gass som kan samle seg i bygninger og øke risikoen for lungekreft ved langvarig eksponering."
    },
        {
        "question": "Hva er enheten for radioaktivitet?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "A",
        "explanation": "Becquerel (Bq) er enheten for radioaktivitet og måler antall henfall per sekund."
    },
    {
        "question": "Hva er enheten for absorbert dose?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "C",
        "explanation": "Gray (Gy) er enheten for absorbert dose og måler mengden energi som absorberes per enhet masse."
    },
    {
        "question": "Hva er enheten for ekvivalent dose?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "B",
        "explanation": "Sievert (Sv) er enheten for ekvivalent dose og tar hensyn til biologisk effekt av strålingen."
    },
    {
        "question": "Hva er enheten for effektiv dose?",
        "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
        "answer": "B",
        "explanation": "Sievert (Sv) er også enheten for effektiv dose, som vurderer risikoen for stråling på ulike organer og vev."
    },
    {
        "question": "Hva er elektromagnetisk stråling?",
        "options": ["A. Stråling som består av partikler", "B. Stråling som består av elektriske og magnetiske felt", "C. Stråling som består av lyd", "D. Stråling som består av varme"],
        "answer": "B",
        "explanation": "Elektromagnetisk stråling består av elektriske og magnetiske felt som svinger i takt."
    },
    {
        "question": "Hvilken type stråling har høyest energi?",
        "options": ["A. Radiobølger", "B. Mikrobølger", "C. Synlig lys", "D. Gammastråling"],
        "answer": "D",
        "explanation": "Gammastråling har høyest energi blant de nevnte typene stråling."
    },
    {
        "question": "Hva er halveringstid?",
        "options": ["A. Tiden det tar for halvparten av atomene i et radioaktivt stoff å henfalle", "B. Tiden det tar for et stoff å miste halvparten av sin energi", "C. Tiden det tar for et stoff å fordoble sin masse", "D. Tiden det tar for et stoff å nå sin maksimale energi"],
        "answer": "A",
        "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling."
    },
    {
        "question": "Hva er trådløs kommunikasjon?",
        "options": ["A. Kommunikasjon som bruker kabler", "B. Kommunikasjon som bruker radiobølger", "C. Kommunikasjon som bruker fiberoptikk", "D. Kommunikasjon som bruker lyd"],
        "answer": "B",
        "explanation": "Trådløs kommunikasjon bruker radiobølger til å sende og motta informasjon uten bruk av kabler."
    },
    {
        "question": "Hva brukes radiobølger til i trådløs kommunikasjon?",
        "options": ["A. Å sende og motta informasjon", "B. Å varme opp mat", "C. Å lage lyd", "D. Å produsere lys"],
        "answer": "A",
        "explanation": "Radiobølger brukes til å sende og motta informasjon i trådløs kommunikasjon."
    },
    {
        "question": "Hva er en bølgelengde?",
        "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
        "answer": "A",
        "explanation": "Bølgelengde er avstanden mellom to bølgetopper eller to bølgebunner."
    },
    {
        "question": "Hva er frekvens?",
        "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
        "answer": "D",
        "explanation": "Frekvens er antall bølger som passerer et punkt per sekund."
    },
    {
        "question": "Hva er sammenhengen mellom bølgelengde og frekvens?",
        "options": ["A. De er uavhengige av hverandre", "B. Når bølgelengden øker, øker også frekvensen", "C. Når bølgelengden øker, minker frekvensen", "D. Når bølgelengden minker, minker også frekvensen"],
        "answer": "C",
        "explanation": "Når bølgelengden øker, minker frekvensen."
    },
    {
        "question": "Hva er elektromagnetisk stråling?",
        "options": ["A. Stråling som består av partikler", "B. Stråling som består av elektriske og magnetiske felt", "C. Stråling som består av lyd", "D. Stråling som består av varme"],
        "answer": "B",
        "explanation": "Elektromagnetisk stråling består av elektriske og magnetiske felt."
    },
    {
        "question": "Hva er ioniserende stråling?",
        "options": ["A. Stråling som kan ionisere atomer og molekyler", "B. Stråling som kan varme opp objekter", "C. Stråling som kan kjøle ned objekter", "D. Stråling som kan skape lys"],
        "answer": "A",
        "explanation": "Ioniserende stråling kan ionisere atomer og molekyler."
    },
    {
        "question": "Hva er alfastråling?",
        "options": ["A. Stråling som består av elektroner", "B. Stråling som består av heliumkjerner", "C. Stråling som består av fotoner", "D. Stråling som består av nøytroner"],
        "answer": "B",
        "explanation": "Alfastråling består av heliumkjerner."
    },
    {
        "question": "Hva er betastråling?",
        "options": ["A. Stråling som består av elektroner eller positroner", "B. Stråling som består av heliumkjerner", "C. Stråling som består av fotoner", "D. Stråling som består av nøytroner"],
        "answer": "A",
        "explanation": "Betastråling består av elektroner eller positroner."
    },
    {
        "question": "Hva er gammastråling?",
        "options": ["A. Stråling som består av elektroner", "B. Stråling som består av heliumkjerner", "C. Stråling som består av fotoner", "D. Stråling som består av nøytroner"],
        "answer": "C",
        "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner."
    },
        {
            "question": "Hvilket grunnstoff har 30 protoner i kjernen?",
            "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Kobber (Cu)"],
            "answer": "C",
            "explanation": "Sink (Zn) har 30 protoner i kjernen, siden det har atomnummer 30 i periodesystemet.",
        },
        {
            "question": "Hva er halveringstid?",
            "options": ["A. Tiden det tar for et stoff å brytes ned", "B. Tiden det tar for stråling å forsvinne", "C. Tiden det tar for et atom å bli stabilt", "D. Tiden det tar før halvparten av atomkjernene har sendt ut radioaktiv stråling"],
            "answer": "D",
            "explanation": "Halveringstid er tiden det tar før halvparten av atomkjernene i et radioaktivt stoff har sendt ut radioaktiv stråling.",
        },
        {
            "question": "Hva er gammastråling?",
            "options": ["A. Partikkelstråling", "B. Synlig lys", "C. Ultrafiolett stråling", "D. Svært energirik elektromagnetisk stråling"],
            "answer": "D",
            "explanation": "Gammastråling er svært energirik elektromagnetisk stråling og består av fotoner.",
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling med lav energi", "B. Stråling som ikke påvirker atomer", "C. Stråling som består av partikler", "D. Stråling som har nok energi til å fjerne elektroner fra et atom eller molekyl"],
            "answer": "D",
            "explanation": "Ioniserende stråling har nok energi til å fjerne elektroner fra et atom eller molekyl.",
        },
        {
            "question": "Hvorfor skal vi være redde for radon i huset vårt?",
            "options": [
                "A. Radon er en edelgass som kan forårsake eksplosjoner",
                "B. Radon er en radioaktiv gass som kan øke risikoen for lungekreft",
                "C. Radon kan føre til elektriske feil i huset",
                "D. Radon kan forårsake vannforurensning"
            ],
            "answer": "B",
            "explanation": "Radon er en radioaktiv gass som kan øke risikoen for lungekreft.",
        },
        {
            "question": "Hva er enheten for radioaktivitet?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "A",
            "explanation": "Enheten for radioaktivitet er Becquerel (Bq).",
        },
        {
            "question": "Hva er enheten for absorbert dose?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "C",
            "explanation": "Enheten for absorbert dose er Gray (Gy).",
        },
        {
            "question": "Hva er enheten for ekvivalent dose?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "B",
            "explanation": "Enheten for ekvivalent dose er Sievert (Sv).",
        },
        {
            "question": "Hva er enheten for effektiv dose?",
            "options": ["A. Becquerel (Bq)", "B. Sievert (Sv)", "C. Gray (Gy)", "D. Curie (Ci)"],
            "answer": "B",
            "explanation": "Enheten for effektiv dose er Sievert (Sv).",
        },
        {
            "question": "Hva er elektromagnetisk stråling?",
            "options": ["A. Stråling som består av partikler", "B. Stråling som består av elektriske og magnetiske felt", "C. Stråling som består av lyd", "D. Stråling som består av varme"],
            "answer": "B",
            "explanation": "Elektromagnetisk stråling består av elektriske og magnetiske felt.",
        },
        {
            "question": "Hvilken type stråling har høyest energi?",
            "options": ["A. Radiobølger", "B. Mikrobølger", "C. Synlig lys", "D. Gammastråling"],
            "answer": "D",
            "explanation": "Gammastråling har høyest energi.",
        },
        {
            "question": "Hva er trådløs kommunikasjon?",
            "options": ["A. Kommunikasjon som bruker kabler", "B. Kommunikasjon som bruker radiobølger", "C. Kommunikasjon som bruker fiberoptikk", "D. Kommunikasjon som bruker lyd"],
            "answer": "B",
            "explanation": "Trådløs kommunikasjon bruker radiobølger.",
        },
        {
            "question": "Hva brukes radiobølger til i trådløs kommunikasjon?",
            "options": ["A. Å sende og motta informasjon", "B. Å varme opp mat", "C. Å lage lyd", "D. Å produsere lys"],
            "answer": "A",
            "explanation": "Radiobølger brukes til å sende og motta informasjon.",
        },
        {
            "question": "Hva er en bølgelengde?",
            "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
            "answer": "A",
            "explanation": "Bølgelengde er avstanden mellom to bølgetopper eller to bølgebunner.",
        },
        {
            "question": "Hva er frekvens?",
            "options": ["A. Avstanden mellom to bølgetopper", "B. Høyden på en bølge", "C. Tiden det tar for en bølge å passere et punkt", "D. Antall bølger som passerer et punkt per sekund"],
            "answer": "D",
            "explanation": "Frekvens er antall bølger som passerer et punkt per sekund.",
        },
        {
            "question": "Hva er sammenhengen mellom bølgelengde og frekvens?",
            "options": ["A. De er uavhengige av hverandre", "B. Når bølgelengden øker, øker også frekvensen", "C. Når bølgelengden øker, minker frekvensen", "D. Når bølgelengden minker, minker også frekvensen"],
            "answer": "C",
            "explanation": "Når bølgelengden øker, minker frekvensen.",
        },
        {
            "question": "Hva er ioniserende stråling?",
            "options": ["A. Stråling som kan ionisere atomer og molekyler", "B. Stråling som kan varme opp objekter", "C. Stråling som kan kjøle ned objekter", "D. Stråling som kan skape lys"],
            "answer": "A",
            "explanation": "Ioniserende stråling kan ionisere atomer og molekyler.",
        }
]

# --- THE UBER ENGINE ---

class UberQuiz:
    def __init__(self, data):
        self.original_data = data
        self.quiz_data = []
        self.current_index = 0
        self.score = 0
        self.streak = 0 # NYTT: Streak teller
        
        display(HTML(custom_css))
        
        # UI Oppsett
        self.header_html = widgets.HTML()
        self.content_box = widgets.VBox([])
        self.content_box.add_class('quiz-content-area')
        
        self.card = widgets.VBox([self.header_html, self.content_box])
        self.card.add_class('quiz-card-container')
        
        # Bakgrunnsorber + Kort
        self.universe = widgets.HTML(f"""
            <div class='universe-container'>
                <div class='orb orb-1'></div>
                <div class='orb orb-2'></div>
                <div class='orb orb-3'></div>
            </div>
        """)
        
        # Triks for å legge kortet OPPÅ universet. 
        # I ren HTML/CSS er dette enklere, men i Widgets må vi være kreative.
        # Vi printer universet først som bakgrunn, og kortet renderes som en widget inni.
        # Her gjør jeg det enkelt: Universet pakker inn kortet via en VBox som har stylingen.
        self.main_display = widgets.VBox([self.card])
        self.main_display.add_class('universe-container') # Her bruker vi universet som wrapper
        
        display(self.main_display)
        self.start_game()

    def update_header(self):
        # Dynamisk Overskrift skrives ved brand-title
        streak_fire = "🔥" * (min(self.streak, 5)) # Maks 5 flammer
        self.header_html.value = f"""
        <div class='quiz-header-widget'>
            <div class='brand-title'>QUIZ</div>   
            <div class='streak-counter'>
                <span>STREAK</span>
                <span>{self.streak}</span>
                <span>{streak_fire}</span>
            </div>
        </div>
        """

    def start_game(self):
        self.quiz_data = self.original_data.copy()
        random.shuffle(self.quiz_data)
        self.current_index = 0
        self.score = 0
        self.streak = 0
        self.render_question()

    def render_question(self):
        if self.current_index >= len(self.quiz_data):
            self.render_results()
            return

        q_data = self.quiz_data[self.current_index]
        total_q = len(self.quiz_data)
        self.update_header()
        
        # Fjern shake-effekt hvis den henger igjen
        self.card.remove_class('shake-effect')
        
        # Progresjon
        prog_pct = (self.current_index / total_q) * 100
        
        html_top = f"""
        <div class="progress-track">
            <div class="progress-bar-fill" style="width: {prog_pct}%;"></div>
        </div>
        <div class='question-text'>{q_data['question']}</div>
        """
        top_widget = widgets.HTML(html_top)

        # Knapper
        self.buttons = []
        for idx, option in enumerate(q_data['options']):
            btn = widgets.Button(description=option)
            btn.layout = widgets.Layout(width='100%', height='auto')
            btn.on_click(lambda b, i=idx, code=chr(65+idx): self.handle_answer(b, i, code))
            self.buttons.append(btn)
            
        btn_box = widgets.VBox(self.buttons)
        self.feedback_out = widgets.Output()
        
        self.content_box.children = [top_widget, btn_box, self.feedback_out]

    def handle_answer(self, clicked_btn, index, selected_code):
        q_data = self.quiz_data[self.current_index]
        correct_code = q_data['answer']
        is_correct = (selected_code == correct_code)

        if is_correct:
            self.score += 1
            self.streak += 1
            color = "#10b981"
            title = "✨ FANTASTISK!"
            # Spill av "riktig"-lyd (visuelt)
            clicked_btn.add_class('correct-flash')
        else:
            self.streak = 0 # Reset streak :(
            color = "#ef4444"
            title = "❌ AI, DEN VAR LEI"
            clicked_btn.add_class('wrong-answer')
            # RISTE EFFEKT PÅ HELE KORTET
            self.card.add_class('shake-effect')

        self.update_header() # Oppdater streak-visning

        # Deaktiver knapper
        for btn in self.buttons:
            btn.disabled = True
            if btn.description.startswith(correct_code) and not is_correct:
                btn.add_class('fasit-highlight')

        with self.feedback_out:
            clear_output()
            feedback_html = f"""
            <div class='feedback-container' style='border-color: {color};'>
                <div style="font-weight:900; font-size:22px; color:{color}; margin-bottom:8px;">{title}</div>
                <div style="font-size:18px; color:#334155; line-height:1.5;">{q_data['explanation']}</div>
            </div>
            """
            display(HTML(feedback_html))
            
            # Neste knapp med gradient stil
            btn_next = widgets.Button(
                description="FORTSETT REISEN ➜", 
                layout=widgets.Layout(width='100%', margin='25px 0 0 0') 
            )
            btn_next.style.button_color = '#1e293b' # Mørk
            btn_next.style.text_color = 'white'
            btn_next.style.font_weight = 'bold'
            btn_next.on_click(lambda b: self.next_q())
            display(btn_next)

    def next_q(self):
        self.current_index += 1
        self.render_question()

    def generate_confetti(self):
        html = "<div style='position:absolute; top:0; left:0; width:100%; height:100%; pointer-events:none; overflow:hidden;'>"
        colors = ['#ec4899', '#8b5cf6', '#3b82f6', '#10b981', '#f59e0b']
        for i in range(80):
            left = random.randint(0, 100)
            dur = random.uniform(2, 5)
            delay = random.uniform(0, 2)
            bg = random.choice(colors)
            size = random.randint(8, 15)
            html += f"<div class='confetti' style='left:{left}%; background:{bg}; animation-duration:{dur}s; animation-delay:{delay}s; width:{size}px; height:{size}px;'></div>"
        html += "</div>"
        return html

    def render_results(self):
        total = len(self.quiz_data)
        percent = (self.score / total) * 100
        
        confetti = self.generate_confetti() if percent >= 50 else ""
        
        if percent >= 90:
            msg = "FANTASTISK! Dette får du til! 👑"
            sub_msg = "En meget god prestasjon."
        elif percent >= 50:
            msg = "MEGET BRA LEVERT! 🔥"
            sub_msg = "Du har god kontroll."
        else:
            msg = "NESTE GANG TAR DU DEN! 💪"
            sub_msg = "Øvelse gjør mester."

        # Resultat HTML
        res_html = widgets.HTML(f"""
        <div style="text-align:center; padding: 20px 0; position:relative;">
            {confetti}
            <div style="font-size:16px; font-weight:700; color:#64748b; letter-spacing:2px; text-transform:uppercase;">Endelig Resultat</div>
            <div class='score-big'>{self.score}/{total}</div>
            <div style="font-size:24px; font-weight:800; color:#1e293b; margin-bottom:10px;">{msg}</div>
            <div style="font-size:18px; color:#64748b;">{sub_msg}</div>
        </div>
        """)

        # Graf
        out_chart = widgets.Output()
        with out_chart:
            fig, ax = plt.subplots(figsize=(6, 6), facecolor='none')
            sizes = [self.score, total - self.score]
            colors = ['#10b981', '#ef4444']
            wedges, texts, autotexts = ax.pie(
                sizes, labels=['Riktig', 'Feil'], colors=colors, 
                autopct='%1.0f%%', startangle=90, 
                textprops={'color':"#334155", 'fontsize': 14, 'weight':'bold'},
                explode=(0.05, 0), shadow=True
            )
            plt.setp(autotexts, size=18, weight="bold", color="white")
            plt.close(fig)
            display(fig)
            
        chart_box = widgets.HBox([out_chart], layout=widgets.Layout(justify_content='center'))
        
        btn_restart = widgets.Button(
            description="SPILL IGJEN ⚡",
            layout=widgets.Layout(width='100%', height='70px', margin='40px 0 0 0')
        )
        btn_restart.style.button_color = '#2563eb'
        btn_restart.style.text_color = 'white'
        btn_restart.style.font_weight = '900'
        btn_restart.style.font_size = '20px'
        btn_restart.on_click(lambda b: self.start_game())
        
        self.content_box.children = [res_html, chart_box, btn_restart]
        self.header_html.value = "" # Skjul header streak på resultat

# --- LAUNCH ---
app = UberQuiz(original_quiz_data)
```
