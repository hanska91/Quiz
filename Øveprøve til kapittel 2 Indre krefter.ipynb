```python
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (for å unngå feil med ttkbootstrap)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# Opprinnelige quiz-data (lagres separat for reset)
original_quiz_data = [
    {
        "question": "Hva var Pangaea?",
        "options": [
            "A. En stor havbunnsrygg",
            "B. Et superkontinent",
            "C. En vulkansk hotspot",
            "D. En dyphavsgrop"
        ],
        "answer": "B",
        "explanation": "Pangaea var et superkontinent som eksisterte for ca. 300–200 millioner år siden, der alle dagens kontinenter var samlet. Det ble senere delt opp og drev fra hverandre."
    },
    {
        "question": "Hva er Alfred Wegener mest kjent for?",
        "options": [
            "A. Teorien om platedrift",
            "B. Teorien om havbunnsspredning",
            "C. Teorien om kontinentaldrift",
            "D. Teorien om jordrotasjon"
        ],
        "answer": "C",
        "explanation": "Alfred Wegener lanserte teorien om kontinentaldrift i 1912, som beskrev hvordan kontinentene en gang var samlet i Pangaea og senere drev fra hverandre."
    },
    {
        "question": "Hva kjennetegner Jordens indre kjerne?",
        "options": [
            "A. Flytende jern og nikkel",
            "B. Fast jern og nikkel",
            "C. Smeltet stein (magma)",
            "D. Silikatbergarter"
        ],
        "answer": "B",
        "explanation": "Indre kjerne består av fast jern og nikkel på grunn av det enorme trykket, selv om temperaturen er svært høy (over 5000 °C)."
    },
    {
        "question": "Hva består Jordens ytre kjerne av?",
        "options": [
            "A. Fast jern og nikkel",
            "B. Flytende jern og nikkel",
            "C. Smeltet silikat",
            "D. Magma og lava"
        ],
        "answer": "B",
        "explanation": "Ytre kjerne består av flytende jern og nikkel. Bevegelsen i dette laget skaper Jordens magnetfelt."
    },
    {
        "question": "Hva er mantelen?",
        "options": [
            "A. Et lag med flytende metall",
            "B. Et lag med faste bergarter som kan flyte over tid",
            "C. Jordens atmosfære",
            "D. Et lag med vann under jordskorpen"
        ],
        "answer": "B",
        "explanation": "Mantelen ligger mellom jordskorpen og kjernen og består av faste bergarter som kan deformeres og flyte langsomt. Her oppstår konveksjonstrømmer som driver platedriften."
    },
    {
        "question": "Hva er astenosfæren?",
        "options": [
            "A. Den ytre, stive delen av jorden",
            "B. Et plastisk, seigflytende lag i øvre mantel under litosfæren",
            "C. Den indre kjernen",
            "D. Et bånd av jordskjelv rundt Stillehavet"
        ],
        "answer": "B",
        "explanation": "Astenosfæren er et seigflytende (plastisk) lag i øvre mantel som litosfæreplatene glir på."
    },
    {
        "question": "Hva er litosfæren?",
        "options": [
            "A. Jordens atmosfære",
            "B. Den stive, ytre delen av jorden (jordskorpe + øverste mantel)",
            "C. Hele mantelen",
            "D. Den flytende kjernen"
        ],
        "answer": "B",
        "explanation": "Litosfæren er jordskorpen og den øverste, stive delen av mantelen. Den er oppdelt i plater."
    },
    {
        "question": "Hva menes med konveksjonstrømmer i mantelen?",
        "options": [
            "A. Vannstrømmer i jordskorpa",
            "B. Sirkulasjon av varm, seig bergmasse som stiger og synker",
            "C. Vindstrømmer i atmosfæren",
            "D. Magnetiske bølger i kjernen"
        ],
        "answer": "B",
        "explanation": "Konveksjonstrømmer er varmedrevne sirkulasjoner i mantelen der varmere materiale stiger og kjøligere synker. Dette er en hoveddrivkraft for platebevegelsene."
    },
    {
        "question": "Hvilke tre hovedtyper av platebevegelse finnes?",
        "options": [
            "A. Erosjon, forvitring og avsetning",
            "B. Divergens, konvergens og transformbevegelse",
            "C. Subduksjon, kollisjon og sprekkdannelse",
            "D. Løfting, folding og forkastning"
        ],
        "answer": "B",
        "explanation": "Platene kan bevege seg fra hverandre (divergens), mot hverandre (konvergens) eller glir sidelengs forbi hverandre langs en forkastning (transform)."
    },
    {
        "question": "Hva kjennetegner en havbunnsplate (oseanisk plate)?",
        "options": [
            "A. Tykk, lett og rik på granitt",
            "B. Tynn, tung og rik på basalt (mafiske mineraler)",
            "C. Tynn, lett og rik på kalkstein",
            "D. Tykk, tung og rik på granitt"
        ],
        "answer": "B",
        "explanation": "Havbunnsplater er tynnere og tettere enn kontinentalplater, og består hovedsakelig av mafiske, basaltiske bergarter. Ordet mafisk kombinerer ordene magnesium og ferro (jern). De har høyt innhold av tyngre metaller som jern, magnesium, mangan og titan m.fl, mens det er lavt innhold av lettere silikater og aluminium. Mafiske bergarter er vanligere i havbunnsskorpe enn i landfast kontinentalskorpe, havbunnsskorpe består primært av mafiske bergarter som er rike på magnesium (derav betegnelsen sima) - som eksempelvis bergartene basalt, diabas og gabbro."
    },
    {
        "question": "Hva kjennetegner Den atlantiske midthavsryggen?",
        "options": [
            "A. En kollisjonssone mellom to kontinentplater",
            "B. En spredningsrygg der ny havbunn dannes",
            "C. En transformforkastning",
            "D. En dyphavsgrop"
        ],
        "answer": "B",
        "explanation": "Den atlantiske midthavsryggen er en spredningssone (divergent plategrense) der magma stiger opp, størkner og danner ny havbunn. Den midtatlantiske ryggen er en undersjøisk fjellkjede (midthavsrygg) som går midt gjennom Atlanterhavet, fra Nordishavet til Sørishavet. Den omfatter den vulkanske midtsonen i en riftdal som oppstod da Afrika og Sør-Amerika drev fra hverandre for 100–200 millioner år siden. Ryggen ble til for rundt 55 millioner år siden og vokser fortsatt, slik at Amerika og Afrika/Europa fjerner seg fra hverandre. Vulkanaktivitet langs fjellkjeden har skapt Jan Mayen, Island, Asorene og Penedos de São Pedro e São Paulo nord for ekvator og Ascension, Tristan da Cunha, Gough Island og Bouvetøya sør for ekvator."
    },
    {
        "question": "Hva er en hotspot?",
        "options": [
            "A. Et område der to kontinenter kolliderer",
            "B. Et område med langvarig oppstrøm av varm mantel (søylestrøm) som skaper vulkanisme",
            "C. En sone med hyppige jordskjelv langs transformgrenser",
            "D. Et kaldt område i mantelen"
        ],
        "answer": "B",
        "explanation": "En hotspot er knyttet til en oppstrøm av varm mantel (mantelplume) som kan skape vulkanske kjeder, ofte uavhengig av plategrenser (f.eks. Hawaii)."
    },
    {
        "question": "Hva er en landform?",
        "options": [
            "A. Et mineral med bestemt krystallstruktur",
            "B. En form på jordoverflaten skapt av indre og/eller ytre prosesser",
            "C. En type bergart",
            "D. Et værfenomen"
        ],
        "answer": "B",
        "explanation": "Landformer er terrengformer som fjell, daler, platåer og sletter, skapt av indre krefter (tektonikk/vulkanisme) og ytre krefter (erosjon, is, vind, vann)."
    },
    {
        "question": "Hva menes med havbunnsspredning?",
        "options": [
            "A. At havet blir dypere over tid",
            "B. Danning av ny oseanisk skorpe ved midthavsrygger når magma størkner",
            "C. At kontinenter synker under havnivå",
            "D. At havstrømmer endrer retning"
        ],
        "answer": "B",
        "explanation": "Havbunnsspredning er prosessen der ny havbunn dannes ved midthavsrygger. Magnetiske striper i bergartene dokumenterer at havbunnen har spredt seg symmetrisk."
    },
    {
        "question": "Hva slags plategrense er San Andreas-forkastningen?",
        "options": [
            "A. Konvergent (subduksjon)",
            "B. Divergent (spredning)",
            "C. Transform (sideveis bevegelse)",
            "D. Kollisjon mellom to kontinentplater"
        ],
        "answer": "C",
        "explanation": "San Andreas-forkastningen i California er en transform grense mellom Stillehavsplaten og Den nordamerikanske platen (høyresidig strike-slip)."
    },
    {
        "question": "Hva er en dyphavsgrop?",
        "options": [
            "A. En undersjøisk fjellrygg ved spredningssone",
            "B. En smal, dyp sone der en plate synker ned under en annen (subduksjon)",
            "C. Et kollapset vulkankrater",
            "D. Et område med tykk kontinental skorpe"
        ],
        "answer": "B",
        "explanation": "Dyphavsgroper dannes ved subduksjon der oseaniske plater synker ned i mantelen, som Marianegropa i Stillehavet."
    },
    {
        "question": "Hva er en øybue?",
        "options": [
            "A. En kjede av koralløyer ved tropiske kyster",
            "B. En vulkansk øykjede dannet ved subduksjon av havbunn under havbunn",
            "C. En barriereøy dannet av bølger og strømmer",
            "D. En rekke øyer dannet ved iserosjon"
        ],
        "answer": "B",
        "explanation": "Vulkanske øybuer dannes der en oseanisk plate synker under en annen oseanisk plate. Smeltemasse stiger opp og danner en bue av vulkanske øyer (f.eks. Japan, Aleutene)."
    },
    {
        "question": "Hva er et jordskjelv?",
        "options": [
            "A. En langsom bevegelse i isbreer",
            "B. En plutselig frigjøring av energi når bergarter ryker langs en forkastning",
            "C. En type vulkanutbrudd",
            "D. Vindkast som rister bakken"
        ],
        "answer": "B",
        "explanation": "Jordskjelv oppstår når spenning i jordskorpen utløses ved brudd langs forkastninger, og energi sprer seg som seismiske bølger."
    },
    {
        "question": "Hvilket av følgende er et moderne bevis for platetektonikk (platedrift)?",
        "options": [
            "A. Lik nedbør på ulike kontinenter",
            "B. GPS-målinger som viser platebevegelser i sanntid",
            "C. At elver renner mot havet",
            "D. At fjellkjeder finnes i flere verdensdeler"
        ],
        "answer": "B",
        "explanation": "Satellitt-GPS måler årlige platebevegelser i cm/år og bekrefter platetektonikk. I tillegg støttes teorien av jordskjelvfordeling, vulkanbelter og havbunnens magnetiske striper."
    },
    {
        "question": "Hvilket funn brukte Wegener som bevis for kontinentaldrift?",
        "options": [
            "A. GPS-målinger",
            "B. Identiske fossiler og bergarter på kontinenter som nå ligger langt fra hverandre",
            "C. Havbunnens magnetiske striper",
            "D. Satellittbilder av midthavsrygger"
        ],
        "answer": "B",
        "explanation": "Wegener pekte på sammenfallende kystlinjer og like fossiler/bergarter på tvers av hav (f.eks. Mesosaurus i Afrika og Sør-Amerika) som bevis for kontinentaldrift."
    },
    {
        "question": "Hva er magma?",
        "options": [
            "A. Smeltet stein under jordoverflaten",
            "B. Smeltet stein på jordoverflaten",
            "C. Herdet vulkansk aske",
            "D. Et mineral rikt på jern"
        ],
        "answer": "A",
        "explanation": "Magma er smeltet (delvis smeltet) steinmasse med oppløste gasser under jordoverflaten."
    },
    {
        "question": "Hva er lava?",
        "options": [
            "A. Smeltet stein under jordoverflaten",
            "B. Smeltet stein som har nådd jordoverflaten",
            "C. Fint sand fra ørkener",
            "D. Herdet aske fra eksplosive utbrudd"
        ],
        "answer": "B",
        "explanation": "Når magma når overflaten ved et vulkanutbrudd, kalles det lava."
    },
    {
        "question": "Hva menes med landheving i Norge?",
        "options": [
            "A. At havet synker globalt",
            "B. Isostatisk heving av landet etter at innlandsisen smeltet",
            "C. At fjell vokser ved vulkanisme",
            "D. Kunstig oppfylling av land ved kysten"
        ],
        "answer": "B",
        "explanation": "Etter siste istid ble jordskorpen presset ned av isen. Når isen forsvant, begynte landet å heve seg igjen (isostatisk kompensasjon)."
    },
    {
        "question": "Hva er fokus (hyposenter) i et jordskjelv?",
        "options": [
            "A. Punktet på jordoverflaten rett over bruddet",
            "B. Punktet inne i jorden der bruddet starter",
            "C. Hele bruddsonen",
            "D. Senteret av en vulkan"
        ],
        "answer": "B",
        "explanation": "Fokus (hyposenter) er stedet i berggrunnen der jordskjelvet faktisk begynner og hvor energien først frigjøres. Energien fra de seismiske bølgene begynner å spre seg utover i alle retninger. Punktet på jordoverflaten som ligger rett over fokuset kalles episenteret. "
    },
    {
        "question": "Hva er episenteret i et jordskjelv?",
        "options": [
            "A. Den dypeste delen av bruddet",
            "B. Punktet på overflaten rett over fokus",
            "C. Hele området som rister",
            "D. Senteret av en tsunami"
        ],
        "answer": "B",
        "explanation": "Episenteret ligger vertikalt over fokus på jordoverflaten og er ofte stedet med sterkest rystelser nær kilden. Det er vanligvis i episenteret at skadene er mest omfattende, da de seismiske bølgene har en tendens til å være kraftigst ved overflaten. "
    },
    {
        "question": "Hva er en tsunami?",
        "options": [
            "A. En vanlig stormbølge skapt av vind",
            "B. En serie lange bølger skapt ved plutselig forflytning av vannmassene (f.eks. jordskjelv på havbunnen)",
            "C. En tidevannsbølge skapt av månen",
            "D. En bølge fra isbrekollaps i innsjøer"
        ],
        "answer": "B",
        "explanation": "Tsunamier utløses typisk av havbunnsjordskjelv, skred eller vulkanutbrudd som fortrenger store vannmasser."
    },
    {
        "question": "Hva kjennetegner en sammensatt vulkan (stratovulkan)?",
        "options": [
            "A. Bred, slak og bygget av tynntflytende basaltlava",
            "B. Bratt, lagdelt av lava og tefra, ofte eksplosiv",
            "C. Flat, lav ås av askelag",
            "D. Undersjøisk vulkan uten utbrudd"
        ],
        "answer": "B",
        "explanation": "Stratovulkaner består av vekslende lag med lava, aske og pyroklastisk materiale. De har ofte seigere (andesitt/rhyolittisk) magma og kan gi eksplosive utbrudd. Eksempler på stratovulkaner inkluderer Vesuv i Italia, Fuji i Japan, Cotopaxi i Ecuador, Beerenberg på Jan Mayen, og Mount St. Helens i USA, som alle er kjegleformede vulkaner bygget opp av lag av lava og aske. "
    },
    {
        "question": "Hva kjennetegner en skjoldvulkan?",
        "options": [
            "A. Bratt konisk form og eksplosive utbrudd",
            "B. Bred, slak profil bygd opp av tynntflytende basaltlava",
            "C. Domformet, seig lava og lav aktivitet",
            "D. Undersjøisk vulkan ved subduksjonssoner"
        ],
        "answer": "B",
        "explanation": "Et eksempel på en skjoldvulkan er Mauna Loa på Hawaii. Skjoldvulkaner er store, flate vulkaner med slake skråninger, som formes av tyntflytende, basaltisk lava. Andre eksempler inkluderer Kīlauea (også på Hawaii), Fagradalsfjall på Island, og den undersjøiske vulkanen Tamumassivet. "
    },
    {
        "question": "Hva er magmatiske (størknings-) bergarter?",
        "options": [
            "A. Bergarter dannet ved trykk og temperatur uten smelte",
            "B. Bergarter dannet når magma eller lava størkner",
            "C. Bergarter dannet ved avsetning og sementering av sedimenter",
            "D. Bergarter dannet av organiske rester i myr"
        ],
        "answer": "B",
        "explanation": "Magmatiske bergarter dannes ved størkning av magma (Dypbergarter er magmatiske bergarter som størkner dypt under jordoverflaten under høyt trykk og langsom avkjøling, noe som gir store, synlige krystaller. Gangbergarter er magmatiske bergarter som størkner i ganger eller sprekker i jordskorpen, typisk mellom dypbergarter og dagbergarter.) eller lava (dagbergarter). Diabas er et eksempel på en gangbergart hvor tilsvarende dypbergart er gabbro og tilsvarende dagbergart er basalt."
    },
    {
        "question": "Hva er sedimentære (avsetnings-) bergarter?",
        "options": [
            "A. Bergarter dannet ved størkning av magma",
            "B. Bergarter dannet ved omvandling av eldre bergarter",
            "C. Bergarter dannet ved avsetning, sammenpressing og sementering av sedimenter",
            "D. Bergarter dannet i kjernen"
        ],
        "answer": "C",
        "explanation": "Sedimentære bergarter dannes når løsmasser avsettes i lag, kompakteres og sementeres. De kan inneholde fossiler (f.eks. sandstein, kalkstein, leirskifer)."
    },
    {
        "question": "Hva er metamorfe (omdannede) bergarter?",
        "options": [
            "A. Bergarter dannet direkte fra magma",
            "B. Bergarter omdannet fra eldre bergarter ved trykk/temperatur uten fullstendig smelting",
            "C. Bergarter dannet av organisk materiale",
            "D. Bergarter dannet i elveleier"
        ],
        "answer": "B",
        "explanation": "Metamorfe bergarter oppstår når eksisterende bergarter omdannes ved høyt trykk og/eller temperatur, f.eks. gneis fra granitt, marmor fra kalkstein."
    },
    {
        "question": "Hva er en bergart?",
        "options": [
            "A. Et enkelt grunnstoff",
            "B. En naturlig blanding (aggregat) av ett eller flere mineraler",
            "C. Smeltet stein i mantelen",
            "D. Herdet lava uten mineraler"
        ],
        "answer": "B",
        "explanation": "En bergart er en naturlig sammensetning av mineraler (f.eks. granitt består av kvarts, feltspat og glimmer)."
    },
    {
        "question": "Hva er et mineral?",
        "options": [
            "A. Et organisk stoff med variabel sammensetning",
            "B. Et naturlig, uorganisk fast stoff med bestemt kjemisk sammensetning og krystallstruktur",
            "C. En type bergart",
            "D. Små biter av stein uten struktur"
        ],
        "answer": "B",
        "explanation": "Mineraler er byggesteinene i bergarter og har definert kjemisk sammensetning og krystallstruktur (f.eks. kvarts, feltspat)."
    }

]

# Initialiser GUI
root = tk.Tk()
root.title("Quiz App")
root.geometry("600x550")
style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 16))
style.configure("TButton", font=("Helvetica", 14))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=500, padding=10)
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10, wraplength=500, justify="left")
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=400, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    qs_label.config(text=question["question"])
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        button = ttk.Button(
            choices_frame,
            text=option,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="")
    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index):
    global score
    question = quiz_data[current_question]
    selected_letter = chr(65 + choice_index)
    correct_letter = question["answer"]
    correct_index = ord(correct_letter) - 65
    previous = user_answers[current_question]

    # Juster score hvis bytter svar
    if previous is not None and chr(65 + previous) == correct_letter:
        score -= 1

    user_answers[current_question] = choice_index

    if selected_letter == correct_letter:
        score += 1
        feedback_label.config(text="Riktig!\n" + question['explanation'], foreground="green")
    else:
        feedback_label.config(
    text=f"Galt! Det riktige svaret er {correct_letter}.\n{question['explanation']}",
    foreground="red"
        )

    score_label.config(text="Score: {}/{}".format(score, len(quiz_data)))
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index):
    question = quiz_data[current_question]
    correct_index = ord(question["answer"]) - 65

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
