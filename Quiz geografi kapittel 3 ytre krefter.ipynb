```python
# -*- coding: utf-8 -*-
import locale
import tkinter as tk
from tkinter import messagebox, ttk
import random

# Lokalitet (noen temaer i ttkbootstrap feiler hvis LC_ALL ikke settes først)
try:
    locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
except locale.Error:
    locale.setlocale(locale.LC_ALL, '')

from ttkbootstrap import Style
locale.setlocale(locale.LC_ALL, '')

# --- Spørsmålsbank ---
# NB: Bruker nøkler: q (tekst), options (4 alternativer), answer (indeks 0–3), explain (forklaring)
original_quiz_data = [
    {
        "q": "Hva menes med forvitring?",
        "options": [
            "Transport av løsmasser med vind, vann eller is",
            "Nedbrytning av bergarter på jordoverflaten ved fysiske eller kjemiske prosesser",
            "Oppbygning av nye bergarter ved sedimentasjon og herdning",
            "Smelting av isbreer som fører til havnivåstigning",
        ],
        "answer": 1,
        "explain": "Forvitring er nedbrytning av bergarter på overflaten – enten mekanisk (fysisk) eller kjemisk."
    },
    {
        "q": "Hva kjennetegner mekanisk forvitring?",
        "options": [
            "Mineraler reagerer med vann og CO₂, og omdannes kjemisk",
            "Bergarten brytes fysisk opp uten kjemiske endringer",
            "Løsmasser transporteres av rennende vann",
            "Bergart oppløses fullstendig og blir til slam",
        ],
        "answer": 1,
        "explain": "Mekanisk forvitring er fysisk oppsprekking/nedknusing uten kjemisk endring av mineralene."
    },
    {
        "q": "Hva kjennetegner kjemisk forvitring?",
        "options": [
            "Bergart sprekker opp på grunn av trykkavlastning",
            "Mineraler reagerer med vann/luft og oppløses eller omdannes",
            "Is og snø presser bergarten sammen",
            "Elver transporterer blokker langs bunnen",
        ],
        "answer": 1,
        "explain": "Kjemisk forvitring endrer mineralene gjennom reaksjoner, for eksempel oppløsning eller hydrolyse."
    },
    {
        "q": "Hva er frostforvitring?",
        "options": [
            "Stein varmes opp og utvider seg i sollys",
            "Vann fryser i sprekker, utvider seg og sprenger løs blokker",
            "Mineraler oksiderer og blir sprø",
            "Isbreer skurer berggrunnen",
        ],
        "answer": 1,
        "explain": "Vann i sprekker fryser, utvider seg ~9 %, og kan sprenge løs stein."
    },
    {
        "q": "Hva er erosjon?",
        "options": [
            "Prosessen der løsmasser dannes ved forvitring",
            "Prosessen der materiale slites løs, fraktes og avsettes av ytre krefter",
            "Kun transport av slam i elver",
            "Kun bølgepåvirkning på kyst",
        ],
        "answer": 1,
        "explain": "Erosjon omfatter både nedsliting, transport og avsetning av materiale av vind, vann og is."
    },
    {
        "q": "Hvordan dannes en V-dal?",
        "options": [
            "Ved at isbreer graver ut dype daler i fjell",
            "Ved elveerosjon som graver seg ned og skaper V-form",
            "Ved jordskjelv som kløyver landskapet",
            "Ved bølger som former kysten",
        ],
        "answer": 1,
        "explain": "Elver graver seg ned i løsmasser/berg og danner V-formede daler."
    },
    {
        "q": "Hva er et elvegjel (canyon)?",
        "options": [
            "En grunn, bred elveslette",
            "En dyp, trang dal med bratte sider dannet av elveerosjon",
            "Et sideleie som bare brukes ved flom",
            "Et delta ytterst ved kysten",
        ],
        "answer": 1,
        "explain": "Gjel/canyon er en dyp og trang dal hvor elva har skåret seg kraftig ned."
    },
    {
        "q": "Hvordan dannes en jettegryte?",
        "options": [
            "Av en isbre som løfter stein",
            "Av vind som pusser fjelloverflaten",
            "Ved at steiner roterer i en virvel og sliper runde groper i fjell",
            "Ved vulkanske eksplosjoner",
        ],
        "answer": 2,
        "explain": "Virvlende vann med stein fungerer som boremaskin og sliper runde gryter i fjell."
    },
    {
        "q": "Hva er en elvevifte?",
        "options": [
            "Vifteformet avsetning der elv mister fart i brattere terreng",
            "Vifteformet avsetning der elv går fra bratt til slakt terreng",
            "Havvifte utenfor kontinentalskråningen",
            "En kunstig kanal for flom",
        ],
        "answer": 1,
        "explain": "Når elva kommer ut på slakere terreng, mister den transportevne og avsetter i vifteform."
    },
    {
        "q": "Hva er et meanderløp?",
        "options": [
            "Rett elveløp i bratt terreng",
            "Forgrenet elveløp (braided)",
            "Svingete elveløp med store buer",
            "Underjordisk elveløp i kalkstein",
        ],
        "answer": 2,
        "explain": "Meandere er store svinger i elva, typisk på flate elvesletter."
    },
    {
        "q": "Hva er en elveslette?",
        "options": [
            "Et flatt avsetningsområde langs elva, ofte dannet ved gjentatte flommer",
            "Et bratt heng langs en elv",
            "Et delta i innsjø",
            "En jettegryte med vannoppfylling",
        ],
        "answer": 0,
        "explain": "Elvesletter dannes ved at elva avsetter materiale over tid i flomsonen."
    },
    {
        "q": "Hva er en kroksjø?",
        "options": [
            "En brakkvannslagune ved kysten",
            "En innsjø dannet når en meandersving kuttes av fra hovedløpet",
            "En innsjø dannet av isblokk i sandur",
            "En innsjø i en botn",
        ],
        "answer": 1,
        "explain": "Når en meander avsnøres, blir sidebuen til en halvmåneformet kroksjø."
    },
    {
        "q": "Hva er et delta?",
        "options": [
            "Avsetningsform der elv renner ut i stillestående vann og mister transportevne",
            "En undersjøisk canyon",
            "En ravine i marin leire",
            "En endemorene",
        ],
        "answer": 0,
        "explain": "Delta dannes der elv møter stillestående vann (innsjø/hav) og avsetter finstoff, sand og grus."
    },
    {
        "q": "Hva betyr slamtransport i en elv?",
        "options": [
            "Transport av store blokker som ruller på bunnen",
            "Transport av oppløst materiale (ions)",
            "Transport av små partikler (leire/silt) som holdes svevende i vannet",
            "Ingen transport, kun avsetning",
        ],
        "answer": 2,
        "explain": "Slamtransport er finstoff (leire/silt) som bæres i suspensjon av strømmen."
    },
    {
        "q": "Hva betyr bunntransport i en elv?",
        "options": [
            "Partikler som flyter i overflaten",
            "Oppløste salter i vannet",
            "Sand, grus og stein som ruller, glir eller hopper langs bunnen",
            "Leirepartikler som holdes svevende",
        ],
        "answer": 2,
        "explain": "Bunntransport er grovere materiale som fraktes nær bunnen ved rulling/hopping (saltasjon)."
    },
    {
        "q": "Hva er en botn (cirque)?",
        "options": [
            "En halvsirkelformet grop skapt av en liten bre ved fjellsiden",
            "En elvevifte i høyfjellet",
            "En liten fjordarm",
            "Et delta foran en bre",
        ],
        "answer": 0,
        "explain": "Botn er en skålformet grop i fjell skapt av en botnbre."
    },
    {
        "q": "Hva er en dalbre?",
        "options": [
            "En liten bre i en botn",
            "En isbre som fyller og beveger seg i en dal",
            "En bre som dekker et platå",
            "En innlandsis",
        ],
        "answer": 1,
        "explain": "Dalbreer følger topografien i dalene og strømmer nedover."
    },
    {
        "q": "Hva er en botnbre?",
        "options": [
            "En svært stor ismasse som dekker kontinenter",
            "En liten bre som ligger i en botn",
            "En bre som flyter på havet",
            "En bre med sterk bunntransport",
        ],
        "answer": 1,
        "explain": "Botnbre er liten, ligger i en botn (cirque) og eroderer skålformede nisjer."
    },
    {
        "q": "Hva er en platåbre?",
        "options": [
            "Bre som dekker et fjellplatå og sender strømmer ned i daler",
            "Bre som kun ligger i skyggepartier",
            "Bre som skurer bunnmorene",
            "Bre som flyter i havnivå",
        ],
        "answer": 0,
        "explain": "Platåbre dekker store flater (platå), ofte med utløpere ned i daler."
    },
    {
        "q": "Hva er innlandsis?",
        "options": [
            "Is som legger seg i elveløp om vinteren",
            "Større ismasse som dekker store landområder, som Grønland",
            "Tynn havis nær kysten",
            "Ferskvannsis i innsjøer",
        ],
        "answer": 1,
        "explain": "Innlandsis er kontinentdekkende is, f.eks. Antarktis og Grønland."
    },
    {
        "q": "Hva er Raet i Norge et eksempel på?",
        "options": [
            "En fjordterskel",
            "En stor morenerygg fra siste istid",
            "En elvevifte",
            "En strandflate",
        ],
        "answer": 1,
        "explain": "Raet er en sammenhengende endemorene-/randmorene-rygg fra siste istid."
    },
    {
        "q": "Hva menes med skuring under en isbre?",
        "options": [
            "Smelting av isen på overflaten",
            "Slipende virkning når stein i isen riper og polerer underlaget",
            "Oppsprekking av isens overflate",
            "Avsnøring av dødis",
        ],
        "answer": 1,
        "explain": "Skuring: stein i/fryset til isen virker som sandpapir mot fjellet og lager skuringsstriper."
    },
    {
        "q": "Hva menes med plukking under en isbre?",
        "options": [
            "Breen plukker opp vann fra smeltingen",
            "Breen fryser fast i berg og river/“plukker” løs blokker",
            "Breen plukker opp sand fra delta",
            "Breen plukker opp løv og organisk materiale",
        ],
        "answer": 1,
        "explain": "Plukking: trykkendringer og frost gjør at isen løsner og tar med blokker fra berggrunnen."
    },
    {
        "q": "Hva kjennetegner en U-dal?",
        "options": [
            "Tverrprofil formet som en U etter breerosjon",
            "Tverrprofil formet som en V etter elveerosjon",
            "En dyp canyon i kalkstein",
            "En skrå elveslette",
        ],
        "answer": 0,
        "explain": "Breer utvider og overdyper daler til brede U-profiler."
    },
    {
        "q": "Hva er en fjord?",
        "options": [
            "En grunn bukt dannet av bølger",
            "En dyp, smal havarm i en nedskurt U-dal",
            "En innsjø avsnørt fra is",
            "Et elvedelta i havet",
        ],
        "answer": 1,
        "explain": "Fjorder er druknede U-daler – skapt av breer og senere fylt av hav."
    },
    {
        "q": "Hva er et isfrontdelta?",
        "options": [
            "Delta dannet i havet foran en stillestående eller tilbaketrekkende brefront",
            "Delta i en kroksjø",
            "Avsetning bak en endemorene",
            "En strandvoll i leire",
        ],
        "answer": 0,
        "explain": "Smeltevann fra breen avsetter sand/grus foran brefronten – ofte store, flate deltaformer."
    },
    {
        "q": "Hva er bunnmorene?",
        "options": [
            "Løsmasser som ligger under breen og blir liggende igjen når isen smelter",
            "Rygg foran breen",
            "Sandstrand ved kysten",
            "Elveavsetning i sving",
        ],
        "answer": 0,
        "explain": "Bunnmorene er usortert materiale (till) avsatt under isen."
    },
    {
        "q": "Hva er endemorene?",
        "options": [
            "Rygg av usorterte løsmasser avsatt ved brefrontens stillstand",
            "Leire avsatt i havet",
            "Sanddyner ved kysten",
            "Elvevifte i bratt terreng",
        ],
        "answer": 0,
        "explain": "Endemorene bygges opp foran brefronten når isen står relativt i ro."
    },
    {
        "q": "Hva er en strandflate?",
        "options": [
            "En flat bølge- og frostformet kystslette nær havnivå",
            "En smal strandvoll av skjell",
            "Et delta i kystsonen",
            "En jettegryte ved kyst",
        ],
        "answer": 0,
        "explain": "Strandflata er en lav, relativt flat kystsone formet av bølger, frost og landheving."
    },
    {
        "q": "Hva menes med fjordkyst?",
        "options": [
            "Kyst dominert av brede sandstrender",
            "Kyst der daler gravd av breer er fylt av hav og danner fjorder",
            "Kyst med mye korallrev",
            "Kyst som kun består av strandflater",
        ],
        "answer": 1,
        "explain": "Fjordkyst er preget av dype, smale fjorder i nedskurte U-daler."
    },
    {
        "q": "Hva er en løsmasskekyst?",
        "options": [
            "Kyst dominert av løsmasser som sand, grus og leire",
            "Kyst kun av hardt fjell",
            "Kyst formet av isbreer",
            "Kyst uten strandprosesser",
        ],
        "answer": 0,
        "explain": "Løsmasskekyster preges av bølge- og strømtransport av sand, grus og leire."
    },
    {
        "q": "Hva er en fjordsjø?",
        "options": [
            "En innsjø avsnørt fra havet i en tidligere fjord",
            "En dyp sjø på kontinentalsokkelen",
            "En innsjø i en botn",
            "En innsjø i en jettegryte",
        ],
        "answer": 0,
        "explain": "Fjordsjø oppstår når en fjord avsnøres, f.eks. av landheving eller terskler."
    },
    {
        "q": "Hva er marin grense (MG)?",
        "options": [
            "Laveste nivå havet nådde etter istid",
            "Høyeste nivå havet nådde etter istid, markert i terrenget",
            "Gjennomsnittlig tidevannsnivå",
            "Grensen for ferskvann i fjorder",
        ],
        "answer": 1,
        "explain": "Marin grense er høyeste havnivå etter siste istid, forskjellig fra sted til sted pga. landheving."
    },
    {
        "q": "Hva er marin leire?",
        "options": [
            "Leire avsatt i en elv",
            "Leire avsatt i havet etter istid, ofte saltpåvirket",
            "Leire avsatt i en kroksjø",
            "Leire dannet ved forvitring av granitt",
        ],
        "answer": 1,
        "explain": "Marin leire ble avsatt i saltvann; kan bli ustabil hvis saltet vaskes ut."
    },
    {
        "q": "Hva er en ravine?",
        "options": [
            "En smal, dyp erosjonsgrøft i finkornige masser (ofte marin leire)",
            "En bue i en meander",
            "En strandvoll på kysten",
            "En morenerygg",
        ],
        "answer": 0,
        "explain": "Raviner dannes ved at rennende vann graver seg ned i løsmasser og skaper bratte sider."
    },
    {
        "q": "Hva kjennetegner kvikkleire?",
        "options": [
            "Leire som alltid er tørr og stabil",
            "Marin leire som ved rystelser kan kollapse og bli nesten flytende",
            "Leire som er issementert",
            "Leire som bare finnes i ørken",
        ],
        "answer": 1,
        "explain": "Når saltet som binder leiren vaskes ut, kan strukturen kollapse og gi kvikkleireskred."
    },
    {
        "q": "Hva er forskjellen på skred og ras (i dagligtale)?",
        "options": [
            "Ingen forskjell – identiske begreper i all bruk",
            "Ras brukes ofte om bratte, plutselige hendelser; skred er bredere (snø, jord, stein, leire)",
            "Skred er alltid små; ras er alltid store",
            "Ras gjelder bare snø; skred gjelder bare stein",
        ],
        "answer": 1,
        "explain": "I praksis brukes ‘skred’ som samlebetegnelse, mens ‘ras’ ofte brukes om bratte, plutselige utglidninger."
    }
    
]

# --- GUI ---
root = tk.Tk()
root.title("Quiz App")
root.geometry("650x560")

style = Style(theme="flatly")
style.configure("TLabel", font=("Helvetica", 14))
style.configure("TButton", font=("Helvetica", 13))

# Globale variabler
score = 0
current_question = 0
choice_btns = []
quiz_data = []
user_answers = []

# Widgets
qs_label = ttk.Label(root, anchor="center", wraplength=560, padding=10, justify="center")
qs_label.pack(pady=10)

choices_frame = ttk.Frame(root)
choices_frame.pack(pady=10, fill=tk.X)

feedback_label = ttk.Label(root, anchor="center", padding=10, wraplength=560, justify="left")
feedback_label.pack(pady=10)

score_label = ttk.Label(root, text="", anchor="center", padding=10)
score_label.pack(pady=10)

nav_frame = ttk.Frame(root)
nav_frame.pack(pady=10)

prev_btn = ttk.Button(nav_frame, text="Tilbake", command=lambda: previous_question(), state="disabled")
prev_btn.pack(side="left", padx=10)

next_btn = ttk.Button(nav_frame, text="Neste", command=lambda: next_question(), state="disabled")
next_btn.pack(side="left", padx=10)

retry_btn = ttk.Button(root, text="Prøv igjen", command=lambda: restart_quiz(), state="disabled")
retry_btn.pack(pady=10)

progress = ttk.Progressbar(root, orient="horizontal", length=420, mode="determinate")
progress.pack(pady=10)

# --- Funksjoner ---

def restart_quiz():
    """Resetter quiz og starter på nytt."""
    global score, current_question, choice_btns, quiz_data, user_answers
    score = 0
    current_question = 0
    quiz_data = original_quiz_data.copy()
    random.shuffle(quiz_data)
    user_answers = [None] * len(quiz_data)
    progress.config(maximum=len(quiz_data), value=0)
    retry_btn.config(state="disabled")
    next_btn.config(state="disabled")
    prev_btn.config(state="disabled")
    show_question()

def show_question():
    """Viser nåværende spørsmål og alternativer."""
    global choice_btns
    for btn in choice_btns:
        btn.destroy()
    choice_btns = []

    question = quiz_data[current_question]
    # BRUK "q" i stedet for "question"
    qs_label.config(text=f"Spørsmål {current_question+1} av {len(quiz_data)}:\n{question['q']}")
    options = question["options"]

    selected_index = user_answers[current_question]

    for i, option in enumerate(options):
        btn_text = f"{'ABCD'[i]}) {option}"
        button = ttk.Button(
            choices_frame,
            text=btn_text,
            command=lambda i=i: check_answer(i),
            style="TButton"
        )
        button.pack(pady=5, fill=tk.X)
        choice_btns.append(button)

    feedback_label.config(text="", foreground="black")
    score_label.config(text=f"Score: {score}/{len(quiz_data)}")
    prev_btn.config(state="normal" if current_question > 0 else "disabled")
    next_btn.config(state="disabled")
    retry_btn.config(state="disabled")

    # Hvis brukeren har svart før, vis fargefeedback og aktiver 'Neste'
    if selected_index is not None:
        apply_feedback_colors(selected_index)
        next_btn.config(state="normal")

def check_answer(choice_index: int):
    """Håndterer brukersvar, oppdaterer poeng og feedback."""
    global score
    question = quiz_data[current_question]
    correct_index = question["answer"]  # <-- indeks (0–3), ikke bokstav
    previous = user_answers[current_question]

    # Juster score hvis brukeren endrer fra korrekt til noe annet
    if previous is not None and previous == correct_index:
        score -= 1

    user_answers[current_question] = choice_index

    if choice_index == correct_index:
        score += 1
        feedback_label.config(text=f"✅ Riktig!\n{question['explain']}", foreground="green4")
    else:
        correct_letter = "ABCD"[correct_index]
        feedback_label.config(
            text=f"❌ Galt! Det riktige svaret er {correct_letter}.\n{question['explain']}",
            foreground="red4"
        )

    score_label.config(text=f"Score: {score}/{len(quiz_data)}")
    progress['value'] = current_question + 1

    apply_feedback_colors(choice_index)
    next_btn.config(state="normal")

def apply_feedback_colors(selected_index: int):
    """Farger knapper basert på korrekt/feil og deaktiverer dem."""
    question = quiz_data[current_question]
    correct_index = question["answer"]  # <-- indeks (0–3)

    for i, button in enumerate(choice_btns):
        button.config(state="disabled")
        if i == correct_index:
            button.config(style="success.TButton")
        elif i == selected_index:
            button.config(style="danger.TButton")
        else:
            button.config(style="secondary.TButton")

def next_question():
    """Går til neste spørsmål, eller fullfører."""
    global current_question
    if current_question < len(quiz_data) - 1:
        current_question += 1
        show_question()
    else:
        messagebox.showinfo(
            "Quiz gjennomført",
            f"Du fikk {score} av {len(quiz_data)} riktige.\nTakk for at du deltok!"
        )
        retry_btn.config(state="normal")
        next_btn.config(state="disabled")
        prev_btn.config(state="disabled")

def previous_question():
    """Går til forrige spørsmål."""
    global current_question
    if current_question > 0:
        current_question -= 1
        show_question()

# Start første gang
restart_quiz()
root.mainloop()
```
