```python
import ipywidgets as widgets
from IPython.display import display, HTML, Javascript, clear_output, display
import random
import matplotlib.pyplot as plt

# --- CSS: FIXED GOD TIER ---
custom_css = """
<style>
    /* --- 1. LEVENDE BAKGRUNN --- */
    .master-container {
        background: linear-gradient(-45deg, #ee7752, #e73c7e, #23a6d5, #23d5ab);
        background-size: 400% 400%;
        animation: gradientBG 15s ease infinite;
        padding: 50px;
        border-radius: 20px;
        font-family: 'Inter', 'Segoe UI', Helvetica, Arial, sans-serif;
    }

    @keyframes gradientBG {
        0% { background-position: 0% 50%; }
        50% { background-position: 100% 50%; }
        100% { background-position: 0% 50%; }
    }

    /* --- 2. GLASS-KORTET --- */
    .quiz-card-container {
        background-color: rgba(255, 255, 255, 0.92);
        backdrop-filter: blur(15px);
        border-radius: 24px;
        box-shadow: 0 25px 80px rgba(0,0,0,0.3);
        width: 100%;
        margin: 0 auto;
        overflow: visible; /* VIKTIG: Lar konfettien flyte fritt */
        border: 1px solid rgba(255, 255, 255, 0.6);
        position: relative;
    }

    /* Header - FIKSET TEKSTKUTT */
    .quiz-header-widget {
        background: transparent;
        padding: 30px 0 10px 0;
        text-align: center;
        width: 100%;
        border-bottom: 2px solid rgba(0,0,0,0.05);
    }
    
    .quiz-title {
        margin: 0;
        font-size: 48px;
        font-weight: 900;
        letter-spacing: 5px;
        text-transform: uppercase;
        
        /* Gradient tekst */
        background: linear-gradient(to right, #2563eb, #9333ea);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        
        /* FIKS FOR KUTTET TEKST: Mer høyde og padding */
        line-height: 1.4; 
        padding-bottom: 10px; 
        display: inline-block; /* Sikrer at boksen omslutter teksten korrekt */
    }

    /* Innhold */
    .quiz-content-area {
        padding: 20px 50px 50px 50px;
        position: relative;
        z-index: 1;
    }

    /* Animasjoner */
    .slide-in-up {
        animation: slideUp 0.6s cubic-bezier(0.16, 1, 0.3, 1) forwards;
        opacity: 0;
        transform: translateY(20px);
    }
    @keyframes slideUp { to { opacity: 1; transform: translateY(0); } }

    /* Progresjon */
    .progress-wrapper {
        margin-top: 15px;
        margin-bottom: 40px;
    }
    .progress-bg {
        width: 100%;
        background: #e5e7eb;
        height: 12px;
        border-radius: 6px;
    }
    .progress-fill {
        background: linear-gradient(90deg, #2563eb, #9333ea);
        height: 12px;
        border-radius: 6px;
        box-shadow: 0 0 15px rgba(147, 51, 234, 0.4);
        transition: width 0.6s cubic-bezier(0.4, 0, 0.2, 1);
    }

    /* Spørsmål */
    .question-text {
        font-size: 28px;
        color: #1f2937;
        font-weight: 800;
        margin-bottom: 35px;
        line-height: 1.3;
        text-align: center;
    }

    /* Knapper */
    .widget-button { 
        box-sizing: border-box !important;
        font-size: 20px !important; 
        border-radius: 16px !important; 
        border: 2px solid #e5e7eb !important; 
        background-color: white !important;
        color: #4b5563 !important;
        margin: 10px 0 !important; 
        padding: 22px 30px !important;
        height: auto !important;
        width: 98% !important; 
        margin-left: 1% !important;
        text-align: left !important;
        outline: none !important;
        box-shadow: 0 4px 6px rgba(0,0,0,0.03) !important;
        transition: all 0.3s cubic-bezier(0.25, 0.8, 0.25, 1) !important;
    }
    
    .widget-button:hover:enabled {
        border-color: #8b5cf6 !important;
        transform: translateY(-4px) scale(1.005);
        box-shadow: 0 15px 30px rgba(139, 92, 246, 0.2) !important;
        z-index: 10;
    }

    /* Feedback Box */
    .feedback-box {
        margin-top: 30px;
        padding: 30px;
        border-radius: 16px;
        background-color: #f9fafb;
        border-left: 8px solid #ccc;
        font-size: 18px;
        line-height: 1.6;
        animation: slideUp 0.4s ease-out;
    }
    
    /* Resultat Melding */
    .result-msg {
        font-size: 24px;
        margin-top: 25px;
        font-weight: 700;
        padding: 30px;
        border-radius: 20px;
        text-align: center;
        box-shadow: 0 10px 30px rgba(0,0,0,0.1);
        position: relative;
        z-index: 20; /* Foran konfetti */
    }

    /* --- KONFETTI MOTOR (FIKSET) --- */
    .confetti-container {
        position: absolute;
        top: 0;
        left: 0;
        width: 100%;
        height: 100%;
        pointer-events: none; /* Lar deg klikke gjennom konfettien */
        z-index: 100; /* Høyere enn alt annet */
        overflow: hidden;
    }
    
    .confetti {
        position: absolute;
        width: 12px; 
        height: 12px;
        opacity: 0.8;
        top: -20px; /* Starter over kanten */
        animation: fall linear forwards;
    }
    
    @keyframes fall {
        0% { transform: translateY(-20px) rotate(0deg); opacity: 1;}
        100% { transform: translateY(600px) rotate(720deg); opacity: 0; }
    }
    
    /* Styling for Flash/Highlight */
    .correct-flash { animation: flash-success 0.6s forwards !important; font-weight: bold !important; z-index: 20 !important; }
    @keyframes flash-success {
        0% { background-color: white; border-color: #10b981; }
        50% { background-color: #10b981; border-color: #10b981; color: white; transform: scale(1.02); }
        100% { background-color: #10b981; border-color: #10b981; color: white; }
    }
    .wrong-answer { background-color: #ef4444 !important; color: white !important; border-color: #ef4444 !important; }
    .fasit-highlight { background-color: #ecfdf5 !important; color: #047857 !important; border-color: #10b981 !important; border-width: 3px !important; font-weight: bold !important; }

</style>
"""

# --- DATA ---
original_quiz_data = [
  {
            "question": "Hva er et kromosom?",
            "options": ["A. En type protein", "B. En del av cellemembranen", "C. En struktur som inneholder DNA og protein", "D. En type lipid"],
            "answer": "C",
            "explanation": "Et kromosom er en struktur som inneholder DNA og proteiner, og finnes i cellekjernen."
        },
        {
            "question": "DNA-tråden er 2 meter lang hvis den strekkes helt ut. En celle er mindre enn en tidels millimeter i diameter. Forklar hvordan det er mulig å få så mye DNA inni hver enkelt celle.",
            "options": ["A. DNA-et er pakket tett sammen i kromosomer", "B. DNA-et er flytende i cytoplasmaet", "C. DNA-et er løst pakket i cellemembranen", "D. DNA-et er pakket i mitokondriene"],
            "answer": "A",
            "explanation": "DNA-et er pakket tett sammen i kromosomer ved hjelp av proteiner som kalles histoner, noe som gjør det mulig å få plass til mye DNA i hver celle."
        },
        {
            "question": "Hvor finner vi DNA-et?",
            "options": ["A. I cellemembranen", "B. I cellekjernen", "C. I ribosomene", "D. I mitokondriene"],
            "answer": "B",
            "explanation": "DNA-et finnes hovedsakelig i cellekjernen."
        },
        {
            "question": "Hva er DNA?",
            "options": ["A. En type karbohydrat", "B. En type protein", "C. En type nukleinsyre", "D. En type lipid"],
            "answer": "C",
            "explanation": "DNA er en type nukleinsyre som inneholder den genetiske informasjonen i cellene."
        },
        {
            "question": "Hvordan er et DNA-molekyl bygget opp?",
            "options": ["A. Enkelttrådet med ribose", "B. Dobbelttrådet med deoksyribose", "C. Enkelttrådet med deoksyribose", "D. Dobbelttrådet med ribose"],
            "answer": "B",
            "explanation": "Et DNA-molekyl er dobbelttrådet og består av deoksyribose, fosfatgrupper og nitrogenbaser."
        },
        {
            "question": "Hva koder genene i DNA for?",
            "options": ["A. Karbohydrater", "B. Lipider", "C. Proteiner", "D. Vitaminer"],
            "answer": "C",
            "explanation": "Genene i DNA koder for proteiner som utfører ulike funksjoner i kroppen."
        },
        {
            "question": "Hvilke 4 nitrogenbaser finnes i DNA?",
            "options": ["A. Adenin, Tymin, Guanin, Cytosin", "B. Adenin, Uracil, Guanin, Cytosin", "C. Adenin, Tymin, Guanin, Uracil", "D. Adenin, Cytosin, Guanin, Uracil"],
            "answer": "A",
            "explanation": "De fire nitrogenbasene i DNA er Adenin, Tymin, Guanin og Cytosin."
        },
        {
            "question": "Hvilke 4 nitrogenbaser finnes i RNA?",
            "options": ["A. Adenin, Tymin, Guanin, Cytosin", "B. Adenin, Uracil, Guanin, Cytosin", "C. Adenin, Tymin, Guanin, Uracil", "D. Adenin, Cytosin, Guanin, Uracil"],
            "answer": "B",
            "explanation": "De fire nitrogenbasene i RNA er Adenin, Uracil, Guanin og Cytosin."
        },
        {
            "question": "Hvilke nitrogenbaser finnes i DNA og RNA, og hvilke nitrogenbaser danner basepar?",
            "options": ["A. DNA: Adenin, Tymin, Guanin, Cytosin; RNA: Adenin, Uracil, Guanin, Cytosin; Basepar: A-T, G-C, A-U", "B. DNA: Adenin, Uracil, Guanin, Cytosin; RNA: Adenin, Tymin, Guanin, Cytosin; Basepar: A-T, G-C, A-U", "C. DNA: Adenin, Tymin, Guanin, Uracil; RNA: Adenin, Uracil, Guanin, Cytosin; Basepar: A-T, G-C, A-U", "D. DNA: Adenin, Tymin, Guanin, Cytosin; RNA: Adenin, Tymin, Guanin, Cytosin; Basepar: A-T, G-C, A-U"],
            "answer": "A",
            "explanation": "I DNA er nitrogenbasene Adenin, Tymin, Guanin og Cytosin, og i RNA er de Adenin, Uracil, Guanin og Cytosin. Baseparene er A-T (i DNA), G-C (i DNA og RNA), og A-U (i RNA)."
        },
        {
            "question": "Hva slags bindinger holder basene sammen?",
            "options": ["A. Kovalente bindinger", "B. Hydrogenbindinger", "C. Ioniske bindinger", "D. Metallbindinger"],
            "answer": "B",
            "explanation": "Hydrogenbindinger holder nitrogenbasene sammen i DNA-molekylet."
        },
        {
            "question": "Hva er DNA-syntesen?",
            "options": ["A. Prosessen der DNA kopieres", "B. Prosessen der RNA lages", "C. Prosessen der proteiner lages", "D. Prosessen der lipider lages"],
            "answer": "A",
            "explanation": "DNA-syntesen er prosessen der DNA kopieres før celledeling."
        },
        {
            "question": "Hvilken rolle spiller enzymet helicase i DNA-syntesen?",
            "options": ["A. Binder sammen DNA-trådene", "B. Åpner DNA-dobbelttråden", "C. Legger til nye nukleotider", "D. Fjerner feil i DNA"],
            "answer": "B",
            "explanation": "Helicase åpner DNA-dobbelttråden ved å bryte hydrogenbindingene mellom basene."
        },
        {
            "question": "Hvilken rolle spiller enzymet DNA-polymerase i DNA-syntesen?",
            "options": ["A. Binder sammen DNA-trådene", "B. Åpner DNA-dobbelttråden", "C. Legger til nye nukleotider", "D. Fjerner feil i DNA"],
            "answer": "C",
            "explanation": "DNA-polymerase legger til nye nukleotider på den voksende DNA-tråden under DNA-syntesen."
        },
        {
            "question": "Hvilken rolle spiller enzymet ligase i DNA-syntesen?",
            "options": ["A. Binder sammen DNA-trådene", "B. Åpner DNA-dobbelttråden", "C. Legger til nye nukleotider", "D. Fjerner feil i DNA"],
            "answer": "A",
            "explanation": "Ligase binder sammen Okazaki-fragmentene på den laggende tråden under DNA-syntesen."
        },
        {
            "question": "Hva er proteinsyntesen?",
            "options": ["A. Prosessen der DNA kopieres", "B. Prosessen der RNA lages", "C. Prosessen der proteiner lages", "D. Prosessen der lipider lages"],
            "answer": "C",
            "explanation": "Proteinsyntesen er prosessen der proteiner lages fra mRNA i ribosomene."
        },
        {
            "question": "Forklar proteinsyntesen så detaljert du kan.",
            "options": ["A. Transkripsjon og translasjon", "B. Replikasjon og transkripsjon", "C. Translasjon og replikasjon", "D. Replikasjon og translasjon"],
            "answer": "A",
            "explanation": "Proteinsyntesen består av to hovedtrinn: transkripsjon (der DNA omdannes til mRNA) og translasjon (der mRNA oversettes til et protein ved ribosomene)."
        },
        {
            "question": "Nevn noen ulike egenskaper proteiner kan ha i kroppen din.",
            "options": ["A. Enzymer, hormoner, antistoffer", "B. Karbohydrater, lipider, vitaminer", "C. Nukleinsyrer, polysakkarider, fettsyrer", "D. Mineraler, vann, oksygen"],
            "answer": "A",
            "explanation": "Proteiner kan fungere som enzymer, hormoner og antistoffer, blant andre roller."
        },
        {
            "question": "Hvor mange ulike aminosyrer har du i kroppen din?",
            "options": ["A. 10", "B. 20", "C. 30", "D. 40"],
            "answer": "B",
            "explanation": "Det finnes 20 ulike aminosyrer som brukes til å bygge proteiner i kroppen."
        },
        {
            "question": "Hvilken funksjon har aminosyrene i kroppen din?",
            "options": ["A. Byggesteiner for proteiner", "B. Energikilde", "C. Lagring av genetisk informasjon", "D. Transport av oksygen"],
            "answer": "A",
            "explanation": "Aminosyrene fungerer som byggesteiner for proteiner, som er essensielle for kroppens struktur og funksjon."
        },
        {
            "question": "Hvor i cellen skjer proteinsyntesen?",
            "options": ["A. I cellekjernen", "B. I mitokondriene", "C. I ribosomene", "D. I cellemembranen"],
            "answer": "C",
            "explanation": "Proteinsyntesen skjer i ribosomene, som finnes i cytoplasmaet."
        },
        {
            "question": "Hva heter de molekylene som finnes i cytoplasmaet og transporterer aminosyrer til ribosomene der proteinsyntesen foregår?",
            "options": ["A. mRNA", "B. tRNA", "C. rRNA", "D. DNA"],
            "answer": "B",
            "explanation": "tRNA (transfer RNA) er molekylene som transporterer aminosyrer til ribosomene under proteinsyntesen."
        },
        {
            "question": "Hva er forskjellen på DNA og RNA?",
            "options": ["A. DNA er dobbelttrådet, RNA er enkelttrådet", "B. DNA inneholder ribose, RNA inneholder deoksyribose", "C. DNA inneholder uracil, RNA inneholder tymin", "D. DNA er enkelttrådet, RNA er dobbelttrådet"],
            "answer": "A",
            "explanation": "DNA er dobbelttrådet og inneholder deoksyribose, mens RNA er enkelttrådet og inneholder ribose."
        },
        {
            "question": "Hva er en mutasjon?",
            "options": ["A. En endring i DNA-sekvensen", "B. En endring i RNA-sekvensen", "C. En endring i proteinstrukturen", "D. En endring i cellemembranen"],
            "answer": "A",
            "explanation": "En mutasjon er en endring i DNA-sekvensen som kan påvirke genfunksjonen."
        },
        {
            "question": "Hva er en mutasjon, og hvilke 3 mutasjoner har vi?",
            "options": ["A. Insersjon, delesjon, basebytte", "B. Transkripsjon, translasjon, replikasjon", "C. Punktmutasjon, kromosommutasjon, genommutasjon", "D. Inversjon, translokasjon, duplikasjon"],
            "answer": "A",
            "explanation": "De tre typene mutasjoner er insersjon (innsetting av ekstra basepar), delesjon (fjerning av basepar), og basebytte (endring av et basepar til et annet)."
        },
        {
            "question": "Forklar hva som menes med insersjon, delesjon og basebytte.",
            "options": ["A. Insersjon: fjerning av basepar, delesjon: innsetting av basepar, basebytte: endring av basepar", "B. Insersjon: innsetting av basepar, delesjon: fjerning av basepar, basebytte: endring av basepar", "C. Insersjon: endring av basepar, delesjon: innsetting av basepar, basebytte: fjerning av basepar", "D. Insersjon: innsetting av basepar, delesjon: endring av basepar, basebytte: fjerning av basepar"],
            "answer": "B",
            "explanation": "Insersjon er innsetting av ekstra basepar, delesjon er fjerning av basepar, og basebytte er endring av et basepar til et annet."
        },
        {
            "question": "Hva er sammenhengen mellom mutasjoner og evolusjon?",
            "options": ["A. Mutasjoner fører alltid til sykdommer", "B. Mutasjoner kan gi opphav til nye egenskaper som kan være fordelaktige", "C. Mutasjoner har ingen betydning for evolusjon", "D. Mutasjoner stopper evolusjon"],
            "answer": "B",
            "explanation": "Mutasjoner kan gi opphav til nye egenskaper som kan være fordelaktige og dermed bidra til evolusjon ved naturlig seleksjon."
        },
        {
            "question": "Hvem var Gregor Mendel? Og hva ble han kjent for?",
            "options": ["A. En fysiker kjent for relativitetsteorien", "B. En kjemiker kjent for periodesystemet", "C. En biolog kjent for arvelæren", "D. En lege kjent for vaksinasjon"],
            "answer": "C",
            "explanation": "Gregor Mendel var en biolog kjent for sine eksperimenter med erteplanter som la grunnlaget for arvelæren."
        },
        {
            "question": "Forklar begrepet dominant.",
            "options": ["A. Et gen som alltid uttrykkes", "B. Et gen som aldri uttrykkes", "C. Et gen som uttrykkes kun i homozygot tilstand", "D. Et gen som uttrykkes kun i heterozygot tilstand"],
            "answer": "A",
            "explanation": "Et dominant gen er et gen som alltid uttrykkes i fenotypen, selv om det bare er til stede i ett eksemplar."
        },
        {
            "question": "Forklar begrepet recessiv.",
            "options": ["A. Et gen som alltid uttrykkes", "B. Et gen som aldri uttrykkes", "C. Et gen som uttrykkes kun i homozygot tilstand", "D. Et gen som uttrykkes kun i heterozygot tilstand"],
            "answer": "C",
            "explanation": "Et recessivt gen er et gen som bare uttrykkes i fenotypen når det er til stede i to eksemplarer (homozygot tilstand)."
        },
        {
            "question": "Forklar begrepet homozygot.",
            "options": ["A. To like alleler for et gen", "B. To ulike alleler for et gen", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "A",
            "explanation": "Homozygot betyr at en organisme har to like alleler for et bestemt gen."
        },
        {
            "question": "Forklar begrepet heterozygot.",
            "options": ["A. To like alleler for et gen", "B. To ulike alleler for et gen", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "B",
            "explanation": "Heterozygot betyr at en organisme har to ulike alleler for et bestemt gen."
        },
        {
            "question": "Forklar begrepet genotype.",
            "options": ["A. Den fysiske uttrykksformen av et gen", "B. Den genetiske sammensetningen av et individ", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "B",
            "explanation": "Genotype refererer til den genetiske sammensetningen av et individ, inkludert alleler for bestemte gener."
        },
        {
            "question": "Forklar begrepet fenotype.",
            "options": ["A. Den fysiske uttrykksformen av et gen", "B. Den genetiske sammensetningen av et individ", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "A",
            "explanation": "Fenotype refererer til den fysiske uttrykksformen av et gen, inkludert observerbare egenskaper."
        },
        {
            "question": "Er påstanden riktig eller gal: Fire og fire nitrogenbaser er kode for en aminosyre når det blir laget proteiner. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Tre og tre nitrogenbaser (kalt et kodon) koder for en aminosyre under proteinsyntesen."
        },
        {
            "question": "Er påstanden riktig eller gal: vi har dobbelt opp av alle gener. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. Vi har to sett med kromosomer, ett fra hver forelder, og dermed dobbelt opp av alle gener."
        },
        {
            "question": "Er påstanden riktig eller gal: Det er mange gener i hvert kromosom. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. Hvert kromosom inneholder mange gener som koder for ulike proteiner."
        },
        {
            "question": "Er påstanden riktig eller gal: Arvestoffet vårt oppbevares i ribosomene. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Arvestoffet vårt (DNA) oppbevares i cellekjernen, ikke i ribosomene."
        },
        {
            "question": "Er påstanden riktig eller gal: Når to brunøyde foreldre får barn, kan de ikke få blåøyde barn. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. To brunøyde foreldre kan bære recessive gener for blå øyne og dermed få blåøyde barn."
        },
        {
            "question": "Er påstanden riktig eller gal: Når to blåøyde foreldre får barn, kan de ikke få brunøyde barn. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. To blåøyde foreldre har kun recessive gener for blå øyne og kan derfor ikke få brunøyde barn."
        },
        {
            "question": "Er påstanden riktig eller gal: m-RNA-molekyler binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av t-RNA. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Det er t-RNA som binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av m-RNA."
        },
        {
            "question": "Er påstanden riktig eller gal: m-RNA-molekyler binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av t-RNA. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Det er t-RNA som binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av m-RNA."
        },
        {
            "question": "Er påstanden riktig eller gal: RNA inneholder ikke Adenin. Det inneholder Uracil i stedet. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. RNA inneholder både Adenin og Uracil. Uracil erstatter Tymin som finnes i DNA."
        },
        {
            "question": "Er påstanden riktig eller gal: Proteinsyntesen skjer kun i cellekjernen. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Proteinsyntesen skjer i ribosomene som finnes i cytoplasmaet, ikke i cellekjernen."
        },
        {
            "question": "Er påstanden riktig eller gal: Alle mutasjoner fører til dårligere DNA/ulemper for individet. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Ikke alle mutasjoner er skadelige; noen kan være nøytrale eller til og med fordelaktige."
        },
        {
            "question": "Er påstanden riktig eller gal: Menn har kun ett X-kromosom og er derfor mer utsatt for kjønnsbundne sykdommer. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. Menn har ett X- og ett Y-kromosom, og er derfor mer utsatt for kjønnsbundne sykdommer som er knyttet til X-kromosomet."
        },
        {
            "question": "Fregner er en dominant egenskap. En mann og en kvinne som begge er heterozygote får barn sammen. Har de to foreldrene fregner eller ikke? Begrunn svaret ditt.",
            "options": ["A. Ja, begge har fregner", "B. Nei, ingen har fregner", "C. Bare mannen har fregner", "D. Bare kvinnen har fregner"],
            "answer": "A",
            "explanation": "Begge foreldrene har fregner fordi de har minst ett dominant allel for fregner (Ff)."
        },
        {
            "question": "Genet B er dominant for brune øyne, og genet b er recessivt for blå øyne. Far har genotypen BB og mor har genotypen Bb. Hvor stor sannsynlighet er det for at de får et barn med blå øyne? Sett gjerne opp et krysningsskjema.",
            "options": ["A. 0%", "B. 25%", "C. 50%", "D. 75%"],
            "answer": "A",
            "explanation": "Sannsynligheten er 0% fordi alle barna vil ha minst ett dominant B-allel og dermed brune øyne."
        },
        {
            "question": "Genet B er dominant for brune øyne, og genet b er recessivt for blå øyne. Far har genotypen BB og mor har genotypen Bb. Hvor stor sannsynlighet er det for at et av deres barn får brune øyne? Sett opp et krysningsskjema.",
            "options": ["A. 0%", "B. 25%", "C. 50%", "D. 100%"],
            "answer": "D",
            "explanation": "Sannsynligheten er 100% fordi alle barna vil ha minst ett dominant B-allel og dermed brune øyne."
        },
        {
            "question": "Hvem har størst sjanse for å bli rødgrønn fargeblinde, menn eller kvinner?",
            "options": ["A. Menn", "B. Kvinner", "C. Begge har lik sjanse", "D. Ingen av dem"],
            "answer": "A",
            "explanation": "Menn har større sjanse for å bli rødgrønn fargeblinde fordi de har bare ett X-kromosom, og hvis det inneholder det recessive allelet for fargeblindhet, vil de uttrykke egenskapen."
        },
        {
            "question": "Hvem har størst sjanse for å bli rødgrønn fargeblinde, menn eller kvinner? Begrunn svaret.",
            "options": ["A. Menn", "B. Kvinner", "C. Begge har lik sjanse", "D. Ingen av dem"],
            "answer": "A",
            "explanation": "Menn har større sjanse for å bli rødgrønn fargeblinde fordi de har bare ett X-kromosom. Hvis det inneholder det recessive allelet for fargeblindhet, vil de uttrykke egenskapen. Kvinner har to X-kromosomer, så de må ha to kopier av det recessive allelet for å være fargeblinde."
        },
        {
            "question": "En mann kan ikke gi rødgrønn fargeblindhet til sine sønner. Hvorfor ikke?",
            "options": ["A. Fordi fargeblindhet er ikke arvelig", "B. Fordi fargeblindhet er knyttet til Y-kromosomet", "C. Fordi fargeblindhet er knyttet til X-kromosomet", "D. Fordi menn ikke kan få sønner"],
            "answer": "C",
            "explanation": "En mann kan ikke gi rødgrønn fargeblindhet til sine sønner fordi fargeblindhet er knyttet til X-kromosomet, og sønner arver sitt X-kromosom fra moren."
        },
        {
            "question": "Hvordan er kjønnskromosomene til en mann?",
            "options": ["A. XX", "B. XY", "C. YY", "D. XZ"],
            "answer": "B",
            "explanation": "Kjønnskromosomene til en mann er XY."
        },
        {
            "question": "Hvordan er kjønnskromosomene til en kvinne?",
            "options": ["A. XX", "B. XY", "C. YY", "D. XZ"],
            "answer": "A",
            "explanation": "Kjønnskromosomene til en kvinne er XX."
        },
        {
            "question": "Hvilket gen bestemmer kjønn?",
            "options": ["A. SRY-genet", "B. BRCA1-genet", "C. HBB-genet", "D. CFTR-genet"],
            "answer": "A",
            "explanation": "SRY-genet på Y-kromosomet bestemmer mannlig kjønn."
        },
        {
            "question": "Hvor mange gener har et Y-kromosom sammenlignet med X-kromosomet?",
            "options": ["A. Flere", "B. Færre", "C. Like mange", "D. Ingen"],
            "answer": "B",
            "explanation": "Y-kromosomet har færre gener sammenlignet med X-kromosomet."
        },
        {
            "question": "Nevn noen kjønnsbundne sykdommer.",
            "options": ["A. Hemofili, Duchenne muskeldystrofi", "B. Diabetes, Alzheimers", "C. Kreft, Hjerteinfarkt", "D. Astma, Allergi"],
            "answer": "A",
            "explanation": "Hemofili og Duchenne muskeldystrofi er eksempler på kjønnsbundne sykdommer."
        },
        {
            "question": "To «friske» foreldre får en sønn som er fargeblind. Hva slags genotype har foreldrene?",
            "options": ["A. Mor: X^NX^N, Far: X^NY", "B. Mor: X^NX^n, Far: X^NY", "C. Mor: X^nX^n, Far: X^NY", "D. Mor: X^NX^N, Far: X^nY"],
            "answer": "B",
            "explanation": "Mor må være bærer (X^NX^n) og far må ha normalt syn (X^NY) for at de kan få en fargeblind sønn."
        },
        {
            "question": "Et par får en datter som er rødgrønn fargeblind. Hvilke genotyper kan foreldrene ha?",
            "options": ["A. Mor: X^NX^N, Far: X^nY", "B. Mor: X^NX^n, Far: X^nY", "C. Mor: X^nX^n, Far: X^NY", "D. Mor: X^nX^n, Far: X^nY"],
            "answer": "D",
            "explanation": "Begge foreldrene må ha det recessive allelet for fargeblindhet. Mor må være X^nX^n og far må være X^nY."
        },
        {
            "question": "I en kvinne skjer det en mutasjon i et gen i en celle i leveren. Genet virker ikke lenger som det skal. Diskuter om denne genfeilen kan bli overført til neste generasjon.",
            "options": ["A. Ja, alltid", "B. Nei, aldri", "C. Bare hvis det skjer i kjønnscellene", "D. Bare hvis det skjer i blodcellene"],
            "answer": "C",
            "explanation": "Genfeilen kan bare bli overført til neste generasjon hvis mutasjonen skjer i kjønnscellene (eggcellene). Mutasjoner i somatiske celler (som leverceller) overføres ikke til avkommet."
         }   
]

# --- LOGIC ENGINE ---

class CapoFinalFixQuiz:
    def __init__(self, data):
        self.original_data = data
        self.quiz_data = []
        self.current_index = 0
        self.score = 0
        
        display(HTML(custom_css))
        
#---------------------------------------------------------------------# 1. Overskrift på quizen skrives under her ---------------------------------------
        self.header_widget = widgets.HTML(
            value="<div class='quiz-header-widget'><h1 class='quiz-title'>Quiz</h1></div>",
            layout=widgets.Layout(width='100%')
        )
        
        # 2. Content Box
        self.content_box = widgets.VBox([])
        self.content_box.add_class('quiz-content-area')
        self.content_box.layout = widgets.Layout(width='100%')
        
        # 3. Glass Card Container
        self.glass_card = widgets.VBox([self.header_widget, self.content_box])
        self.glass_card.add_class('quiz-card-container')
        
        # 4. Master Background
        self.master_bg = widgets.VBox([self.glass_card])
        self.master_bg.add_class('master-container')
        
        display(self.master_bg)
        self.start_game()

    def start_game(self):
        self.quiz_data = self.original_data.copy()
        random.shuffle(self.quiz_data)
        self.current_index = 0
        self.score = 0
        self.render_question()

    def render_question(self):
        if self.current_index >= len(self.quiz_data):
            self.render_results()
            return

        q_data = self.quiz_data[self.current_index]
        total_q = len(self.quiz_data)
        
        prog_val = ((self.current_index) / total_q) * 100
        
        html_content = f"""
        <div class="slide-in-up">
            <div class="progress-wrapper">
                <div style="display:flex; justify-content:space-between; color:#6b7280; font-size:13px; margin-bottom:10px; font-weight:700; letter-spacing:1px; text-transform:uppercase;">
                    <span>Progresjon</span>
                    <span>{self.current_index + 1} / {total_q}</span>
                </div>
                <div class="progress-bg">
                    <div class="progress-fill" style="width:{prog_val}%;"></div>
                </div>
            </div>
            <div class='question-text'>{q_data['question']}</div>
        </div>
        """
        top_html = widgets.HTML(html_content)

        self.buttons = []
        for idx, option in enumerate(q_data['options']):
            btn = widgets.Button(description=option)
            btn.layout = widgets.Layout(width='100%', height='auto')
            btn.on_click(lambda b, i=idx, code=chr(65+idx): self.handle_answer(b, i, code))
            self.buttons.append(btn)
        
        btn_box = widgets.VBox(self.buttons)
        btn_box.layout = widgets.Layout(width='100%')
        btn_box.add_class('slide-in-up')
        
        self.feedback_out = widgets.Output()
        self.feedback_out.layout = widgets.Layout(width='100%')

        self.content_box.children = [top_html, btn_box, self.feedback_out]

    def handle_answer(self, clicked_btn, index, selected_code):
        q_data = self.quiz_data[self.current_index]
        correct_code = q_data['answer']
        is_correct = (selected_code == correct_code)

        if is_correct:
            self.score += 1

        for btn in self.buttons:
            btn.disabled = True
            if btn.description.startswith(correct_code) and not is_correct:
                btn.add_class('fasit-highlight')
            
        if is_correct:
            clicked_btn.add_class('correct-flash')
            color = "#10b981"
            title = "✨ HELT RIKTIG!"
        else:
            clicked_btn.add_class('wrong-answer')
            color = "#ef4444"
            title = "❌ DESVERRE FEIL"

        with self.feedback_out:
            clear_output()
            feedback_html = f"""
            <div class='feedback-box' style='border-left-color: {color};'>
                <strong style='color:{color}; font-size:20px; letter-spacing:1px;'>{title}</strong><br>
                <div style="margin-top:10px; color:#374151;">{q_data['explanation']}</div>
            </div>
            """
            display(HTML(feedback_html))
            
            btn_next = widgets.Button(
                description="GÅ VIDERE ➜", 
                button_style='primary', 
                layout=widgets.Layout(width='100%', margin='30px 0 0 0') 
            )
            btn_next.style.font_weight = '900'
            btn_next.style.font_size = '18px'
            btn_next.style.button_color = '#2563eb'
            btn_next.on_click(lambda b: self.next_q())
            display(btn_next)

    def next_q(self):
        self.current_index += 1
        self.render_question()

    def generate_confetti(self):
        """Genererer en 'sky' av konfetti divs som flyter over skjermen."""
        confetti_html = "<div class='confetti-container'>"
        colors = ['#ef476f', '#ffd166', '#06d6a0', '#118ab2', '#073b4c', '#e73c7e']
        
        # Genererer 60 biter konfetti
        for i in range(60): 
            left = random.randint(0, 100)
            # Raskere fall: mellom 2s og 5s
            duration = random.uniform(2, 5) 
            delay = random.uniform(0, 3)
            bg = random.choice(colors)
            size = random.randint(8, 14)
            
            confetti_html += f"""
            <div class='confetti' style='
                left:{left}%; 
                animation-duration:{duration}s; 
                animation-delay:{delay}s; 
                background-color:{bg};
                width:{size}px;
                height:{size}px;
            '></div>
            """
        confetti_html += "</div>"
        return confetti_html

    def render_results(self):
        total = len(self.quiz_data)
        percent = (self.score / total) * 100 if total > 0 else 0
        
        # Generer konfetti HTML hvis man består
        confetti_effect = self.generate_confetti() if percent >= 50 else ""
        
        if percent >= 80:
            msg = "Fantastisk! Du er en ekte mester! 🏆"
            bg = "#d1fae5"
            txt = "#065f46"
        elif percent >= 50:
            msg = "Veldig bra jobbet! 📚"
            bg = "#fef3c7"
            txt = "#92400e"
        else:
            msg = "Prøv igjen – øvelse gjør mester! 💪"
            bg = "#fee2e2"
            txt = "#991b1b"

        # Her injiserer vi konfetti ØVERST i HTML-blokken, men inni containeren
        res_html = widgets.HTML(f"""
        <div style="position:relative; width:100%; height:100%;">
            {confetti_effect} 
            <div style="text-align:center; padding-top:20px; position:relative; z-index:10;" class="slide-in-up">
                <div style="font-size:16px; color:#6b7280; letter-spacing:3px; font-weight:700; text-transform:uppercase;">Resultat</div>
                <div style="font-size:100px; font-weight:900; color:#111827; margin: 10px 0; line-height:1; letter-spacing:-4px;">
                    {self.score}/{total}
                </div>
                <div class='result-msg' style="background-color: {bg}; color: {txt};">
                    {msg}
                </div>
            </div>
        </div>
        """)
        
        out_chart = widgets.Output()
        with out_chart:
            fig, ax = plt.subplots(figsize=(6, 6), facecolor='none') 
            sizes = [self.score, total - self.score]
            colors = ['#10b981', '#ef4444']
            wedges, texts, autotexts = ax.pie(
                sizes, labels=['Riktig', 'Feil'], colors=colors, 
                autopct='%1.0f%%', startangle=90, textprops={'color':"#374151", 'fontsize': 14, 'weight':'bold'},
                explode=(0.05, 0)
            )
            plt.setp(autotexts, size=18, weight="bold", color="white")
            plt.close(fig)
            display(fig)

        chart_box = widgets.HBox([out_chart], layout=widgets.Layout(justify_content='center'))

        btn_restart = widgets.Button(
            description="START PÅ NYTT 🔄",
            layout=widgets.Layout(width='100%', height='80px', margin='40px 0 0 0')
        )
        btn_restart.style.button_color = '#2563eb'
        btn_restart.style.text_color = 'white'
        btn_restart.style.font_weight = 'bold'
        btn_restart.style.font_size = '24px'
        btn_restart.on_click(lambda b: self.start_game())

        self.content_box.children = [res_html, chart_box, btn_restart]

# --- START APP ---
app = CapoFinalFixQuiz(original_quiz_data)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, Javascript, clear_output, display
import random
import matplotlib.pyplot as plt

# --- CSS: THE SINGULARITY EDITION ---
custom_css = """
<style>
    /* --- 1. ATMOSFÆRE & BAKGRUNN --- */
    .universe-container {
        background: radial-gradient(circle at center, #1e293b, #0f172a);
        position: relative;
        padding: 60px;
        border-radius: 24px;
        overflow: hidden;
        font-family: 'Inter', system-ui, -apple-system, sans-serif;
        box-shadow: inset 0 0 100px rgba(0,0,0,0.5);
    }

    /* Flytende Orber */
    .orb {
        position: absolute;
        border-radius: 50%;
        filter: blur(40px);
        opacity: 0.4;
        animation: floatOrb 20s infinite ease-in-out;
        z-index: 0;
    }
    .orb-1 { top: -10%; left: -10%; width: 300px; height: 300px; background: #3b82f6; animation-delay: 0s; }
    .orb-2 { bottom: -10%; right: -10%; width: 400px; height: 400px; background: #8b5cf6; animation-delay: -5s; }
    .orb-3 { top: 40%; left: 60%; width: 200px; height: 200px; background: #ec4899; animation-delay: -10s; }

    @keyframes floatOrb {
        0%, 100% { transform: translate(0, 0) scale(1); }
        33% { transform: translate(30px, -50px) scale(1.1); }
        66% { transform: translate(-20px, 20px) scale(0.9); }
    }

    /* --- 2. THE HUD CARD (GLASSMORPHISM ULTIMATE) --- */
    .quiz-card-container {
        background: rgba(255, 255, 255, 0.85);
        backdrop-filter: blur(20px);
        -webkit-backdrop-filter: blur(20px);
        border-radius: 24px;
        box-shadow: 
            0 20px 50px rgba(0,0,0,0.2), 
            0 0 0 1px rgba(255, 255, 255, 0.5) inset;
        width: 100%;
        max-width: 900px; /* Bred kino-format */
        margin: 0 auto;
        position: relative;
        z-index: 10;
        overflow: hidden;
        transition: transform 0.3s ease;
    }

    /* Riste-animasjon ved feil svar */
    .shake-effect {
        animation: shake 0.5s cubic-bezier(.36,.07,.19,.97) both;
    }
    @keyframes shake {
        10%, 90% { transform: translate3d(-1px, 0, 0); }
        20%, 80% { transform: translate3d(2px, 0, 0); }
        30%, 50%, 70% { transform: translate3d(-4px, 0, 0); }
        40%, 60% { transform: translate3d(4px, 0, 0); }
    }

    /* Header */
    .quiz-header-widget {
        display: flex;
        justify-content: space-between;
        align-items: center;
        padding: 25px 40px;
        border-bottom: 1px solid rgba(0,0,0,0.05);
        background: rgba(255,255,255,0.4);
    }
    
    .brand-title {
        font-size: 28px;
        font-weight: 900;
        letter-spacing: 2px;
        background: linear-gradient(135deg, #2563eb, #db2777);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin: 0;
    }

    .streak-counter {
        font-size: 18px;
        font-weight: 700;
        color: #f59e0b;
        background: rgba(254, 243, 199, 0.8);
        padding: 8px 16px;
        border-radius: 20px;
        box-shadow: 0 2px 10px rgba(245, 158, 11, 0.2);
        display: flex;
        align-items: center;
        gap: 8px;
    }

    /* Innhold */
    .quiz-content-area {
        padding: 40px 50px;
    }

    /* Spørsmål */
    .question-text {
        font-size: 30px;
        color: #0f172a;
        font-weight: 800;
        line-height: 1.3;
        margin-bottom: 30px;
        animation: fadeSlideIn 0.5s ease-out;
    }

    @keyframes fadeSlideIn {
        from { opacity: 0; transform: translateY(10px); }
        to { opacity: 1; transform: translateY(0); }
    }

    /* --- KNAPPER MED RIPPLE & GLOW --- */
    .widget-button { 
        position: relative;
        box-sizing: border-box !important;
        font-size: 18px !important; 
        font-weight: 600 !important;
        border-radius: 16px !important; 
        border: 2px solid rgba(0,0,0,0.05) !important;
        background: white !important;
        color: #334155 !important;
        padding: 20px 30px !important;
        margin-bottom: 12px !important;
        width: 100% !important;
        text-align: left !important;
        transition: all 0.2s cubic-bezier(0.4, 0, 0.2, 1) !important;
        overflow: hidden; /* For ripple effekten */
    }
    
    .widget-button:hover:enabled {
        border-color: #3b82f6 !important;
        background: #f8fafc !important;
        transform: scale(1.01) translateY(-2px);
        box-shadow: 0 10px 25px rgba(59, 130, 246, 0.15) !important;
        z-index: 2;
    }

    /* Fasit & Svar Styles */
    .correct-flash { 
        background: #10b981 !important; 
        color: white !important; 
        border-color: #10b981 !important;
        box-shadow: 0 0 30px rgba(16, 185, 129, 0.4) !important;
        font-weight: 800 !important;
    }
    
    .wrong-answer {
        background: #ef4444 !important;
        color: white !important;
        border-color: #ef4444 !important;
    }

    .fasit-highlight {
        background: #ecfdf5 !important;
        color: #059669 !important;
        border: 2px solid #10b981 !important;
    }

    /* Feedback Box */
    .feedback-container {
        margin-top: 30px;
        padding: 25px;
        background: #f1f5f9;
        border-radius: 16px;
        border-left: 6px solid;
        animation: popIn 0.4s cubic-bezier(0.175, 0.885, 0.32, 1.275);
    }
    
    @keyframes popIn {
        from { opacity: 0; transform: scale(0.95); }
        to { opacity: 1; transform: scale(1); }
    }

    /* Progresjonslinje */
    .progress-track {
        height: 6px;
        background: #e2e8f0;
        border-radius: 3px;
        margin-bottom: 30px;
        overflow: hidden;
    }
    .progress-bar-fill {
        height: 100%;
        background: linear-gradient(90deg, #3b82f6, #8b5cf6, #ec4899);
        background-size: 200% 100%;
        animation: gradientMove 3s linear infinite;
        transition: width 0.8s cubic-bezier(0.4, 0, 0.2, 1);
    }
    @keyframes gradientMove { 0% { background-position: 100% 0; } 100% { background-position: -100% 0; } }

    /* Resultat */
    .score-big {
        font-size: 100px;
        font-weight: 900;
        line-height: 1;
        background: -webkit-linear-gradient(#3b82f6, #8b5cf6);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin: 20px 0;
    }

    /* Konfetti er 'must' */
    .confetti {
        position: absolute;
        width: 10px; height: 10px;
        top: -10px;
        z-index: 100;
        animation: confettiFall linear forwards;
    }
    @keyframes confettiFall {
        to { transform: translateY(600px) rotate(720deg); }
    }

</style>
"""

# --- DATA ---
original_quiz_data = [
  {
            "question": "Hva er et kromosom?",
            "options": ["A. En type protein", "B. En del av cellemembranen", "C. En struktur som inneholder DNA og protein", "D. En type lipid"],
            "answer": "C",
            "explanation": "Et kromosom er en struktur som inneholder DNA og proteiner, og finnes i cellekjernen."
        },
        {
            "question": "DNA-tråden er 2 meter lang hvis den strekkes helt ut. En celle er mindre enn en tidels millimeter i diameter. Forklar hvordan det er mulig å få så mye DNA inni hver enkelt celle.",
            "options": ["A. DNA-et er pakket tett sammen i kromosomer", "B. DNA-et er flytende i cytoplasmaet", "C. DNA-et er løst pakket i cellemembranen", "D. DNA-et er pakket i mitokondriene"],
            "answer": "A",
            "explanation": "DNA-et er pakket tett sammen i kromosomer ved hjelp av proteiner som kalles histoner, noe som gjør det mulig å få plass til mye DNA i hver celle."
        },
        {
            "question": "Hvor finner vi DNA-et?",
            "options": ["A. I cellemembranen", "B. I cellekjernen", "C. I ribosomene", "D. I mitokondriene"],
            "answer": "B",
            "explanation": "DNA-et finnes hovedsakelig i cellekjernen."
        },
        {
            "question": "Hva er DNA?",
            "options": ["A. En type karbohydrat", "B. En type protein", "C. En type nukleinsyre", "D. En type lipid"],
            "answer": "C",
            "explanation": "DNA er en type nukleinsyre som inneholder den genetiske informasjonen i cellene."
        },
        {
            "question": "Hvordan er et DNA-molekyl bygget opp?",
            "options": ["A. Enkelttrådet med ribose", "B. Dobbelttrådet med deoksyribose", "C. Enkelttrådet med deoksyribose", "D. Dobbelttrådet med ribose"],
            "answer": "B",
            "explanation": "Et DNA-molekyl er dobbelttrådet og består av deoksyribose, fosfatgrupper og nitrogenbaser."
        },
        {
            "question": "Hva koder genene i DNA for?",
            "options": ["A. Karbohydrater", "B. Lipider", "C. Proteiner", "D. Vitaminer"],
            "answer": "C",
            "explanation": "Genene i DNA koder for proteiner som utfører ulike funksjoner i kroppen."
        },
        {
            "question": "Hvilke 4 nitrogenbaser finnes i DNA?",
            "options": ["A. Adenin, Tymin, Guanin, Cytosin", "B. Adenin, Uracil, Guanin, Cytosin", "C. Adenin, Tymin, Guanin, Uracil", "D. Adenin, Cytosin, Guanin, Uracil"],
            "answer": "A",
            "explanation": "De fire nitrogenbasene i DNA er Adenin, Tymin, Guanin og Cytosin."
        },
        {
            "question": "Hvilke 4 nitrogenbaser finnes i RNA?",
            "options": ["A. Adenin, Tymin, Guanin, Cytosin", "B. Adenin, Uracil, Guanin, Cytosin", "C. Adenin, Tymin, Guanin, Uracil", "D. Adenin, Cytosin, Guanin, Uracil"],
            "answer": "B",
            "explanation": "De fire nitrogenbasene i RNA er Adenin, Uracil, Guanin og Cytosin."
        },
        {
            "question": "Hvilke nitrogenbaser finnes i DNA og RNA, og hvilke nitrogenbaser danner basepar?",
            "options": ["A. DNA: Adenin, Tymin, Guanin, Cytosin; RNA: Adenin, Uracil, Guanin, Cytosin; Basepar: A-T, G-C, A-U", "B. DNA: Adenin, Uracil, Guanin, Cytosin; RNA: Adenin, Tymin, Guanin, Cytosin; Basepar: A-T, G-C, A-U", "C. DNA: Adenin, Tymin, Guanin, Uracil; RNA: Adenin, Uracil, Guanin, Cytosin; Basepar: A-T, G-C, A-U", "D. DNA: Adenin, Tymin, Guanin, Cytosin; RNA: Adenin, Tymin, Guanin, Cytosin; Basepar: A-T, G-C, A-U"],
            "answer": "A",
            "explanation": "I DNA er nitrogenbasene Adenin, Tymin, Guanin og Cytosin, og i RNA er de Adenin, Uracil, Guanin og Cytosin. Baseparene er A-T (i DNA), G-C (i DNA og RNA), og A-U (i RNA)."
        },
        {
            "question": "Hva slags bindinger holder basene sammen?",
            "options": ["A. Kovalente bindinger", "B. Hydrogenbindinger", "C. Ioniske bindinger", "D. Metallbindinger"],
            "answer": "B",
            "explanation": "Hydrogenbindinger holder nitrogenbasene sammen i DNA-molekylet."
        },
        {
            "question": "Hva er DNA-syntesen?",
            "options": ["A. Prosessen der DNA kopieres", "B. Prosessen der RNA lages", "C. Prosessen der proteiner lages", "D. Prosessen der lipider lages"],
            "answer": "A",
            "explanation": "DNA-syntesen er prosessen der DNA kopieres før celledeling."
        },
        {
            "question": "Hvilken rolle spiller enzymet helicase i DNA-syntesen?",
            "options": ["A. Binder sammen DNA-trådene", "B. Åpner DNA-dobbelttråden", "C. Legger til nye nukleotider", "D. Fjerner feil i DNA"],
            "answer": "B",
            "explanation": "Helicase åpner DNA-dobbelttråden ved å bryte hydrogenbindingene mellom basene."
        },
        {
            "question": "Hvilken rolle spiller enzymet DNA-polymerase i DNA-syntesen?",
            "options": ["A. Binder sammen DNA-trådene", "B. Åpner DNA-dobbelttråden", "C. Legger til nye nukleotider", "D. Fjerner feil i DNA"],
            "answer": "C",
            "explanation": "DNA-polymerase legger til nye nukleotider på den voksende DNA-tråden under DNA-syntesen."
        },
        {
            "question": "Hvilken rolle spiller enzymet ligase i DNA-syntesen?",
            "options": ["A. Binder sammen DNA-trådene", "B. Åpner DNA-dobbelttråden", "C. Legger til nye nukleotider", "D. Fjerner feil i DNA"],
            "answer": "A",
            "explanation": "Ligase binder sammen Okazaki-fragmentene på den laggende tråden under DNA-syntesen."
        },
        {
            "question": "Hva er proteinsyntesen?",
            "options": ["A. Prosessen der DNA kopieres", "B. Prosessen der RNA lages", "C. Prosessen der proteiner lages", "D. Prosessen der lipider lages"],
            "answer": "C",
            "explanation": "Proteinsyntesen er prosessen der proteiner lages fra mRNA i ribosomene."
        },
        {
            "question": "Forklar proteinsyntesen så detaljert du kan.",
            "options": ["A. Transkripsjon og translasjon", "B. Replikasjon og transkripsjon", "C. Translasjon og replikasjon", "D. Replikasjon og translasjon"],
            "answer": "A",
            "explanation": "Proteinsyntesen består av to hovedtrinn: transkripsjon (der DNA omdannes til mRNA) og translasjon (der mRNA oversettes til et protein ved ribosomene)."
        },
        {
            "question": "Nevn noen ulike egenskaper proteiner kan ha i kroppen din.",
            "options": ["A. Enzymer, hormoner, antistoffer", "B. Karbohydrater, lipider, vitaminer", "C. Nukleinsyrer, polysakkarider, fettsyrer", "D. Mineraler, vann, oksygen"],
            "answer": "A",
            "explanation": "Proteiner kan fungere som enzymer, hormoner og antistoffer, blant andre roller."
        },
        {
            "question": "Hvor mange ulike aminosyrer har du i kroppen din?",
            "options": ["A. 10", "B. 20", "C. 30", "D. 40"],
            "answer": "B",
            "explanation": "Det finnes 20 ulike aminosyrer som brukes til å bygge proteiner i kroppen."
        },
        {
            "question": "Hvilken funksjon har aminosyrene i kroppen din?",
            "options": ["A. Byggesteiner for proteiner", "B. Energikilde", "C. Lagring av genetisk informasjon", "D. Transport av oksygen"],
            "answer": "A",
            "explanation": "Aminosyrene fungerer som byggesteiner for proteiner, som er essensielle for kroppens struktur og funksjon."
        },
        {
            "question": "Hvor i cellen skjer proteinsyntesen?",
            "options": ["A. I cellekjernen", "B. I mitokondriene", "C. I ribosomene", "D. I cellemembranen"],
            "answer": "C",
            "explanation": "Proteinsyntesen skjer i ribosomene, som finnes i cytoplasmaet."
        },
        {
            "question": "Hva heter de molekylene som finnes i cytoplasmaet og transporterer aminosyrer til ribosomene der proteinsyntesen foregår?",
            "options": ["A. mRNA", "B. tRNA", "C. rRNA", "D. DNA"],
            "answer": "B",
            "explanation": "tRNA (transfer RNA) er molekylene som transporterer aminosyrer til ribosomene under proteinsyntesen."
        },
        {
            "question": "Hva er forskjellen på DNA og RNA?",
            "options": ["A. DNA er dobbelttrådet, RNA er enkelttrådet", "B. DNA inneholder ribose, RNA inneholder deoksyribose", "C. DNA inneholder uracil, RNA inneholder tymin", "D. DNA er enkelttrådet, RNA er dobbelttrådet"],
            "answer": "A",
            "explanation": "DNA er dobbelttrådet og inneholder deoksyribose, mens RNA er enkelttrådet og inneholder ribose."
        },
        {
            "question": "Hva er en mutasjon?",
            "options": ["A. En endring i DNA-sekvensen", "B. En endring i RNA-sekvensen", "C. En endring i proteinstrukturen", "D. En endring i cellemembranen"],
            "answer": "A",
            "explanation": "En mutasjon er en endring i DNA-sekvensen som kan påvirke genfunksjonen."
        },
        {
            "question": "Hva er en mutasjon, og hvilke 3 mutasjoner har vi?",
            "options": ["A. Insersjon, delesjon, basebytte", "B. Transkripsjon, translasjon, replikasjon", "C. Punktmutasjon, kromosommutasjon, genommutasjon", "D. Inversjon, translokasjon, duplikasjon"],
            "answer": "A",
            "explanation": "De tre typene mutasjoner er insersjon (innsetting av ekstra basepar), delesjon (fjerning av basepar), og basebytte (endring av et basepar til et annet)."
        },
        {
            "question": "Forklar hva som menes med insersjon, delesjon og basebytte.",
            "options": ["A. Insersjon: fjerning av basepar, delesjon: innsetting av basepar, basebytte: endring av basepar", "B. Insersjon: innsetting av basepar, delesjon: fjerning av basepar, basebytte: endring av basepar", "C. Insersjon: endring av basepar, delesjon: innsetting av basepar, basebytte: fjerning av basepar", "D. Insersjon: innsetting av basepar, delesjon: endring av basepar, basebytte: fjerning av basepar"],
            "answer": "B",
            "explanation": "Insersjon er innsetting av ekstra basepar, delesjon er fjerning av basepar, og basebytte er endring av et basepar til et annet."
        },
        {
            "question": "Hva er sammenhengen mellom mutasjoner og evolusjon?",
            "options": ["A. Mutasjoner fører alltid til sykdommer", "B. Mutasjoner kan gi opphav til nye egenskaper som kan være fordelaktige", "C. Mutasjoner har ingen betydning for evolusjon", "D. Mutasjoner stopper evolusjon"],
            "answer": "B",
            "explanation": "Mutasjoner kan gi opphav til nye egenskaper som kan være fordelaktige og dermed bidra til evolusjon ved naturlig seleksjon."
        },
        {
            "question": "Hvem var Gregor Mendel? Og hva ble han kjent for?",
            "options": ["A. En fysiker kjent for relativitetsteorien", "B. En kjemiker kjent for periodesystemet", "C. En biolog kjent for arvelæren", "D. En lege kjent for vaksinasjon"],
            "answer": "C",
            "explanation": "Gregor Mendel var en biolog kjent for sine eksperimenter med erteplanter som la grunnlaget for arvelæren."
        },
        {
            "question": "Forklar begrepet dominant.",
            "options": ["A. Et gen som alltid uttrykkes", "B. Et gen som aldri uttrykkes", "C. Et gen som uttrykkes kun i homozygot tilstand", "D. Et gen som uttrykkes kun i heterozygot tilstand"],
            "answer": "A",
            "explanation": "Et dominant gen er et gen som alltid uttrykkes i fenotypen, selv om det bare er til stede i ett eksemplar."
        },
        {
            "question": "Forklar begrepet recessiv.",
            "options": ["A. Et gen som alltid uttrykkes", "B. Et gen som aldri uttrykkes", "C. Et gen som uttrykkes kun i homozygot tilstand", "D. Et gen som uttrykkes kun i heterozygot tilstand"],
            "answer": "C",
            "explanation": "Et recessivt gen er et gen som bare uttrykkes i fenotypen når det er til stede i to eksemplarer (homozygot tilstand)."
        },
        {
            "question": "Forklar begrepet homozygot.",
            "options": ["A. To like alleler for et gen", "B. To ulike alleler for et gen", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "A",
            "explanation": "Homozygot betyr at en organisme har to like alleler for et bestemt gen."
        },
        {
            "question": "Forklar begrepet heterozygot.",
            "options": ["A. To like alleler for et gen", "B. To ulike alleler for et gen", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "B",
            "explanation": "Heterozygot betyr at en organisme har to ulike alleler for et bestemt gen."
        },
        {
            "question": "Forklar begrepet genotype.",
            "options": ["A. Den fysiske uttrykksformen av et gen", "B. Den genetiske sammensetningen av et individ", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "B",
            "explanation": "Genotype refererer til den genetiske sammensetningen av et individ, inkludert alleler for bestemte gener."
        },
        {
            "question": "Forklar begrepet fenotype.",
            "options": ["A. Den fysiske uttrykksformen av et gen", "B. Den genetiske sammensetningen av et individ", "C. Et gen som alltid uttrykkes", "D. Et gen som aldri uttrykkes"],
            "answer": "A",
            "explanation": "Fenotype refererer til den fysiske uttrykksformen av et gen, inkludert observerbare egenskaper."
        },
        {
            "question": "Er påstanden riktig eller gal: Fire og fire nitrogenbaser er kode for en aminosyre når det blir laget proteiner. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Tre og tre nitrogenbaser (kalt et kodon) koder for en aminosyre under proteinsyntesen."
        },
        {
            "question": "Er påstanden riktig eller gal: vi har dobbelt opp av alle gener. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. Vi har to sett med kromosomer, ett fra hver forelder, og dermed dobbelt opp av alle gener."
        },
        {
            "question": "Er påstanden riktig eller gal: Det er mange gener i hvert kromosom. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. Hvert kromosom inneholder mange gener som koder for ulike proteiner."
        },
        {
            "question": "Er påstanden riktig eller gal: Arvestoffet vårt oppbevares i ribosomene. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Arvestoffet vårt (DNA) oppbevares i cellekjernen, ikke i ribosomene."
        },
        {
            "question": "Er påstanden riktig eller gal: Når to brunøyde foreldre får barn, kan de ikke få blåøyde barn. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. To brunøyde foreldre kan bære recessive gener for blå øyne og dermed få blåøyde barn."
        },
        {
            "question": "Er påstanden riktig eller gal: Når to blåøyde foreldre får barn, kan de ikke få brunøyde barn. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. To blåøyde foreldre har kun recessive gener for blå øyne og kan derfor ikke få brunøyde barn."
        },
        {
            "question": "Er påstanden riktig eller gal: m-RNA-molekyler binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av t-RNA. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Det er t-RNA som binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av m-RNA."
        },
        {
            "question": "Er påstanden riktig eller gal: m-RNA-molekyler binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av t-RNA. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Det er t-RNA som binder seg til aminosyrer og bringer dem inn på riktig plass og i riktig rekkefølge bestemt av m-RNA."
        },
        {
            "question": "Er påstanden riktig eller gal: RNA inneholder ikke Adenin. Det inneholder Uracil i stedet. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. RNA inneholder både Adenin og Uracil. Uracil erstatter Tymin som finnes i DNA."
        },
        {
            "question": "Er påstanden riktig eller gal: Proteinsyntesen skjer kun i cellekjernen. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Proteinsyntesen skjer i ribosomene som finnes i cytoplasmaet, ikke i cellekjernen."
        },
        {
            "question": "Er påstanden riktig eller gal: Alle mutasjoner fører til dårligere DNA/ulemper for individet. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "B",
            "explanation": "Påstanden er gal. Ikke alle mutasjoner er skadelige; noen kan være nøytrale eller til og med fordelaktige."
        },
        {
            "question": "Er påstanden riktig eller gal: Menn har kun ett X-kromosom og er derfor mer utsatt for kjønnsbundne sykdommer. Begrunn svaret ditt.",
            "options": ["A. Riktig", "B. Gal"],
            "answer": "A",
            "explanation": "Påstanden er riktig. Menn har ett X- og ett Y-kromosom, og er derfor mer utsatt for kjønnsbundne sykdommer som er knyttet til X-kromosomet."
        },
        {
            "question": "Fregner er en dominant egenskap. En mann og en kvinne som begge er heterozygote får barn sammen. Har de to foreldrene fregner eller ikke? Begrunn svaret ditt.",
            "options": ["A. Ja, begge har fregner", "B. Nei, ingen har fregner", "C. Bare mannen har fregner", "D. Bare kvinnen har fregner"],
            "answer": "A",
            "explanation": "Begge foreldrene har fregner fordi de har minst ett dominant allel for fregner (Ff)."
        },
        {
            "question": "Genet B er dominant for brune øyne, og genet b er recessivt for blå øyne. Far har genotypen BB og mor har genotypen Bb. Hvor stor sannsynlighet er det for at de får et barn med blå øyne? Sett gjerne opp et krysningsskjema.",
            "options": ["A. 0%", "B. 25%", "C. 50%", "D. 75%"],
            "answer": "A",
            "explanation": "Sannsynligheten er 0% fordi alle barna vil ha minst ett dominant B-allel og dermed brune øyne."
        },
        {
            "question": "Genet B er dominant for brune øyne, og genet b er recessivt for blå øyne. Far har genotypen BB og mor har genotypen Bb. Hvor stor sannsynlighet er det for at et av deres barn får brune øyne? Sett opp et krysningsskjema.",
            "options": ["A. 0%", "B. 25%", "C. 50%", "D. 100%"],
            "answer": "D",
            "explanation": "Sannsynligheten er 100% fordi alle barna vil ha minst ett dominant B-allel og dermed brune øyne."
        },
        {
            "question": "Hvem har størst sjanse for å bli rødgrønn fargeblinde, menn eller kvinner?",
            "options": ["A. Menn", "B. Kvinner", "C. Begge har lik sjanse", "D. Ingen av dem"],
            "answer": "A",
            "explanation": "Menn har større sjanse for å bli rødgrønn fargeblinde fordi de har bare ett X-kromosom, og hvis det inneholder det recessive allelet for fargeblindhet, vil de uttrykke egenskapen."
        },
        {
            "question": "Hvem har størst sjanse for å bli rødgrønn fargeblinde, menn eller kvinner? Begrunn svaret.",
            "options": ["A. Menn", "B. Kvinner", "C. Begge har lik sjanse", "D. Ingen av dem"],
            "answer": "A",
            "explanation": "Menn har større sjanse for å bli rødgrønn fargeblinde fordi de har bare ett X-kromosom. Hvis det inneholder det recessive allelet for fargeblindhet, vil de uttrykke egenskapen. Kvinner har to X-kromosomer, så de må ha to kopier av det recessive allelet for å være fargeblinde."
        },
        {
            "question": "En mann kan ikke gi rødgrønn fargeblindhet til sine sønner. Hvorfor ikke?",
            "options": ["A. Fordi fargeblindhet er ikke arvelig", "B. Fordi fargeblindhet er knyttet til Y-kromosomet", "C. Fordi fargeblindhet er knyttet til X-kromosomet", "D. Fordi menn ikke kan få sønner"],
            "answer": "C",
            "explanation": "En mann kan ikke gi rødgrønn fargeblindhet til sine sønner fordi fargeblindhet er knyttet til X-kromosomet, og sønner arver sitt X-kromosom fra moren."
        },
        {
            "question": "Hvordan er kjønnskromosomene til en mann?",
            "options": ["A. XX", "B. XY", "C. YY", "D. XZ"],
            "answer": "B",
            "explanation": "Kjønnskromosomene til en mann er XY."
        },
        {
            "question": "Hvordan er kjønnskromosomene til en kvinne?",
            "options": ["A. XX", "B. XY", "C. YY", "D. XZ"],
            "answer": "A",
            "explanation": "Kjønnskromosomene til en kvinne er XX."
        },
        {
            "question": "Hvilket gen bestemmer kjønn?",
            "options": ["A. SRY-genet", "B. BRCA1-genet", "C. HBB-genet", "D. CFTR-genet"],
            "answer": "A",
            "explanation": "SRY-genet på Y-kromosomet bestemmer mannlig kjønn."
        },
        {
            "question": "Hvor mange gener har et Y-kromosom sammenlignet med X-kromosomet?",
            "options": ["A. Flere", "B. Færre", "C. Like mange", "D. Ingen"],
            "answer": "B",
            "explanation": "Y-kromosomet har færre gener sammenlignet med X-kromosomet."
        },
        {
            "question": "Nevn noen kjønnsbundne sykdommer.",
            "options": ["A. Hemofili, Duchenne muskeldystrofi", "B. Diabetes, Alzheimers", "C. Kreft, Hjerteinfarkt", "D. Astma, Allergi"],
            "answer": "A",
            "explanation": "Hemofili og Duchenne muskeldystrofi er eksempler på kjønnsbundne sykdommer."
        },
        {
            "question": "To «friske» foreldre får en sønn som er fargeblind. Hva slags genotype har foreldrene?",
            "options": ["A. Mor: X^NX^N, Far: X^NY", "B. Mor: X^NX^n, Far: X^NY", "C. Mor: X^nX^n, Far: X^NY", "D. Mor: X^NX^N, Far: X^nY"],
            "answer": "B",
            "explanation": "Mor må være bærer (X^NX^n) og far må ha normalt syn (X^NY) for at de kan få en fargeblind sønn."
        },
        {
            "question": "Et par får en datter som er rødgrønn fargeblind. Hvilke genotyper kan foreldrene ha?",
            "options": ["A. Mor: X^NX^N, Far: X^nY", "B. Mor: X^NX^n, Far: X^nY", "C. Mor: X^nX^n, Far: X^NY", "D. Mor: X^nX^n, Far: X^nY"],
            "answer": "D",
            "explanation": "Begge foreldrene må ha det recessive allelet for fargeblindhet. Mor må være X^nX^n og far må være X^nY."
        },
        {
            "question": "I en kvinne skjer det en mutasjon i et gen i en celle i leveren. Genet virker ikke lenger som det skal. Diskuter om denne genfeilen kan bli overført til neste generasjon.",
            "options": ["A. Ja, alltid", "B. Nei, aldri", "C. Bare hvis det skjer i kjønnscellene", "D. Bare hvis det skjer i blodcellene"],
            "answer": "C",
            "explanation": "Genfeilen kan bare bli overført til neste generasjon hvis mutasjonen skjer i kjønnscellene (eggcellene). Mutasjoner i somatiske celler (som leverceller) overføres ikke til avkommet."
         }   
]

# --- THE UBER ENGINE ---

class UberQuiz:
    def __init__(self, data):
        self.original_data = data
        self.quiz_data = []
        self.current_index = 0
        self.score = 0
        self.streak = 0 # NYTT: Streak teller
        
        display(HTML(custom_css))
        
        # UI Oppsett
        self.header_html = widgets.HTML()
        self.content_box = widgets.VBox([])
        self.content_box.add_class('quiz-content-area')
        
        self.card = widgets.VBox([self.header_html, self.content_box])
        self.card.add_class('quiz-card-container')
        
        # Bakgrunnsorber + Kort
        self.universe = widgets.HTML(f"""
            <div class='universe-container'>
                <div class='orb orb-1'></div>
                <div class='orb orb-2'></div>
                <div class='orb orb-3'></div>
            </div>
        """)
        
        # Triks for å legge kortet OPPÅ universet. 
        # I ren HTML/CSS er dette enklere, men i Widgets må vi være kreative.
        # Vi printer universet først som bakgrunn, og kortet renderes som en widget inni.
        # Her gjør jeg det enkelt: Universet pakker inn kortet via en VBox som har stylingen.
        self.main_display = widgets.VBox([self.card])
        self.main_display.add_class('universe-container') # Her bruker vi universet som wrapper
        
        display(self.main_display)
        self.start_game()

    def update_header(self):
        # Dynamisk Overskrift skrives ved brand-title
        streak_fire = "🔥" * (min(self.streak, 5)) # Maks 5 flammer
        self.header_html.value = f"""
        <div class='quiz-header-widget'>
            <div class='brand-title'>QUIZ</div>   
            <div class='streak-counter'>
                <span>STREAK</span>
                <span>{self.streak}</span>
                <span>{streak_fire}</span>
            </div>
        </div>
        """

    def start_game(self):
        self.quiz_data = self.original_data.copy()
        random.shuffle(self.quiz_data)
        self.current_index = 0
        self.score = 0
        self.streak = 0
        self.render_question()

    def render_question(self):
        if self.current_index >= len(self.quiz_data):
            self.render_results()
            return

        q_data = self.quiz_data[self.current_index]
        total_q = len(self.quiz_data)
        self.update_header()
        
        # Fjern shake-effekt hvis den henger igjen
        self.card.remove_class('shake-effect')
        
        # Progresjon
        prog_pct = (self.current_index / total_q) * 100
        
        html_top = f"""
        <div class="progress-track">
            <div class="progress-bar-fill" style="width: {prog_pct}%;"></div>
        </div>
        <div class='question-text'>{q_data['question']}</div>
        """
        top_widget = widgets.HTML(html_top)

        # Knapper
        self.buttons = []
        for idx, option in enumerate(q_data['options']):
            btn = widgets.Button(description=option)
            btn.layout = widgets.Layout(width='100%', height='auto')
            btn.on_click(lambda b, i=idx, code=chr(65+idx): self.handle_answer(b, i, code))
            self.buttons.append(btn)
            
        btn_box = widgets.VBox(self.buttons)
        self.feedback_out = widgets.Output()
        
        self.content_box.children = [top_widget, btn_box, self.feedback_out]

    def handle_answer(self, clicked_btn, index, selected_code):
        q_data = self.quiz_data[self.current_index]
        correct_code = q_data['answer']
        is_correct = (selected_code == correct_code)

        if is_correct:
            self.score += 1
            self.streak += 1
            color = "#10b981"
            title = "✨ FANTASTISK!"
            # Spill av "riktig"-lyd (visuelt)
            clicked_btn.add_class('correct-flash')
        else:
            self.streak = 0 # Reset streak :(
            color = "#ef4444"
            title = "❌ AI, DEN VAR LEI"
            clicked_btn.add_class('wrong-answer')
            # RISTE EFFEKT PÅ HELE KORTET
            self.card.add_class('shake-effect')

        self.update_header() # Oppdater streak-visning

        # Deaktiver knapper
        for btn in self.buttons:
            btn.disabled = True
            if btn.description.startswith(correct_code) and not is_correct:
                btn.add_class('fasit-highlight')

        with self.feedback_out:
            clear_output()
            feedback_html = f"""
            <div class='feedback-container' style='border-color: {color};'>
                <div style="font-weight:900; font-size:22px; color:{color}; margin-bottom:8px;">{title}</div>
                <div style="font-size:18px; color:#334155; line-height:1.5;">{q_data['explanation']}</div>
            </div>
            """
            display(HTML(feedback_html))
            
            # Neste knapp med gradient stil
            btn_next = widgets.Button(
                description="FORTSETT REISEN ➜", 
                layout=widgets.Layout(width='100%', margin='25px 0 0 0') 
            )
            btn_next.style.button_color = '#1e293b' # Mørk
            btn_next.style.text_color = 'white'
            btn_next.style.font_weight = 'bold'
            btn_next.on_click(lambda b: self.next_q())
            display(btn_next)

    def next_q(self):
        self.current_index += 1
        self.render_question()

    def generate_confetti(self):
        html = "<div style='position:absolute; top:0; left:0; width:100%; height:100%; pointer-events:none; overflow:hidden;'>"
        colors = ['#ec4899', '#8b5cf6', '#3b82f6', '#10b981', '#f59e0b']
        for i in range(80):
            left = random.randint(0, 100)
            dur = random.uniform(2, 5)
            delay = random.uniform(0, 2)
            bg = random.choice(colors)
            size = random.randint(8, 15)
            html += f"<div class='confetti' style='left:{left}%; background:{bg}; animation-duration:{dur}s; animation-delay:{delay}s; width:{size}px; height:{size}px;'></div>"
        html += "</div>"
        return html

    def render_results(self):
        total = len(self.quiz_data)
        percent = (self.score / total) * 100
        
        confetti = self.generate_confetti() if percent >= 50 else ""
        
        if percent >= 90:
            msg = "FANTASTISK! Dette får du til! 👑"
            sub_msg = "En meget god prestasjon."
        elif percent >= 50:
            msg = "MEGET BRA LEVERT! 🔥"
            sub_msg = "Du har god kontroll."
        else:
            msg = "NESTE GANG TAR DU DEN! 💪"
            sub_msg = "Øvelse gjør mester."

        # Resultat HTML
        res_html = widgets.HTML(f"""
        <div style="text-align:center; padding: 20px 0; position:relative;">
            {confetti}
            <div style="font-size:16px; font-weight:700; color:#64748b; letter-spacing:2px; text-transform:uppercase;">Endelig Resultat</div>
            <div class='score-big'>{self.score}/{total}</div>
            <div style="font-size:24px; font-weight:800; color:#1e293b; margin-bottom:10px;">{msg}</div>
            <div style="font-size:18px; color:#64748b;">{sub_msg}</div>
        </div>
        """)

        # Graf
        out_chart = widgets.Output()
        with out_chart:
            fig, ax = plt.subplots(figsize=(6, 6), facecolor='none')
            sizes = [self.score, total - self.score]
            colors = ['#10b981', '#ef4444']
            wedges, texts, autotexts = ax.pie(
                sizes, labels=['Riktig', 'Feil'], colors=colors, 
                autopct='%1.0f%%', startangle=90, 
                textprops={'color':"#334155", 'fontsize': 14, 'weight':'bold'},
                explode=(0.05, 0), shadow=True
            )
            plt.setp(autotexts, size=18, weight="bold", color="white")
            plt.close(fig)
            display(fig)
            
        chart_box = widgets.HBox([out_chart], layout=widgets.Layout(justify_content='center'))
        
        btn_restart = widgets.Button(
            description="SPILL IGJEN ⚡",
            layout=widgets.Layout(width='100%', height='70px', margin='40px 0 0 0')
        )
        btn_restart.style.button_color = '#2563eb'
        btn_restart.style.text_color = 'white'
        btn_restart.style.font_weight = '900'
        btn_restart.style.font_size = '20px'
        btn_restart.on_click(lambda b: self.start_game())
        
        self.content_box.children = [res_html, chart_box, btn_restart]
        self.header_html.value = "" # Skjul header streak på resultat

# --- LAUNCH ---
app = UberQuiz(original_quiz_data)
```
