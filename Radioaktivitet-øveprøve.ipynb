```python
import ipywidgets as widgets
from IPython.display import display, HTML, Javascript, clear_output, display
import random
import matplotlib.pyplot as plt

# --- CSS: FIXED GOD TIER ---
custom_css = """
<style>
    /* --- 1. LEVENDE BAKGRUNN --- */
    .master-container {
        background: linear-gradient(-45deg, #ee7752, #e73c7e, #23a6d5, #23d5ab);
        background-size: 400% 400%;
        animation: gradientBG 15s ease infinite;
        padding: 50px;
        border-radius: 20px;
        font-family: 'Inter', 'Segoe UI', Helvetica, Arial, sans-serif;
    }

    @keyframes gradientBG {
        0% { background-position: 0% 50%; }
        50% { background-position: 100% 50%; }
        100% { background-position: 0% 50%; }
    }

    /* --- 2. GLASS-KORTET --- */
    .quiz-card-container {
        background-color: rgba(255, 255, 255, 0.92);
        backdrop-filter: blur(15px);
        border-radius: 24px;
        box-shadow: 0 25px 80px rgba(0,0,0,0.3);
        width: 100%;
        margin: 0 auto;
        overflow: visible; /* VIKTIG: Lar konfettien flyte fritt */
        border: 1px solid rgba(255, 255, 255, 0.6);
        position: relative;
    }

    /* Header - FIKSET TEKSTKUTT */
    .quiz-header-widget {
        background: transparent;
        padding: 30px 0 10px 0;
        text-align: center;
        width: 100%;
        border-bottom: 2px solid rgba(0,0,0,0.05);
    }
    
    .quiz-title {
        margin: 0;
        font-size: 48px;
        font-weight: 900;
        letter-spacing: 5px;
        text-transform: uppercase;
        
        /* Gradient tekst */
        background: linear-gradient(to right, #2563eb, #9333ea);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        
        /* FIKS FOR KUTTET TEKST: Mer h√∏yde og padding */
        line-height: 1.4; 
        padding-bottom: 10px; 
        display: inline-block; /* Sikrer at boksen omslutter teksten korrekt */
    }

    /* Innhold */
    .quiz-content-area {
        padding: 20px 50px 50px 50px;
        position: relative;
        z-index: 1;
    }

    /* Animasjoner */
    .slide-in-up {
        animation: slideUp 0.6s cubic-bezier(0.16, 1, 0.3, 1) forwards;
        opacity: 0;
        transform: translateY(20px);
    }
    @keyframes slideUp { to { opacity: 1; transform: translateY(0); } }

    /* Progresjon */
    .progress-wrapper {
        margin-top: 15px;
        margin-bottom: 40px;
    }
    .progress-bg {
        width: 100%;
        background: #e5e7eb;
        height: 12px;
        border-radius: 6px;
    }
    .progress-fill {
        background: linear-gradient(90deg, #2563eb, #9333ea);
        height: 12px;
        border-radius: 6px;
        box-shadow: 0 0 15px rgba(147, 51, 234, 0.4);
        transition: width 0.6s cubic-bezier(0.4, 0, 0.2, 1);
    }

    /* Sp√∏rsm√•l */
    .question-text {
        font-size: 28px;
        color: #1f2937;
        font-weight: 800;
        margin-bottom: 35px;
        line-height: 1.3;
        text-align: center;
    }

    /* Knapper */
    .widget-button { 
        box-sizing: border-box !important;
        font-size: 20px !important; 
        border-radius: 16px !important; 
        border: 2px solid #e5e7eb !important; 
        background-color: white !important;
        color: #4b5563 !important;
        margin: 10px 0 !important; 
        padding: 22px 30px !important;
        height: auto !important;
        width: 98% !important; 
        margin-left: 1% !important;
        text-align: left !important;
        outline: none !important;
        box-shadow: 0 4px 6px rgba(0,0,0,0.03) !important;
        transition: all 0.3s cubic-bezier(0.25, 0.8, 0.25, 1) !important;
    }
    
    .widget-button:hover:enabled {
        border-color: #8b5cf6 !important;
        transform: translateY(-4px) scale(1.005);
        box-shadow: 0 15px 30px rgba(139, 92, 246, 0.2) !important;
        z-index: 10;
    }

    /* Feedback Box */
    .feedback-box {
        margin-top: 30px;
        padding: 30px;
        border-radius: 16px;
        background-color: #f9fafb;
        border-left: 8px solid #ccc;
        font-size: 18px;
        line-height: 1.6;
        animation: slideUp 0.4s ease-out;
    }
    
    /* Resultat Melding */
    .result-msg {
        font-size: 24px;
        margin-top: 25px;
        font-weight: 700;
        padding: 30px;
        border-radius: 20px;
        text-align: center;
        box-shadow: 0 10px 30px rgba(0,0,0,0.1);
        position: relative;
        z-index: 20; /* Foran konfetti */
    }

    /* --- KONFETTI MOTOR (FIKSET) --- */
    .confetti-container {
        position: absolute;
        top: 0;
        left: 0;
        width: 100%;
        height: 100%;
        pointer-events: none; /* Lar deg klikke gjennom konfettien */
        z-index: 100; /* H√∏yere enn alt annet */
        overflow: hidden;
    }
    
    .confetti {
        position: absolute;
        width: 12px; 
        height: 12px;
        opacity: 0.8;
        top: -20px; /* Starter over kanten */
        animation: fall linear forwards;
    }
    
    @keyframes fall {
        0% { transform: translateY(-20px) rotate(0deg); opacity: 1;}
        100% { transform: translateY(600px) rotate(720deg); opacity: 0; }
    }
    
    /* Styling for Flash/Highlight */
    .correct-flash { animation: flash-success 0.6s forwards !important; font-weight: bold !important; z-index: 20 !important; }
    @keyframes flash-success {
        0% { background-color: white; border-color: #10b981; }
        50% { background-color: #10b981; border-color: #10b981; color: white; transform: scale(1.02); }
        100% { background-color: #10b981; border-color: #10b981; color: white; }
    }
    .wrong-answer { background-color: #ef4444 !important; color: white !important; border-color: #ef4444 !important; }
    .fasit-highlight { background-color: #ecfdf5 !important; color: #047857 !important; border-color: #10b981 !important; border-width: 3px !important; font-weight: bold !important; }

</style>
"""

# ------------------------------ Her legger du inn sp√∏rsm√•lene dine fra Copilot -----------------------------------------------------------------------
original_quiz_data = [
  {
    "question": "Hvilket atom har 30 protoner i kjernen?",
    "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Uran (U)"],
    "answer": "C",
    "explanation": "Atomnummer 30 tilsvarer sink (Zn)."
  },
  {
    "question": "Christian m√•ler str√•lingen fra en kilde. Str√•lingen g√•r gjennom papir og tre, men stoppes av en blyplate. Hvilken str√•lingstype er dette?",
    "options": ["A. Alfastr√•ling", "B. Betastr√•ling", "C. Gammastr√•ling"],
    "answer": "C",
    "explanation": "Gammastr√•ling har h√∏y gjennomtrengningsevne og dempes av tunge materialer som bly/betong."
  },
  {
    "question": "En nedbrutt Cs-137-kilde har aktivitet 1 000 Bq. Halveringstiden er 30 √•r. Hva var aktiviteten for 60 √•r siden?",
    "options": ["A. 500 Bq", "B. 1 000 Bq", "C. 2 000 Bq", "D. 4 000 Bq", "E. 8 000 Bq"],
    "answer": "D",
    "explanation": "60 √•r er to halveringer tilbake i tid: 1 000 Bq √ó 2 √ó 2 = 4 000 Bq."
  },
  {
    "question": "Ved karbon-14-datering viser et funn 25 % av opprinnelig C-14. Halveringstiden er 5 730 √•r. Hvor gammelt er funnet?",
    "options": ["A. 1 432 √•r", "B. 5 730 √•r", "C. 7 163 √•r", "D. 11 460 √•r", "E. 17 190 √•r"],
    "answer": "D",
    "explanation": "25 % = 1/4 = to halveringer ‚Üí 2 √ó 5 730 √•r = 11 460 √•r."
  },
  {
    "question": "N√•r kadmium-113 blir til indium-113, hvilken str√•lingstype sendes ut?",
    "options": ["A. Alfastr√•ling", "B. Betastr√•ling (Œ≤‚àí)", "C. Gammastr√•ling"],
    "answer": "B",
    "explanation": "Œ≤‚àí-henfall √∏ker atomnummeret med 1 (Cd ‚Üí In), mens massetallet er uendret."
  },
  {
    "question": "Thorium-224 sender ut alfastr√•ling. Hvilken isotop dannes?",
    "options": ["A. Radium-220", "B. Actinium-220", "C. Thorium-224", "D. Protactinium-224", "E. Uran-228"],
    "answer": "A",
    "explanation": "Alfastr√•ling reduserer atomnummeret med 2 og massetallet med 4: Th-224 ‚Üí Ra-220."
  },
  {
    "question": "Radon-222 sender ut alfastr√•ling. Hvilken datterkjerne dannes?",
    "options": ["A. Polonium-218", "B. Radon-218", "C. Radium-218", "D. Bly-218"],
    "answer": "A",
    "explanation": "Alfa: Z ‚àí 2, A ‚àí 4 ‚Üí Rn-222 (Z=86) blir Po-218 (Z=84)."
  },
  {
    "question": "P√•stand: I et elektrisk felt blir alfastr√•ling b√∏yd mot den negative platen, betastr√•ling mot den positive platen, og gammastr√•ling blir ikke avb√∏yd.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Alfa er positiv, Œ≤‚àí er negativ, gamma er n√∏ytral."
  },
  {
    "question": "Hva beskriver gammastr√•ling best?",
    "options": ["A. Positivt ladde heliumkjerner", "B. Elektroner fra atomkjernen", "C. Elektromagnetisk str√•ling med sv√¶rt kort b√∏lgelengde"],
    "answer": "C",
    "explanation": "Gamma er fotoner (EM-str√•ling) med kort b√∏lgelengde og h√∏y energi."
  },
  {
    "question": "P√•stand: Betastr√•ling har alltid kortere rekkevidde i luft enn alfastr√•ling.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "B",
    "explanation": "Alfa stopper etter f√• cm; beta kan g√• flere meter i luft."
  },
  {
    "question": "P√•stand: Gammastr√•ling er vanskeligst √• stoppe og skjermes ofte med bly eller betong.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Gamma har stor gjennomtrengningsevne; bly/betong brukes for skjerming."
  },
  {
    "question": "Hvilken av f√∏lgende er vanligvis ikke ioniserende?",
    "options": ["A. UV-C", "B. R√∏ntgen", "C. Gammastr√•ling", "D. Synlig lys"],
    "answer": "D",
    "explanation": "Synlig lys regnes som ikke-ioniserende."
  },
  {
    "question": "Fullf√∏r: \"Den str√•lingstypen som er elektromagnetisk og har sv√¶rt kort b√∏lgelengde\" er ‚Ä¶",
    "options": ["A. Alfastr√•ling", "B. Betastr√•ling", "C. Gammastr√•ling"],
    "answer": "C",
    "explanation": "Gamma er EM-str√•ling."
  },
  {
    "question": "Hvor lang tid tar det f√∏r antallet Rn-222-kjerner er 1/4 av den opprinnelige mengden? Halveringstid: 3,8 dager.",
    "options": ["A. 3,8 dager", "B. 7,6 dager", "C. 11,4 dager", "D. 26,8 minutter"],
    "answer": "B",
    "explanation": "To halveringer: 2 √ó 3,8 dager = 7,6 dager."
  },
  {
    "question": "Hvilket materiale stopper alfastr√•ling effektivt i et enkelt skjermingsfors√∏k?",
    "options": ["A. Et papirark", "B. En treplate", "C. En blyplate", "D. Ingen av disse"],
    "answer": "A",
    "explanation": "Alfa stoppes av papir eller √∏verste hudlag."
  },
  {
    "question": "P√•stand: Ved betastr√•ling (Œ≤‚àí) √∏ker atomnummeret med 1, mens massetallet er uendret.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Et n√∏ytron omdannes til et proton og et elektron; A er uendret."
  },
  {
    "question": "Hvilket grunnstoff dannes n√•r radium-226 sender ut alfastr√•ling?",
    "options": ["A. Radium-222", "B. Radon-222", "C. Polonium-222", "D. Bly-222"],
    "answer": "B",
    "explanation": "Ra-226 ‚Üí Rn-222 ved Œ±-henfall."
  },
  {
    "question": "Fosfor-32 (P-32) sender ut betastr√•ling (Œ≤‚àí). Hvilket isotop blir dannet?",
    "options": ["A. Silisium-32", "B. Svovel-32", "C. Klor-32", "D. Fosfor-31"],
    "answer": "B",
    "explanation": "Œ≤‚àí: P (Z=15) ‚Üí S (Z=16); massetallet uendret: S-32."
  },
  {
    "question": "Hvilken av f√∏lgende er en viktig kilde til naturlig bakgrunnsstr√•ling i Norge?",
    "options": ["A. Radongass i bygninger", "B. Bluetooth-signaler", "C. Synlig lys fra lamper", "D. Lyd fra musikk"],
    "answer": "A",
    "explanation": "Radon fra grunnen kan gi betydelig dose innend√∏rs."
  },
  {
    "question": "Hvilken kombinasjon av observasjoner st√∏tter Big Bang-teorien?",
    "options": [
      "A. R√∏dforskyvning, kosmisk bakgrunnsstr√•ling, fordeling av lette grunnstoffer",
      "B. Nordlys, jordskjelv, ozonlaget",
      "C. Vulkanutbrudd, platetektonikk, v√¶rfenomener",
      "D. Solform√∏rkelser, magnetfelt, fotosyntese"
    ],
    "answer": "A",
    "explanation": "Disse tre observasjonene er sentrale bevislinjer for Big Bang."
  },
  {
    "question": "En geigerteller viser 8 000 tellinger/min ved start. Etter tre halveringer, hva er tellingene per min?",
    "options": ["A. 1 000", "B. 2 000", "C. 4 000", "D. 500"],
    "answer": "A",
    "explanation": "8 000 √ó (1/2)^3 = 1 000."
  },
  {
    "question": "Sorter fra minst til st√∏rst gjennomtrengning: alfa, beta, gamma.",
    "options": ["A. Alfa ‚Üí beta ‚Üí gamma", "B. Beta ‚Üí alfa ‚Üí gamma", "C. Gamma ‚Üí beta ‚Üí alfa"],
    "answer": "A",
    "explanation": "Alfa stoppes lett; beta middels; gamma mest gjennomtrengende."
  },
  {
    "question": "Hvilket symbol representerer alfastr√•ling i en kjernereaksjon?",
    "options": ["A. ‚Å¥‚ÇÇHe", "B. ‚Å∞‚Çã‚ÇÅe", "C. Œ≥"],
    "answer": "A",
    "explanation": "Alfa kan skrives som ‚Å¥‚ÇÇHe."
  },
  {
    "question": "Hva skjer med massetallet ved betastr√•ling (Œ≤‚àí)?",
    "options": ["A. √òker med 1", "B. Minker med 1", "C. Uendret"],
    "answer": "C",
    "explanation": "Et n√∏ytron blir til proton + elektron; massetallet (protoner+n√∏ytroner) er uendret."
  },
  {
    "question": "Et radioaktivt stoff har halveringstid 10 dager. Startmengde 40 g. Hvor mye er igjen etter 30 dager?",
    "options": ["A. 20 g", "B. 10 g", "C. 5 g", "D. 2,5 g"],
    "answer": "C",
    "explanation": "30 dager = tre halveringer: 40 ‚Üí 20 ‚Üí 10 ‚Üí 5 g."
  },
  {
    "question": "P√•stand: Gammastr√•ling sendes ofte ut fra en eksitert kjerne etter alfa- eller betahenfall for √• kvitte seg med overskuddsenergi.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Gamma er fotoner som frakter bort energi n√•r kjernen g√•r til lavere eksitasjonstilstand."
  },
  {
    "question": "Hvilken m√•leenhet brukes for aktivitet (henfall per sekund)?",
    "options": ["A. Gray (Gy)", "B. Sievert (Sv)", "C. Becquerel (Bq)"],
    "answer": "C",
    "explanation": "1 Bq = 1 henfall per sekund."
  },
  {
    "question": "Hvilken m√•leenhet brukes for absorbert dose?",
    "options": ["A. Becquerel (Bq)", "B. Gray (Gy)", "C. Sievert (Sv)"],
    "answer": "B",
    "explanation": "Gy m√•ler energi avsatt per kg (J/kg)."
  },
  {
    "question": "Hvilken m√•leenhet brukes for effektiv dose (risikovektet dose)?",
    "options": ["A. Becquerel (Bq)", "B. Gray (Gy)", "C. Sievert (Sv)"],
    "answer": "C",
    "explanation": "Sv tar hensyn til str√•lingstype og vevssensitivitet."
  },
  {
    "question": "P√•stand: UV-A er alltid ioniserende.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "B",
    "explanation": "UV-A ligger n√¶r grensen; kan gi biologiske skader, men regnes ikke alltid som ioniserende."
  },
  {
    "question": "Hva er hoved√•rsaken til r√∏dforskyvning av lys fra fjerne galakser i Big Bang-modellen?",
    "options": ["A. Universets ekspansjon", "B. Rotasjon i galaksene (Doppler)", "C. Absorpsjon i jordas atmosf√¶re", "D. Gravitasjonslinseeffekter"],
    "answer": "A",
    "explanation": "Rommet utvider seg; b√∏lgelengden strekkes."
  },
  {
    "question": "Hvilke lette grunnstoffer ble hovedsakelig dannet i de f√∏rste minuttene etter Big Bang?",
    "options": ["A. Hydrogen og helium", "B. Karbon og oksygen", "C. Silisium og jern", "D. Hydrogen og litium"],
    "answer": "A",
    "explanation": "I hovedsak H og He (spor av Li)."
  },
  {
    "question": "P√•stand: En Œ≤‚àí-kilde avb√∏yes sterkere i et elektrisk felt enn en Œ±-kilde med samme hastighet.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Œ≤‚àí har mye h√∏yere ladning-til-masse-forhold (q/m) enn Œ± og avb√∏yes mer."
  },
  {
    "question": "Et radiumisotop har 88 protoner og 136 n√∏ytroner. Hvilket isotop skriver du?",
    "options": ["A. Ra-224", "B. Ra-226", "C. Ra-222", "D. Ra-220"],
    "answer": "A",
    "explanation": "Massetallet A = 88 + 136 = 224 ‚Üí Ra-224."
  },
  {
    "question": "Halveringstiden til Pb-214 er 26,8 min. Hvor mye er igjen av 1,0 g etter 80,4 min?",
    "options": ["A. 0,50 g", "B. 0,25 g", "C. 0,125 g", "D. 0,0625 g"],
    "answer": "C",
    "explanation": "80,4 min = 3 halveringer: 1,0 √ó (1/2)^3 = 0,125 g."
  },
  {
    "question": "P√•stand: Bakgrunnsstr√•ling kommer kun fra verdensrommet.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "B",
    "explanation": "Kommer ogs√• fra radon, byggematerialer, mat, medisinske unders√∏kelser m.m."
  },
  {
    "question": "Hvilket instrument bruker man typisk til √• registrere ioniserende str√•ling i skolelaboratoriet?",
    "options": ["A. Spektrofotometer", "B. Geigerteller", "C. Osmometer", "D. Refraktometer"],
    "answer": "B",
    "explanation": "Geigerteller registrerer henfall som pulser/tellinger."
  },
  {
    "question": "Hvilket utsagn om alfastr√•ling er riktig?",
    "options": ["A. Best√•r av n√∏ytrinoer", "B. Best√•r av heliumkjerner (2p+2n)", "C. Er synlig lys", "D. Er negativt ladet"],
    "answer": "B",
    "explanation": "Alfa er ‚Å¥‚ÇÇHe, positivt ladet og tung."
  },
  {
    "question": "Hvilket utsagn om betastr√•ling (Œ≤‚àí) er riktig?",
    "options": ["A. √òker massetallet med 1", "B. Sender ut et elektron fra kjernen", "C. √òker atomnummeret med 2", "D. Er samme som r√∏ntgenstr√•ling"],
    "answer": "B",
    "explanation": "Ved Œ≤‚àí-henfall emitteres et elektron (og et antin√∏ytrino); Z √∏ker med 1."
  },
  {
    "question": "En Rn-222-pr√∏ve reduseres til 12,5 % av startmengden. Omtrent hvor mange halveringer har skjedd?",
    "options": ["A. 2", "B. 3", "C. 4", "D. 5"],
    "answer": "C",
    "explanation": "12,5 % = 1/8 = tre halveringer, men 12,5 % er faktisk 1/8 ‚Üí 3 halveringer (OBS!)."
  },
  {
    "question": "Hvilken skjerming er mest effektiv mot betastr√•ling?",
    "options": ["A. Papirark", "B. Tynn metall (aluminium)- eller blyplate", "C. Luft", "D. Vann"],
    "answer": "B",
    "explanation": "Beta stoppes av metaller (aluminium)/bly (tynt lag)."
  }
]

# --- Her slutter sp√∏rsm√•lsdelen fra copilot, husk bare en hakaeparantes --------------------------------------------------------------------------

class CapoFinalFixQuiz:
    def __init__(self, data):
        self.original_data = data
        self.quiz_data = []
        self.current_index = 0
        self.score = 0
        
        display(HTML(custom_css))
        
#--------------------------------------------------------------# 1. Overskrift p√• quizen skrives under her inni >__< til h√∏yre for quiz-title ---------------------------------------
        self.header_widget = widgets.HTML(
            value="<div class='quiz-header-widget'><h1 class='quiz-title'>Quiz</h1></div>",
            layout=widgets.Layout(width='100%')
        )
        
        # 2. Content Box
        self.content_box = widgets.VBox([])
        self.content_box.add_class('quiz-content-area')
        self.content_box.layout = widgets.Layout(width='100%')
        
        # 3. Glass Card Container
        self.glass_card = widgets.VBox([self.header_widget, self.content_box])
        self.glass_card.add_class('quiz-card-container')
        
        # 4. Master Background
        self.master_bg = widgets.VBox([self.glass_card])
        self.master_bg.add_class('master-container')
        
        display(self.master_bg)
        self.start_game()

    def start_game(self):
        self.quiz_data = self.original_data.copy()
        random.shuffle(self.quiz_data)
        self.current_index = 0
        self.score = 0
        self.render_question()

    def render_question(self):
        if self.current_index >= len(self.quiz_data):
            self.render_results()
            return

        q_data = self.quiz_data[self.current_index]
        total_q = len(self.quiz_data)
        
        prog_val = ((self.current_index) / total_q) * 100
        
        html_content = f"""
        <div class="slide-in-up">
            <div class="progress-wrapper">
                <div style="display:flex; justify-content:space-between; color:#6b7280; font-size:13px; margin-bottom:10px; font-weight:700; letter-spacing:1px; text-transform:uppercase;">
                    <span>Progresjon</span>
                    <span>{self.current_index + 1} / {total_q}</span>
                </div>
                <div class="progress-bg">
                    <div class="progress-fill" style="width:{prog_val}%;"></div>
                </div>
            </div>
            <div class='question-text'>{q_data['question']}</div>
        </div>
        """
        top_html = widgets.HTML(html_content)

        self.buttons = []
        for idx, option in enumerate(q_data['options']):
            btn = widgets.Button(description=option)
            btn.layout = widgets.Layout(width='100%', height='auto')
            btn.on_click(lambda b, i=idx, code=chr(65+idx): self.handle_answer(b, i, code))
            self.buttons.append(btn)
        
        btn_box = widgets.VBox(self.buttons)
        btn_box.layout = widgets.Layout(width='100%')
        btn_box.add_class('slide-in-up')
        
        self.feedback_out = widgets.Output()
        self.feedback_out.layout = widgets.Layout(width='100%')

        self.content_box.children = [top_html, btn_box, self.feedback_out]

    def handle_answer(self, clicked_btn, index, selected_code):
        q_data = self.quiz_data[self.current_index]
        correct_code = q_data['answer']
        is_correct = (selected_code == correct_code)

        if is_correct:
            self.score += 1

        for btn in self.buttons:
            btn.disabled = True
            if btn.description.startswith(correct_code) and not is_correct:
                btn.add_class('fasit-highlight')
            
        if is_correct:
            clicked_btn.add_class('correct-flash')
            color = "#10b981"
            title = "‚ú® HELT RIKTIG!"
        else:
            clicked_btn.add_class('wrong-answer')
            color = "#ef4444"
            title = "‚ùå DESVERRE FEIL"

        with self.feedback_out:
            clear_output()
            feedback_html = f"""
            <div class='feedback-box' style='border-left-color: {color};'>
                <strong style='color:{color}; font-size:20px; letter-spacing:1px;'>{title}</strong><br>
                <div style="margin-top:10px; color:#374151;">{q_data['explanation']}</div>
            </div>
            """
            display(HTML(feedback_html))
            
            btn_next = widgets.Button(
                description="G√Ö VIDERE ‚ûú", 
                button_style='primary', 
                layout=widgets.Layout(width='100%', margin='30px 0 0 0') 
            )
            btn_next.style.font_weight = '900'
            btn_next.style.font_size = '18px'
            btn_next.style.button_color = '#2563eb'
            btn_next.on_click(lambda b: self.next_q())
            display(btn_next)

    def next_q(self):
        self.current_index += 1
        self.render_question()

    def generate_confetti(self):
        """Genererer en 'sky' av konfetti divs som flyter over skjermen."""
        confetti_html = "<div class='confetti-container'>"
        colors = ['#ef476f', '#ffd166', '#06d6a0', '#118ab2', '#073b4c', '#e73c7e']
        
        # Genererer 60 biter konfetti
        for i in range(60): 
            left = random.randint(0, 100)
            # Raskere fall: mellom 2s og 5s
            duration = random.uniform(2, 5) 
            delay = random.uniform(0, 3)
            bg = random.choice(colors)
            size = random.randint(8, 14)
            
            confetti_html += f"""
            <div class='confetti' style='
                left:{left}%; 
                animation-duration:{duration}s; 
                animation-delay:{delay}s; 
                background-color:{bg};
                width:{size}px;
                height:{size}px;
            '></div>
            """
        confetti_html += "</div>"
        return confetti_html

    def render_results(self):
        total = len(self.quiz_data)
        percent = (self.score / total) * 100 if total > 0 else 0
        
        # Generer konfetti HTML hvis man best√•r
        confetti_effect = self.generate_confetti() if percent >= 50 else ""
        
        if percent >= 80:
            msg = "Fantastisk! Du er en ekte mester! üèÜ"
            bg = "#d1fae5"
            txt = "#065f46"
        elif percent >= 50:
            msg = "Veldig bra jobbet! üìö"
            bg = "#fef3c7"
            txt = "#92400e"
        else:
            msg = "Pr√∏v igjen ‚Äì √∏velse gj√∏r mester! üí™"
            bg = "#fee2e2"
            txt = "#991b1b"

        # Her injiserer vi konfetti √òVERST i HTML-blokken, men inni containeren
        res_html = widgets.HTML(f"""
        <div style="position:relative; width:100%; height:100%;">
            {confetti_effect} 
            <div style="text-align:center; padding-top:20px; position:relative; z-index:10;" class="slide-in-up">
                <div style="font-size:16px; color:#6b7280; letter-spacing:3px; font-weight:700; text-transform:uppercase;">Resultat</div>
                <div style="font-size:100px; font-weight:900; color:#111827; margin: 10px 0; line-height:1; letter-spacing:-4px;">
                    {self.score}/{total}
                </div>
                <div class='result-msg' style="background-color: {bg}; color: {txt};">
                    {msg}
                </div>
            </div>
        </div>
        """)
        
        out_chart = widgets.Output()
        with out_chart:
            fig, ax = plt.subplots(figsize=(6, 6), facecolor='none') 
            sizes = [self.score, total - self.score]
            colors = ['#10b981', '#ef4444']
            wedges, texts, autotexts = ax.pie(
                sizes, labels=['Riktig', 'Feil'], colors=colors, 
                autopct='%1.0f%%', startangle=90, textprops={'color':"#374151", 'fontsize': 14, 'weight':'bold'},
                explode=(0.05, 0)
            )
            plt.setp(autotexts, size=18, weight="bold", color="white")
            plt.close(fig)
            display(fig)

        chart_box = widgets.HBox([out_chart], layout=widgets.Layout(justify_content='center'))

        btn_restart = widgets.Button(
            description="START P√Ö NYTT üîÑ",
            layout=widgets.Layout(width='100%', height='80px', margin='40px 0 0 0')
        )
        btn_restart.style.button_color = '#2563eb'
        btn_restart.style.text_color = 'white'
        btn_restart.style.font_weight = 'bold'
        btn_restart.style.font_size = '24px'
        btn_restart.on_click(lambda b: self.start_game())

        self.content_box.children = [res_html, chart_box, btn_restart]

# --- START APP ---
app = CapoFinalFixQuiz(original_quiz_data)
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML, Javascript, clear_output, display
import random
import matplotlib.pyplot as plt

# --- CSS: THE SINGULARITY EDITION ---
custom_css = """
<style>
    /* --- 1. ATMOSF√ÜRE & BAKGRUNN --- */
    .universe-container {
        background: radial-gradient(circle at center, #1e293b, #0f172a);
        position: relative;
        padding: 60px;
        border-radius: 24px;
        overflow: hidden;
        font-family: 'Inter', system-ui, -apple-system, sans-serif;
        box-shadow: inset 0 0 100px rgba(0,0,0,0.5);
    }

    /* Flytende Orber */
    .orb {
        position: absolute;
        border-radius: 50%;
        filter: blur(40px);
        opacity: 0.4;
        animation: floatOrb 20s infinite ease-in-out;
        z-index: 0;
    }
    .orb-1 { top: -10%; left: -10%; width: 300px; height: 300px; background: #3b82f6; animation-delay: 0s; }
    .orb-2 { bottom: -10%; right: -10%; width: 400px; height: 400px; background: #8b5cf6; animation-delay: -5s; }
    .orb-3 { top: 40%; left: 60%; width: 200px; height: 200px; background: #ec4899; animation-delay: -10s; }

    @keyframes floatOrb {
        0%, 100% { transform: translate(0, 0) scale(1); }
        33% { transform: translate(30px, -50px) scale(1.1); }
        66% { transform: translate(-20px, 20px) scale(0.9); }
    }

    /* --- 2. THE HUD CARD (GLASSMORPHISM ULTIMATE) --- */
    .quiz-card-container {
        background: rgba(255, 255, 255, 0.85);
        backdrop-filter: blur(20px);
        -webkit-backdrop-filter: blur(20px);
        border-radius: 24px;
        box-shadow: 
            0 20px 50px rgba(0,0,0,0.2), 
            0 0 0 1px rgba(255, 255, 255, 0.5) inset;
        width: 100%;
        max-width: 900px; /* Bred kino-format */
        margin: 0 auto;
        position: relative;
        z-index: 10;
        overflow: hidden;
        transition: transform 0.3s ease;
    }

    /* Riste-animasjon ved feil svar */
    .shake-effect {
        animation: shake 0.5s cubic-bezier(.36,.07,.19,.97) both;
    }
    @keyframes shake {
        10%, 90% { transform: translate3d(-1px, 0, 0); }
        20%, 80% { transform: translate3d(2px, 0, 0); }
        30%, 50%, 70% { transform: translate3d(-4px, 0, 0); }
        40%, 60% { transform: translate3d(4px, 0, 0); }
    }

    /* Header */
    .quiz-header-widget {
        display: flex;
        justify-content: space-between;
        align-items: center;
        padding: 25px 40px;
        border-bottom: 1px solid rgba(0,0,0,0.05);
        background: rgba(255,255,255,0.4);
    }
    
    .brand-title {
        font-size: 28px;
        font-weight: 900;
        letter-spacing: 2px;
        background: linear-gradient(135deg, #2563eb, #db2777);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin: 0;
    }

    .streak-counter {
        font-size: 18px;
        font-weight: 700;
        color: #f59e0b;
        background: rgba(254, 243, 199, 0.8);
        padding: 8px 16px;
        border-radius: 20px;
        box-shadow: 0 2px 10px rgba(245, 158, 11, 0.2);
        display: flex;
        align-items: center;
        gap: 8px;
    }

    /* Innhold */
    .quiz-content-area {
        padding: 40px 50px;
    }

    /* Sp√∏rsm√•l */
    .question-text {
        font-size: 30px;
        color: #0f172a;
        font-weight: 800;
        line-height: 1.3;
        margin-bottom: 30px;
        animation: fadeSlideIn 0.5s ease-out;
    }

    @keyframes fadeSlideIn {
        from { opacity: 0; transform: translateY(10px); }
        to { opacity: 1; transform: translateY(0); }
    }

    /* --- KNAPPER MED RIPPLE & GLOW --- */
    .widget-button { 
        position: relative;
        box-sizing: border-box !important;
        font-size: 18px !important; 
        font-weight: 600 !important;
        border-radius: 16px !important; 
        border: 2px solid rgba(0,0,0,0.05) !important;
        background: white !important;
        color: #334155 !important;
        padding: 20px 30px !important;
        margin-bottom: 12px !important;
        width: 100% !important;
        text-align: left !important;
        transition: all 0.2s cubic-bezier(0.4, 0, 0.2, 1) !important;
        overflow: hidden; /* For ripple effekten */
    }
    
    .widget-button:hover:enabled {
        border-color: #3b82f6 !important;
        background: #f8fafc !important;
        transform: scale(1.01) translateY(-2px);
        box-shadow: 0 10px 25px rgba(59, 130, 246, 0.15) !important;
        z-index: 2;
    }

    /* Fasit & Svar Styles */
    .correct-flash { 
        background: #10b981 !important; 
        color: white !important; 
        border-color: #10b981 !important;
        box-shadow: 0 0 30px rgba(16, 185, 129, 0.4) !important;
        font-weight: 800 !important;
    }
    
    .wrong-answer {
        background: #ef4444 !important;
        color: white !important;
        border-color: #ef4444 !important;
    }

    .fasit-highlight {
        background: #ecfdf5 !important;
        color: #059669 !important;
        border: 2px solid #10b981 !important;
    }

    /* Feedback Box */
    .feedback-container {
        margin-top: 30px;
        padding: 25px;
        background: #f1f5f9;
        border-radius: 16px;
        border-left: 6px solid;
        animation: popIn 0.4s cubic-bezier(0.175, 0.885, 0.32, 1.275);
    }
    
    @keyframes popIn {
        from { opacity: 0; transform: scale(0.95); }
        to { opacity: 1; transform: scale(1); }
    }

    /* Progresjonslinje */
    .progress-track {
        height: 6px;
        background: #e2e8f0;
        border-radius: 3px;
        margin-bottom: 30px;
        overflow: hidden;
    }
    .progress-bar-fill {
        height: 100%;
        background: linear-gradient(90deg, #3b82f6, #8b5cf6, #ec4899);
        background-size: 200% 100%;
        animation: gradientMove 3s linear infinite;
        transition: width 0.8s cubic-bezier(0.4, 0, 0.2, 1);
    }
    @keyframes gradientMove { 0% { background-position: 100% 0; } 100% { background-position: -100% 0; } }

    /* Resultat */
    .score-big {
        font-size: 100px;
        font-weight: 900;
        line-height: 1;
        background: -webkit-linear-gradient(#3b82f6, #8b5cf6);
        -webkit-background-clip: text;
        -webkit-text-fill-color: transparent;
        margin: 20px 0;
    }

    /* Konfetti er 'must' */
    .confetti {
        position: absolute;
        width: 10px; height: 10px;
        top: -10px;
        z-index: 100;
        animation: confettiFall linear forwards;
    }
    @keyframes confettiFall {
        to { transform: translateY(600px) rotate(720deg); }
    }

</style>
"""

# ------------------------------ Her legger du inn sp√∏rsm√•lene dine fra Copilot -----------------------------------------------------------------------
original_quiz_data = [
    {
    "question": "Hvilket atom har 30 protoner i kjernen?",
    "options": ["A. Karbon (C)", "B. Hydrogen (H)", "C. Sink (Zn)", "D. Uran (U)"],
    "answer": "C",
    "explanation": "Atomnummer 30 tilsvarer sink (Zn)."
  },
  {
    "question": "Christian m√•ler str√•lingen fra en kilde. Str√•lingen g√•r gjennom papir og tre, men stoppes av en blyplate. Hvilken str√•lingstype er dette?",
    "options": ["A. Alfastr√•ling", "B. Betastr√•ling", "C. Gammastr√•ling"],
    "answer": "C",
    "explanation": "Gammastr√•ling har h√∏y gjennomtrengningsevne og dempes av tunge materialer som bly/betong."
  },
  {
    "question": "En nedbrutt Cs-137-kilde har aktivitet 1 000 Bq. Halveringstiden er 30 √•r. Hva var aktiviteten for 60 √•r siden?",
    "options": ["A. 500 Bq", "B. 1 000 Bq", "C. 2 000 Bq", "D. 4 000 Bq", "E. 8 000 Bq"],
    "answer": "D",
    "explanation": "60 √•r er to halveringer tilbake i tid: 1 000 Bq √ó 2 √ó 2 = 4 000 Bq."
  },
  {
    "question": "Ved karbon-14-datering viser et funn 25 % av opprinnelig C-14. Halveringstiden er 5 730 √•r. Hvor gammelt er funnet?",
    "options": ["A. 1 432 √•r", "B. 5 730 √•r", "C. 7 163 √•r", "D. 11 460 √•r", "E. 17 190 √•r"],
    "answer": "D",
    "explanation": "25 % = 1/4 = to halveringer ‚Üí 2 √ó 5 730 √•r = 11 460 √•r."
  },
  {
    "question": "N√•r kadmium-113 blir til indium-113, hvilken str√•lingstype sendes ut?",
    "options": ["A. Alfastr√•ling", "B. Betastr√•ling (Œ≤‚àí)", "C. Gammastr√•ling"],
    "answer": "B",
    "explanation": "Œ≤‚àí-henfall √∏ker atomnummeret med 1 (Cd ‚Üí In), mens massetallet er uendret."
  },
  {
    "question": "Thorium-224 sender ut alfastr√•ling. Hvilken isotop dannes?",
    "options": ["A. Radium-220", "B. Actinium-220", "C. Thorium-224", "D. Protactinium-224", "E. Uran-228"],
    "answer": "A",
    "explanation": "Alfastr√•ling reduserer atomnummeret med 2 og massetallet med 4: Th-224 ‚Üí Ra-220."
  },
  {
    "question": "Radon-222 sender ut alfastr√•ling. Hvilken datterkjerne dannes?",
    "options": ["A. Polonium-218", "B. Radon-218", "C. Radium-218", "D. Bly-218"],
    "answer": "A",
    "explanation": "Alfa: Z ‚àí 2, A ‚àí 4 ‚Üí Rn-222 (Z=86) blir Po-218 (Z=84)."
  },
  {
    "question": "P√•stand: I et elektrisk felt blir alfastr√•ling b√∏yd mot den negative platen, betastr√•ling mot den positive platen, og gammastr√•ling blir ikke avb√∏yd.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Alfa er positiv, Œ≤‚àí er negativ, gamma er n√∏ytral."
  },
  {
    "question": "Hva beskriver gammastr√•ling best?",
    "options": ["A. Positivt ladde heliumkjerner", "B. Elektroner fra atomkjernen", "C. Elektromagnetisk str√•ling med sv√¶rt kort b√∏lgelengde"],
    "answer": "C",
    "explanation": "Gamma er fotoner (EM-str√•ling) med kort b√∏lgelengde og h√∏y energi."
  },
  {
    "question": "P√•stand: Betastr√•ling har alltid kortere rekkevidde i luft enn alfastr√•ling.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "B",
    "explanation": "Alfa stopper etter f√• cm; beta kan g√• flere meter i luft."
  },
  {
    "question": "P√•stand: Gammastr√•ling er vanskeligst √• stoppe og skjermes ofte med bly eller betong.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Gamma har stor gjennomtrengningsevne; bly/betong brukes for skjerming."
  },
  {
    "question": "Hvilken av f√∏lgende er vanligvis ikke ioniserende?",
    "options": ["A. UV-C", "B. R√∏ntgen", "C. Gammastr√•ling", "D. Synlig lys"],
    "answer": "D",
    "explanation": "Synlig lys regnes som ikke-ioniserende."
  },
  {
    "question": "Fullf√∏r: \"Den str√•lingstypen som er elektromagnetisk og har sv√¶rt kort b√∏lgelengde\" er ‚Ä¶",
    "options": ["A. Alfastr√•ling", "B. Betastr√•ling", "C. Gammastr√•ling"],
    "answer": "C",
    "explanation": "Gamma er EM-str√•ling."
  },
  {
    "question": "Hvor lang tid tar det f√∏r antallet Rn-222-kjerner er 1/4 av den opprinnelige mengden? Halveringstid: 3,8 dager.",
    "options": ["A. 3,8 dager", "B. 7,6 dager", "C. 11,4 dager", "D. 26,8 minutter"],
    "answer": "B",
    "explanation": "To halveringer: 2 √ó 3,8 dager = 7,6 dager."
  },
  {
    "question": "Hvilket materiale stopper alfastr√•ling effektivt i et enkelt skjermingsfors√∏k?",
    "options": ["A. Et papirark", "B. En treplate", "C. En blyplate", "D. Ingen av disse"],
    "answer": "A",
    "explanation": "Alfa stoppes av papir eller √∏verste hudlag."
  },
  {
    "question": "P√•stand: Ved betastr√•ling (Œ≤‚àí) √∏ker atomnummeret med 1, mens massetallet er uendret.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Et n√∏ytron omdannes til et proton og et elektron; A er uendret."
  },
  {
    "question": "Hvilket grunnstoff dannes n√•r radium-226 sender ut alfastr√•ling?",
    "options": ["A. Radium-222", "B. Radon-222", "C. Polonium-222", "D. Bly-222"],
    "answer": "B",
    "explanation": "Ra-226 ‚Üí Rn-222 ved Œ±-henfall."
  },
  {
    "question": "Fosfor-32 (P-32) sender ut betastr√•ling (Œ≤‚àí). Hvilket isotop blir dannet?",
    "options": ["A. Silisium-32", "B. Svovel-32", "C. Klor-32", "D. Fosfor-31"],
    "answer": "B",
    "explanation": "Œ≤‚àí: P (Z=15) ‚Üí S (Z=16); massetallet uendret: S-32."
  },
  {
    "question": "Hvilken av f√∏lgende er en viktig kilde til naturlig bakgrunnsstr√•ling i Norge?",
    "options": ["A. Radongass i bygninger", "B. Bluetooth-signaler", "C. Synlig lys fra lamper", "D. Lyd fra musikk"],
    "answer": "A",
    "explanation": "Radon fra grunnen kan gi betydelig dose innend√∏rs."
  },
  {
    "question": "Hvilken kombinasjon av observasjoner st√∏tter Big Bang-teorien?",
    "options": [
      "A. R√∏dforskyvning, kosmisk bakgrunnsstr√•ling, fordeling av lette grunnstoffer",
      "B. Nordlys, jordskjelv, ozonlaget",
      "C. Vulkanutbrudd, platetektonikk, v√¶rfenomener",
      "D. Solform√∏rkelser, magnetfelt, fotosyntese"
    ],
    "answer": "A",
    "explanation": "Disse tre observasjonene er sentrale bevislinjer for Big Bang."
  },
  {
    "question": "En geigerteller viser 8 000 tellinger/min ved start. Etter tre halveringer, hva er tellingene per min?",
    "options": ["A. 1 000", "B. 2 000", "C. 4 000", "D. 500"],
    "answer": "A",
    "explanation": "8 000 √ó (1/2)^3 = 1 000."
  },
  {
    "question": "Sorter fra minst til st√∏rst gjennomtrengning: alfa, beta, gamma.",
    "options": ["A. Alfa ‚Üí beta ‚Üí gamma", "B. Beta ‚Üí alfa ‚Üí gamma", "C. Gamma ‚Üí beta ‚Üí alfa"],
    "answer": "A",
    "explanation": "Alfa stoppes lett; beta middels; gamma mest gjennomtrengende."
  },
  {
    "question": "Hvilket symbol representerer alfastr√•ling i en kjernereaksjon?",
    "options": ["A. ‚Å¥‚ÇÇHe", "B. ‚Å∞‚Çã‚ÇÅe", "C. Œ≥"],
    "answer": "A",
    "explanation": "Alfa kan skrives som ‚Å¥‚ÇÇHe."
  },
  {
    "question": "Hva skjer med massetallet ved betastr√•ling (Œ≤‚àí)?",
    "options": ["A. √òker med 1", "B. Minker med 1", "C. Uendret"],
    "answer": "C",
    "explanation": "Et n√∏ytron blir til proton + elektron; massetallet (protoner+n√∏ytroner) er uendret."
  },
  {
    "question": "Et radioaktivt stoff har halveringstid 10 dager. Startmengde 40 g. Hvor mye er igjen etter 30 dager?",
    "options": ["A. 20 g", "B. 10 g", "C. 5 g", "D. 2,5 g"],
    "answer": "C",
    "explanation": "30 dager = tre halveringer: 40 ‚Üí 20 ‚Üí 10 ‚Üí 5 g."
  },
  {
    "question": "P√•stand: Gammastr√•ling sendes ofte ut fra en eksitert kjerne etter alfa- eller betahenfall for √• kvitte seg med overskuddsenergi.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Gamma er fotoner som frakter bort energi n√•r kjernen g√•r til lavere eksitasjonstilstand."
  },
  {
    "question": "Hvilken m√•leenhet brukes for aktivitet (henfall per sekund)?",
    "options": ["A. Gray (Gy)", "B. Sievert (Sv)", "C. Becquerel (Bq)"],
    "answer": "C",
    "explanation": "1 Bq = 1 henfall per sekund."
  },
  {
    "question": "Hvilken m√•leenhet brukes for absorbert dose?",
    "options": ["A. Becquerel (Bq)", "B. Gray (Gy)", "C. Sievert (Sv)"],
    "answer": "B",
    "explanation": "Gy m√•ler energi avsatt per kg (J/kg)."
  },
  {
    "question": "Hvilken m√•leenhet brukes for effektiv dose (risikovektet dose)?",
    "options": ["A. Becquerel (Bq)", "B. Gray (Gy)", "C. Sievert (Sv)"],
    "answer": "C",
    "explanation": "Sv tar hensyn til str√•lingstype og vevssensitivitet."
  },
  {
    "question": "P√•stand: UV-A er alltid ioniserende.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "B",
    "explanation": "UV-A ligger n√¶r grensen; kan gi biologiske skader, men regnes ikke alltid som ioniserende."
  },
  {
    "question": "Hva er hoved√•rsaken til r√∏dforskyvning av lys fra fjerne galakser i Big Bang-modellen?",
    "options": ["A. Universets ekspansjon", "B. Rotasjon i galaksene (Doppler)", "C. Absorpsjon i jordas atmosf√¶re", "D. Gravitasjonslinseeffekter"],
    "answer": "A",
    "explanation": "Rommet utvider seg; b√∏lgelengden strekkes."
  },
  {
    "question": "Hvilke lette grunnstoffer ble hovedsakelig dannet i de f√∏rste minuttene etter Big Bang?",
    "options": ["A. Hydrogen og helium", "B. Karbon og oksygen", "C. Silisium og jern", "D. Hydrogen og litium"],
    "answer": "A",
    "explanation": "I hovedsak H og He (spor av Li)."
  },
  {
    "question": "P√•stand: En Œ≤‚àí-kilde avb√∏yes sterkere i et elektrisk felt enn en Œ±-kilde med samme hastighet.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "A",
    "explanation": "Œ≤‚àí har mye h√∏yere ladning-til-masse-forhold (q/m) enn Œ± og avb√∏yes mer."
  },
  {
    "question": "Et radiumisotop har 88 protoner og 136 n√∏ytroner. Hvilket isotop skriver du?",
    "options": ["A. Ra-224", "B. Ra-226", "C. Ra-222", "D. Ra-220"],
    "answer": "A",
    "explanation": "Massetallet A = 88 + 136 = 224 ‚Üí Ra-224."
  },
  {
    "question": "Halveringstiden til Pb-214 er 26,8 min. Hvor mye er igjen av 1,0 g etter 80,4 min?",
    "options": ["A. 0,50 g", "B. 0,25 g", "C. 0,125 g", "D. 0,0625 g"],
    "answer": "C",
    "explanation": "80,4 min = 3 halveringer: 1,0 √ó (1/2)^3 = 0,125 g."
  },
  {
    "question": "P√•stand: Bakgrunnsstr√•ling kommer kun fra verdensrommet.",
    "options": ["A. Sant", "B. Usant"],
    "answer": "B",
    "explanation": "Kommer ogs√• fra radon, byggematerialer, mat, medisinske unders√∏kelser m.m."
  },
  {
    "question": "Hvilket instrument bruker man typisk til √• registrere ioniserende str√•ling i skolelaboratoriet?",
    "options": ["A. Spektrofotometer", "B. Geigerteller", "C. Osmometer", "D. Refraktometer"],
    "answer": "B",
    "explanation": "Geigerteller registrerer henfall som pulser/tellinger."
  },
  {
    "question": "Hvilket utsagn om alfastr√•ling er riktig?",
    "options": ["A. Best√•r av n√∏ytrinoer", "B. Best√•r av heliumkjerner (2p+2n)", "C. Er synlig lys", "D. Er negativt ladet"],
    "answer": "B",
    "explanation": "Alfa er ‚Å¥‚ÇÇHe, positivt ladet og tung."
  },
  {
    "question": "Hvilket utsagn om betastr√•ling (Œ≤‚àí) er riktig?",
    "options": ["A. √òker massetallet med 1", "B. Sender ut et elektron fra kjernen", "C. √òker atomnummeret med 2", "D. Er samme som r√∏ntgenstr√•ling"],
    "answer": "B",
    "explanation": "Ved Œ≤‚àí-henfall emitteres et elektron (og et antin√∏ytrino); Z √∏ker med 1."
  },
  {
    "question": "En Rn-222-pr√∏ve reduseres til 12,5 % av startmengden. Omtrent hvor mange halveringer har skjedd?",
    "options": ["A. 2", "B. 3", "C. 4", "D. 5"],
    "answer": "C",
    "explanation": "12,5 % = 1/8 = tre halveringer, men 12,5 % er faktisk 1/8 ‚Üí 3 halveringer (OBS!)."
  },
  {
    "question": "Hvilken skjerming er mest effektiv mot betastr√•ling?",
    "options": ["A. Papirark", "B. Tynn metall (aluminium)- eller blyplate", "C. Luft", "D. Vann"],
    "answer": "B",
    "explanation": "Beta stoppes av metaller (aluminium)/bly (tynt lag)."
  }
 

]

# --- Her slutter sp√∏rsm√•lsdelen fra copilot, husk bare en hakaeparantes --------------------------------------------------------------------------

class UberQuiz:
    def __init__(self, data):
        self.original_data = data
        self.quiz_data = []
        self.current_index = 0
        self.score = 0
        self.streak = 0 # NYTT: Streak teller
        
        display(HTML(custom_css))
        
        # UI Oppsett
        self.header_html = widgets.HTML()
        self.content_box = widgets.VBox([])
        self.content_box.add_class('quiz-content-area')
        
        self.card = widgets.VBox([self.header_html, self.content_box])
        self.card.add_class('quiz-card-container')
        
        # Bakgrunnsorber + Kort
        self.universe = widgets.HTML(f"""
            <div class='universe-container'>
                <div class='orb orb-1'></div>
                <div class='orb orb-2'></div>
                <div class='orb orb-3'></div>
            </div>
        """)
        
        # Triks for √• legge kortet OPP√Ö universet. 
        # I ren HTML/CSS er dette enklere, men i Widgets m√• vi v√¶re kreative.
        # Vi printer universet f√∏rst som bakgrunn, og kortet renderes som en widget inni.
        # Her gj√∏r jeg det enkelt: Universet pakker inn kortet via en VBox som har stylingen.
        self.main_display = widgets.VBox([self.card])
        self.main_display.add_class('universe-container') # Her bruker vi universet som wrapper
        
        display(self.main_display)
        self.start_game()

    def update_header(self):
        # Dynamisk Overskrift skrives ved brand-title
        streak_fire = "üî•" * (min(self.streak, 5)) # --------- Skriv inn overskriften til h√∏yre for brand-title inni >___< ---------------------------
        self.header_html.value = f"""
        <div class='quiz-header-widget'>
            <div class='brand-title'>QUIZ</div> 
            <div class='streak-counter'>
                <span>STREAK</span>
                <span>{self.streak}</span>
                <span>{streak_fire}</span>
            </div>
        </div>
        """

    def start_game(self):
        self.quiz_data = self.original_data.copy()
        random.shuffle(self.quiz_data)
        self.current_index = 0
        self.score = 0
        self.streak = 0
        self.render_question()

    def render_question(self):
        if self.current_index >= len(self.quiz_data):
            self.render_results()
            return

        q_data = self.quiz_data[self.current_index]
        total_q = len(self.quiz_data)
        self.update_header()
        
        # Fjern shake-effekt hvis den henger igjen
        self.card.remove_class('shake-effect')
        
        # Progresjon
        prog_pct = (self.current_index / total_q) * 100
        
        html_top = f"""
        <div class="progress-track">
            <div class="progress-bar-fill" style="width: {prog_pct}%;"></div>
        </div>
        <div class='question-text'>{q_data['question']}</div>
        """
        top_widget = widgets.HTML(html_top)

        # Knapper
        self.buttons = []
        for idx, option in enumerate(q_data['options']):
            btn = widgets.Button(description=option)
            btn.layout = widgets.Layout(width='100%', height='auto')
            btn.on_click(lambda b, i=idx, code=chr(65+idx): self.handle_answer(b, i, code))
            self.buttons.append(btn)
            
        btn_box = widgets.VBox(self.buttons)
        self.feedback_out = widgets.Output()
        
        self.content_box.children = [top_widget, btn_box, self.feedback_out]

    def handle_answer(self, clicked_btn, index, selected_code):
        q_data = self.quiz_data[self.current_index]
        correct_code = q_data['answer']
        is_correct = (selected_code == correct_code)

        if is_correct:
            self.score += 1
            self.streak += 1
            color = "#10b981"
            title = "‚ú® FANTASTISK!"
            # Spill av "riktig"-lyd (visuelt)
            clicked_btn.add_class('correct-flash')
        else:
            self.streak = 0 # Reset streak :(
            color = "#ef4444"
            title = "‚ùå AI, DEN VAR LEI"
            clicked_btn.add_class('wrong-answer')
            # RISTE EFFEKT P√Ö HELE KORTET
            self.card.add_class('shake-effect')

        self.update_header() # Oppdater streak-visning

        # Deaktiver knapper
        for btn in self.buttons:
            btn.disabled = True
            if btn.description.startswith(correct_code) and not is_correct:
                btn.add_class('fasit-highlight')

        with self.feedback_out:
            clear_output()
            feedback_html = f"""
            <div class='feedback-container' style='border-color: {color};'>
                <div style="font-weight:900; font-size:22px; color:{color}; margin-bottom:8px;">{title}</div>
                <div style="font-size:18px; color:#334155; line-height:1.5;">{q_data['explanation']}</div>
            </div>
            """
            display(HTML(feedback_html))
            
            # Neste knapp med gradient stil
            btn_next = widgets.Button(
                description="FORTSETT REISEN ‚ûú", 
                layout=widgets.Layout(width='100%', margin='25px 0 0 0') 
            )
            btn_next.style.button_color = '#1e293b' # M√∏rk
            btn_next.style.text_color = 'white'
            btn_next.style.font_weight = 'bold'
            btn_next.on_click(lambda b: self.next_q())
            display(btn_next)

    def next_q(self):
        self.current_index += 1
        self.render_question()

    def generate_confetti(self):
        html = "<div style='position:absolute; top:0; left:0; width:100%; height:100%; pointer-events:none; overflow:hidden;'>"
        colors = ['#ec4899', '#8b5cf6', '#3b82f6', '#10b981', '#f59e0b']
        for i in range(80):
            left = random.randint(0, 100)
            dur = random.uniform(2, 5)
            delay = random.uniform(0, 2)
            bg = random.choice(colors)
            size = random.randint(8, 15)
            html += f"<div class='confetti' style='left:{left}%; background:{bg}; animation-duration:{dur}s; animation-delay:{delay}s; width:{size}px; height:{size}px;'></div>"
        html += "</div>"
        return html

    def render_results(self):
        total = len(self.quiz_data)
        percent = (self.score / total) * 100
        
        confetti = self.generate_confetti() if percent >= 50 else ""
        
        if percent >= 90:
            msg = "FANTASTISK! Dette f√•r du til! üëë"       # ------legg til tilbakemelding her ------------------------------------------------------
            sub_msg = "En meget god prestasjon."
        elif percent >= 50:
            msg = "MEGET BRA LEVERT! üî•"
            sub_msg = "Du har god kontroll."
        else:
            msg = "NESTE GANG TAR DU DEN! üí™"
            sub_msg = "√òvelse gj√∏r mester."

        # Resultat HTML
        res_html = widgets.HTML(f"""
        <div style="text-align:center; padding: 20px 0; position:relative;">
            {confetti}
            <div style="font-size:16px; font-weight:700; color:#64748b; letter-spacing:2px; text-transform:uppercase;">Endelig Resultat</div>
            <div class='score-big'>{self.score}/{total}</div>
            <div style="font-size:24px; font-weight:800; color:#1e293b; margin-bottom:10px;">{msg}</div>
            <div style="font-size:18px; color:#64748b;">{sub_msg}</div>
        </div>
        """)

        # Graf
        out_chart = widgets.Output()
        with out_chart:
            fig, ax = plt.subplots(figsize=(6, 6), facecolor='none')
            sizes = [self.score, total - self.score]
            colors = ['#10b981', '#ef4444']
            wedges, texts, autotexts = ax.pie(
                sizes, labels=['Riktig', 'Feil'], colors=colors, 
                autopct='%1.0f%%', startangle=90, 
                textprops={'color':"#334155", 'fontsize': 14, 'weight':'bold'},
                explode=(0.05, 0), shadow=True
            )
            plt.setp(autotexts, size=18, weight="bold", color="white")
            plt.close(fig)
            display(fig)
            
        chart_box = widgets.HBox([out_chart], layout=widgets.Layout(justify_content='center'))
        
        btn_restart = widgets.Button(
            description="SPILL IGJEN ‚ö°",
            layout=widgets.Layout(width='100%', height='70px', margin='40px 0 0 0')
        )
        btn_restart.style.button_color = '#2563eb'
        btn_restart.style.text_color = 'white'
        btn_restart.style.font_weight = '900'
        btn_restart.style.font_size = '20px'
        btn_restart.on_click(lambda b: self.start_game())
        
        self.content_box.children = [res_html, chart_box, btn_restart]
        self.header_html.value = "" # Skjul header streak p√• resultat

# --- LAUNCH ---
app = UberQuiz(original_quiz_data)
```
